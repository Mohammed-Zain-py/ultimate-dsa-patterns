# Binary Search Problem 15 — Aggressive Cows

## Problem Statement

Given positions of stalls
and number of cows,

place cows in stalls

such that

minimum distance
between any two cows

is maximized.

Return that maximum distance.

# Problem Explanation

Example 1:

Input:

stalls = [1,2,4,8,9]

cows = 3


Output:

3


Explanation:

Place cows at:

1

4

8


Minimum distance:

3



Example 2:

Input:

stalls = [1,2,8,12]

cows = 2


Output:

11


Explanation:

Place cows:

1

12

# Brute Force Approach

Try every possible distance.

Check:

can cows
be placed?

```python
def aggressive_cows_bruteforce(stalls, cows):
    stalls.sort()

    for distance in range(1, stalls[-1] - stalls[0] + 1):
        placed = 1
        last = stalls[0]

        for stall in stalls[1:]:
            if stall - last >= distance:
                placed += 1
                last = stall

        if placed < cows:
            return distance - 1

    return stalls[-1] - stalls[0]
```

# Complexity of Brute Force

Time Complexity:

O(n × max_distance)

# Optimal Idea

Search:

minimum distance

If distance works,

try larger.

Otherwise,

reduce distance.

# Dry Run

stalls = [1,2,4,8,9]

cows = 3


distance = 4


Impossible


Try smaller


distance = 3


Possible

```python
class Solution:
    def aggressiveCows(self, stalls, cows):
        stalls.sort()

        left = 1
        right = stalls[-1] - stalls[0]

        answer = 0

        while left <= right:
            mid = left + (right - left) // 2

            placed = 1
            last = stalls[0]

            for stall in stalls[1:]:
                if stall - last >= mid:
                    placed += 1
                    last = stall

            if placed >= cows:
                answer = mid
                left = mid + 1

            else:
                right = mid - 1

        return answer
```

# Complexity Comparison

Brute Force:

Time → O(n × max_distance)

Space → O(1)


Optimal:

Time → O(n log(max_distance))

Space → O(1)

# Important Observation

Search:

distance

NOT positions.
