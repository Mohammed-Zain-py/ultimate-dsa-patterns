# Binary Search Problem 17 — Book Allocation Problem

## Problem Statement

Given an array `pages`

where pages[i]
represents pages in a book,

and `students`,

allocate books
such that:

each student receives
continuous books.

Return:

minimum possible value

of the maximum pages assigned.

# Problem Explanation

Example 1:

Input:

pages = [12,34,67,90]

students = 2


Output:

113


Explanation:

Student 1:

12 + 34 + 67


Student 2:

90


Maximum:

113



Example 2:

Input:

pages = [10,20,30,40]

students = 2


Output:

60

# Brute Force Approach

Try every possible
page limit.

Check:

can books
be allocated?

```python
def allocate_books_bruteforce(pages, students):
    if students > len(pages):
        return -1

    for limit in range(max(pages), sum(pages) + 1):
        used = 1
        current = 0

        for page in pages:
            if current + page > limit:
                used += 1
                current = 0

            current += page

        if used <= students:
            return limit
```

# Complexity of Brute Force

Time Complexity:

O(n × sum(pages))

# Optimal Idea

Search:

maximum pages

If allocation works,

try smaller.

Otherwise,

increase pages.

# Dry Run

pages = [12,34,67,90]

students = 2


limit = 120


Possible


Try smaller


Answer:

113

```python
class Solution:
    def findPages(self, pages, students):
        if students > len(pages):
            return -1

        left = max(pages)
        right = sum(pages)

        answer = right

        while left <= right:
            mid = left + (right - left) // 2

            used = 1
            current = 0

            for page in pages:
                if current + page > mid:
                    used += 1
                    current = 0

                current += page

            if used <= students:
                answer = mid
                right = mid - 1

            else:
                left = mid + 1

        return answer
```

# Complexity Comparison

Brute Force:

Time → O(n × sum(pages))

Space → O(1)


Optimal:

Time → O(n log(sum(pages)))

Space → O(1)
