# Binary Search Problem 3 — Binary Search

## Problem Statement

Given a sorted array `nums`
and a target value `target`,

return the index of target.

If target does not exist,
return `-1`.

# Problem Explanation

Example 1:

Input:

nums = [-1,0,3,5,9,12]

target = 9


Output:

4


Explanation:

Element 9 exists at index 4.



Example 2:

Input:

nums = [-1,0,3,5,9,12]

target = 2


Output:

-1


Explanation:

Target does not exist.

# Brute Force Approach

Traverse the entire array.

If target is found,

return its index.

```python
def search_bruteforce(nums, target):
    for i in range(len(nums)):
        if nums[i] == target:
            return i

    return -1


print(
    search_bruteforce(
        [-1,0,3,5,9,12],
        9
    )
)
```

# Complexity of Brute Force

Time Complexity:

O(n)

Because every element
may be checked.

# Optimal Idea — Binary Search

Check middle.

If target is smaller,

discard right half.

If target is larger,

discard left half.

# Dry Run

nums = [-1,0,3,5,9,12]

target = 9


left = 0

right = 5


mid = 2

nums[2] = 3


Move right


mid = 4


Found.

```python
class Solution:
    def search(self, nums, target):
        left = 0
        right = len(nums) - 1

        while left <= right:
            mid = left + (
                right - left
            ) // 2

            if nums[mid] == target:
                return mid

            elif nums[mid] < target:
                left = mid + 1

            else:
                right = mid - 1

        return -1
```

# Complexity Comparison

Brute Force:

Time → O(n)

Space → O(1)


Optimal:

Time → O(log n)

Space → O(1)

# Important Note

Binary Search works because:

every iteration removes
half of the search space.
