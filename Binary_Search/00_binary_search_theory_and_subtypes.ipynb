# Binary Search — Theory and Subtypes

Binary Search is one of the most important
DSA patterns for interviews.

It works by repeatedly reducing
the search space by half.

# Core Requirement

Binary Search works when:

search space is sorted
OR

the answer space is monotonic.

# What is Monotonic?

A condition changes only once.

Example:

FFFFFTTTTT

Once True starts,
it remains True.

# Why Binary Search is Powerful

Normal search:
O(n)

Binary Search:
O(log n)

Huge optimization.

# Basic Binary Search Idea

1. find middle
2. compare
3. discard half
4. continue

```python
nums = [1, 2, 3, 4, 5, 6, 7]
target = 5

left = 0
right = len(nums) - 1

while left <= right:
    mid = (left + right) // 2

    if nums[mid] == target:
        print("found at index", mid)
        break

    elif nums[mid] < target:
        left = mid + 1

    else:
        right = mid - 1

if left > right:
    print("not found")
```
