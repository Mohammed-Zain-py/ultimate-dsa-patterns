# Binary Search Problem 5 — Upper Bound

## Problem Statement

Given a sorted array `nums`
and a target `x`,

return the first index
such that:

nums[index] > x

If no such index exists,

return length of array.

# Problem Explanation

Example 1:

Input:

nums = [1,2,2,3,5]

x = 2


Output:

3


Explanation:

Index 3 contains value 3.

This is the first value
strictly greater than 2.



Example 2:

Input:

nums = [1,3,5]

x = 5


Output:

3


Explanation:

No value exists
greater than 5.

Return array length.

# Brute Force Approach

Traverse array.

Return first index
where:

nums[i] > x

```python
def upper_bound_bruteforce(nums, x):
    for i in range(len(nums)):
        if nums[i] > x:
            return i

    return len(nums)


print(upper_bound_bruteforce([1,2,2,3,5],2))
```

# Complexity of Brute Force

Time Complexity:

O(n)

Because every element
may be checked.

# Optimal Idea — Store Candidate

When:

nums[mid] > x

mid becomes
possible answer.

Store it.

Continue searching left.

# Dry Run

nums = [1,2,2,3,5]

x = 2


left = 0

right = 4


mid = 2

nums[mid]=2


Move right


mid = 3


answer = 3

```python
class Solution:
    def upperBound(self, nums, x):
        left = 0
        right = len(nums) - 1

        answer = len(nums)

        while left <= right:
            mid = left + (right - left) // 2

            if nums[mid] > x:
                answer = mid
                right = mid - 1

            else:
                left = mid + 1

        return answer
```

# Complexity Comparison

Brute Force:

Time → O(n)

Space → O(1)


Optimal:

Time → O(log n)

Space → O(1)

# Lower Bound vs Upper Bound

Lower Bound:

first ≥ x


Upper Bound:

first > x

# Pattern Takeaway

Upper Bound is useful for:

- frequency counting
- range search
- first and last occurrence
