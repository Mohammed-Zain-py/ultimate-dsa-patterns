# Binary Search Problem 13 — Koko Eating Bananas

## Problem Statement

Koko loves bananas.

Given an array `piles`
where piles[i] represents bananas,

and an integer `h`
(hours available),

return the minimum eating speed
needed to finish all bananas.

Koko eats:

k bananas per hour.

# Problem Explanation

Example 1:

Input:

piles = [3,6,7,11]

h = 8


Output:

4


Explanation:

Speed:

4

Hours:

1 + 2 + 2 + 3 = 8



Example 2:

Input:

piles = [30,11,23,4,20]

h = 5


Output:

30


Explanation:

Need maximum speed.

# Brute Force Approach

Try every possible speed.

Check:

can all bananas
finish within h?

```python
import math


def koko_bruteforce(piles, h):
    for speed in range(1, max(piles) + 1):
        hours = 0

        for pile in piles:
            hours += math.ceil(pile / speed)

        if hours <= h:
            return speed
```

# Complexity of Brute Force

Time Complexity:

O(max(piles) × n)

# Optimal Idea

Search:

speed

NOT answer directly.

If speed works,

try smaller speed.

Otherwise,

increase speed.

# Dry Run

piles = [3,6,7,11]

h = 8


speed = 6


Possible


Try smaller


speed = 4


Still possible


Answer = 4

```python
class Solution:
    def minEatingSpeed(self, piles, h):
        left = 1
        right = max(piles)

        answer = right

        while left <= right:
            mid = left + (right - left) // 2

            hours = 0

            for pile in piles:
                hours += (pile + mid - 1) // mid

            if hours <= h:
                answer = mid
                right = mid - 1

            else:
                left = mid + 1

        return answer
```

# Complexity Comparison

Brute Force:

Time → O(max(piles) × n)

Space → O(1)


Optimal:

Time → O(n log(max(piles)))

Space → O(1)

# Important Observation

You are NOT searching:

bananas

You are searching:

speed
