# Binary Search Problem 22 — Find Kth Missing Positive Number

## Problem Statement

Given a sorted array `arr`
of positive integers

and an integer `k`,

return the kth missing positive number.

# Problem Explanation

Example 1:

Input:

arr = [2,3,4,7,11]

k = 5


Output:

9


Explanation:

Missing numbers:

1,5,6,8,9,...

5th missing:

9



Example 2:

Input:

arr = [1,2,3,4]

k = 2


Output:

6


Explanation:

Missing:

5,6,...

Answer:

6

# Brute Force Approach

Traverse numbers.

Reduce k

whenever

a number is missing.

```python
def kth_missing_bruteforce(arr, k):
    current = 1
    index = 0

    while k > 0:
        if index < len(arr) and arr[index] == current:
            index += 1
        else:
            k -= 1

        current += 1

    return current - 1


print(kth_missing_bruteforce([2, 3, 4, 7, 11], 5))
```

# Complexity of Brute Force

Time Complexity:

O(answer)

# Optimal Idea

At index i:

missing numbers =

arr[i] − (i + 1)

Use Binary Search

to find

first position

where missing ≥ k

# Dry Run

arr = [2,3,4,7,11]

k = 5


index = 4


missing =

11 − 5


= 6


Answer:

9

```python
class Solution:
    def findKthPositive(self, arr, k):
        left = 0
        right = len(arr) - 1

        while left <= right:
            mid = left + (right - left) // 2

            missing = arr[mid] - (mid + 1)

            if missing < k:
                left = mid + 1

            else:
                right = mid - 1

        return left + k
```

# Complexity Comparison

Brute Force:

Time → O(answer)

Space → O(1)


Optimal:

Time → O(log n)

Space → O(1)

# Important Observation

We search:

count of missing

NOT values.
