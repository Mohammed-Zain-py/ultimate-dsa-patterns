# Binary Search Problem 10 — Find Minimum in Rotated Sorted Array

## Problem Statement

Given a rotated sorted array `nums`,

return the minimum element.

You must solve it in:

O(log n)

# Problem Explanation

Example 1:

Input:

nums = [3,4,5,1,2]


Output:

1


Explanation:

Array was rotated.

Minimum becomes 1.



Example 2:

Input:

nums = [4,5,6,7,0,1,2]


Output:

0


Explanation:

Smallest element is 0.

# Brute Force Approach

Traverse array.

Track minimum value.

```python
def find_min_bruteforce(nums):
    minimum = nums[0]

    for num in nums:
        minimum = min(minimum, num)

    return minimum


print(find_min_bruteforce([3, 4, 5, 1, 2]))
```

# Complexity of Brute Force

Time Complexity:

O(n)

# Optimal Idea

At every step:

one half is sorted.

Minimum always exists
inside the unsorted side.

Keep updating answer.

# Dry Run

nums = [4,5,6,7,0,1,2]


mid = 3


Left sorted


Minimum can still exist
on right


Move right


Found 0

```python
class Solution:
    def findMin(self, nums):
        left = 0
        right = len(nums) - 1

        answer = nums[0]

        while left <= right:
            if nums[left] <= nums[right]:
                answer = min(answer, nums[left])
                break

            mid = left + (right - left) // 2

            answer = min(answer, nums[mid])

            if nums[left] <= nums[mid]:
                left = mid + 1

            else:
                right = mid - 1

        return answer
```

# Complexity Comparison

Brute Force:

Time → O(n)

Space → O(1)


Optimal:

Time → O(log n)

Space → O(1)

# Important Observation

Rotated arrays
still preserve

partial ordering.

# Pattern Takeaway

If one side is sorted,

search the other side.
