# Binary Search Problem 6 — First and Last Occurrence

## Problem Statement

Given a sorted array `nums`
and a target `target`,

return:

[first occurrence, last occurrence]

If target does not exist,

return:

[-1, -1]

# Problem Explanation

Example 1:

Input:

nums = [5,7,7,8,8,10]

target = 8


Output:

[3,4]


Explanation:

First occurrence:

index 3

Last occurrence:

index 4



Example 2:

Input:

nums = [5,7,7,8,8,10]

target = 6


Output:

[-1,-1]


Explanation:

Target does not exist.

# Brute Force Approach

Traverse array.

Store:

first appearance

last appearance

```python
def search_range_bruteforce(nums, target):
    first = -1
    last = -1

    for i in range(len(nums)):
        if nums[i] == target:
            if first == -1:
                first = i

            last = i

    return [first, last]


print(search_range_bruteforce(
    [5, 7, 7, 8, 8, 10],
    8
))
```

# Complexity of Brute Force

Time Complexity:

O(n)

Because the entire array
may be visited.

# Optimal Idea

Use:

Lower Bound

to find first occurrence.


Use:

Upper Bound − 1

to find last occurrence.

# Dry Run

nums = [5,7,7,8,8,10]

target = 8


Lower Bound:

3


Upper Bound:

5


Last:

5 − 1 = 4

```python
class Solution:
    def searchRange(self, nums, target):
        first = -1
        last = -1

        left = 0
        right = len(nums) - 1

        while left <= right:
            mid = left + (right - left) // 2

            if nums[mid] >= target:
                if nums[mid] == target:
                    first = mid

                right = mid - 1

            else:
                left = mid + 1

        left = 0
        right = len(nums) - 1

        while left <= right:
            mid = left + (right - left) // 2

            if nums[mid] <= target:
                if nums[mid] == target:
                    last = mid

                left = mid + 1

            else:
                right = mid - 1

        return [first, last]
```

# Complexity Comparison

Brute Force:

Time → O(n)

Space → O(1)


Optimal:

Time → O(log n)

Space → O(1)

# Important Observation

Range problems often become:

Lower Bound

Upper Bound
