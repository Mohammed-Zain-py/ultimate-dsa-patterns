# Binary Search Problem 16 — Split Array Largest Sum

## Problem Statement

Given an array `nums`
and an integer `k`,

split the array
into exactly `k`
non-empty subarrays.

Return:

minimum possible value

of the largest subarray sum.

# Problem Explanation

Example 1:

Input:

nums = [7,2,5,10,8]

k = 2


Output:

18


Explanation:

Split:

[7,2,5]

[10,8]


Largest sum:

18



Example 2:

Input:

nums = [1,2,3,4,5]

k = 2


Output:

9


Explanation:

Split:

[1,2,3]

[4,5]

# Brute Force Approach

Try every possible
maximum sum.

Check:

can array
be split?

```python
def split_array_bruteforce(nums, k):
    for limit in range(max(nums), sum(nums) + 1):
        parts = 1
        current = 0

        for num in nums:
            if current + num > limit:
                parts += 1
                current = 0

            current += num

        if parts <= k:
            return limit
```

# Complexity of Brute Force

Time Complexity:

O(n × sum(nums))

# Optimal Idea

Search:

largest allowed sum

If splitting works,

try smaller.

Otherwise,

increase limit.

# Dry Run

nums = [7,2,5,10,8]

k = 2


limit = 21


Possible


Try smaller


Eventually:

18

```python
class Solution:
    def splitArray(self, nums, k):
        left = max(nums)
        right = sum(nums)

        answer = right

        while left <= right:
            mid = left + (right - left) // 2

            parts = 1
            current = 0

            for num in nums:
                if current + num > mid:
                    parts += 1
                    current = 0

                current += num

            if parts <= k:
                answer = mid
                right = mid - 1

            else:
                left = mid + 1

        return answer
```

# Complexity Comparison

Brute Force:

Time → O(n × sum(nums))

Space → O(1)


Optimal:

Time → O(n log(sum(nums)))

Space → O(1)

# Important Observation

Search:

largest subarray sum

NOT split positions.
