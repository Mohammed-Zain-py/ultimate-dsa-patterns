# How to Identify Binary Search Problems

Binary Search is NOT identified by the words:

"binary search"

Instead,

we identify the structure of the problem.

# The Biggest Signal

Ask:

Can I remove half of the search space
after checking one value?

If YES,

Binary Search may work.

# Signal 1 — Sorted Data

Examples:

- sorted array
- sorted matrix
- ordered values

Classic Binary Search starts here.

# Signal 2 — Monotonic Behaviour

Look for:

FFFFTTTT

or

TTTTFFFF

After crossing one point,

the answer behavior never changes.

# Signal 3 — Answer Search Space

Many problems do NOT search elements.

They search answers.

Examples:

- minimum capacity
- maximum distance
- smallest valid answer

# Example Thinking

Question:

Minimum speed to finish work.

Think:

Can speed 10 work?

Can speed 20 work?

Can speed 30 work?

Notice:

answers become monotonic.

# Common Binary Search Keywords

Watch for:

- sorted
- search efficiently
- minimize maximum
- maximize minimum
- smallest valid
- largest possible

```python
nums = [1, 3, 5, 7, 9]
target = 7

left = 0
right = len(nums) - 1

while left <= right:
    mid = (left + right) // 2

    if nums[mid] == target:
        print("Binary Search Possible")
        break

    elif nums[mid] < target:
        left = mid + 1

    else:
        right = mid - 1
```

# Common Mistakes

Mistake 1:

Using Binary Search without monotonicity


Mistake 2:

Applying Binary Search to unsorted data


Mistake 3:

Not defining search space correctly

# Recognition Checklist

Ask:

1. Can half be removed?

2. Is data sorted?

3. Is answer monotonic?

4. Can brute force become logarithmic?

# Final Takeaway

Do NOT memorize code.

Learn to identify:

WHY Binary Search works.
