# Search Space and Monotonic Functions

This notebook explains the most important concept
in advanced Binary Search.

Binary Search is not limited to arrays.

Sometimes,

we search over possible answers.

# What is Search Space?

Search Space means:

all possible values
where the answer can exist.

Examples:

Search index:
0 → n−1

Search answer:
1 → 1000000

# Problem Explanation

Example 1:

Question:

Find square root of 100.

Possible answers:

1 → 100

We search answers.


Example 2:

Question:

Minimum speed to finish work.

Possible answers:

1 → maximum speed

Again,

searching answers.

# What is Monotonic Behaviour?

Monotonic means:

once behavior changes,

it never returns.

# Example of Monotonic Pattern

Speed:

1 → False

2 → False

3 → False

4 → True

5 → True

6 → True

Binary Search finds transition point.

# Core Thinking

Do NOT ask:

Which value is answer?

Ask:

Can this value become answer?

# Recognition Formula

If:

FFFFTTTT

Find first True


OR


TTTTFFFF

Find last True

```python
def can_finish(speed):
    return speed >= 7


left = 1
right = 10

while left <= right:
    mid = (left + right) // 2

    if can_finish(mid):
        right = mid - 1

    else:
        left = mid + 1


print("answer =", left)
```

# Why This Works

Every check removes
half of the answer space.

That gives:

O(log n)
