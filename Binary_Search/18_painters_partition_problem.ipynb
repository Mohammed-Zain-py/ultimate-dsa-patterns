# Binary Search Problem 18 — Painter's Partition Problem

## Problem Statement

Given an array `boards`

where boards[i]
represents board length,

and `painters`,

allocate continuous boards.

Return:

minimum possible time

to paint all boards.

# Problem Explanation

Example 1:

Input:

boards = [10,20,30,40]

painters = 2


Output:

60


Explanation:

Painter 1:

10 + 20 + 30


Painter 2:

40


Maximum time:

60



Example 2:

Input:

boards = [5,10,30,20]

painters = 2


Output:

35

# Brute Force Approach

Try every possible
time limit.

Check:

can boards
be allocated?

```python
def painters_bruteforce(boards, painters):
    for limit in range(max(boards), sum(boards) + 1):
        used = 1
        current = 0

        for board in boards:
            if current + board > limit:
                used += 1
                current = 0

            current += board

        if used <= painters:
            return limit


print(
    painters_bruteforce(
        [10, 20, 30, 40],
        2
    )
)
```

# Complexity of Brute Force

Time Complexity:

O(n × sum(boards))

# Optimal Idea

Search:

maximum allowed time

If valid,

try smaller.

Otherwise,

increase.

# Dry Run

boards = [10,20,30,40]

painters = 2


time = 60


Possible


Try smaller


Final:

60

```python
class Solution:
    def minTime(self, boards, painters):
        left = max(boards)
        right = sum(boards)

        answer = right

        while left <= right:
            mid = left + (right - left) // 2

            used = 1
            current = 0

            for board in boards:
                if current + board > mid:
                    used += 1
                    current = 0

                current += board

            if used <= painters:
                answer = mid
                right = mid - 1

            else:
                left = mid + 1

        return answer


obj = Solution()

print(
    obj.minTime(
        [10, 20, 30, 40],
        2
    )
)
```

# Complexity Comparison

Brute Force:

Time → O(n × sum(boards))

Space → O(1)


Optimal:

Time → O(n log(sum(boards)))

Space → O(1)
