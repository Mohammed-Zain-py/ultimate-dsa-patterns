# Binary Search Problem 4 — Lower Bound

## Problem Statement

Given a sorted array `nums`
and a target `x`,

return the first index
such that:

nums[index] ≥ x

If no such index exists,

return length of array.

# Problem Explanation

Example 1:

Input:

nums = [1,2,2,3,5]

x = 2


Output:

1


Explanation:

Index 1 contains 2.

This is the first value
greater than or equal to 2.



Example 2:

Input:

nums = [1,3,5]

x = 4


Output:

2


Explanation:

nums[2]=5

This is the first value
greater than or equal to 4.

# Brute Force Approach

Traverse array.

Return the first position
where:

nums[i] ≥ x

```python
def lower_bound_bruteforce(nums, x):
    for i in range(len(nums)):
        if nums[i] >= x:
            return i

    return len(nums)


print(
    lower_bound_bruteforce(
        [1,2,2,3,5],
        2
    )
)
```

# Complexity of Brute Force

Time Complexity:

O(n)

Because traversal may
check all elements.

# Optimal Idea — Store Candidate

When:

nums[mid] ≥ x

mid can become answer.

Store it.

Then search left
for earlier positions.

# Dry Run

nums = [1,2,2,3,5]

x = 2


left = 0

right = 4


mid = 2

answer = 2


Move left


mid = 0


Move right


mid = 1


answer = 1

```python
class Solution:
    def lowerBound(self, nums, x):
        left = 0
        right = len(nums) - 1

        answer = len(nums)

        while left <= right:
            mid = left + (
                right - left
            ) // 2

            if nums[mid] >= x:
                answer = mid
                right = mid - 1

            else:
                left = mid + 1

        return answer
```

# Complexity Comparison

Brute Force:

Time → O(n)

Space → O(1)


Optimal:

Time → O(log n)

Space → O(1)

# Important Observation

Lower Bound means:

first valid position.

Not exact match.

# Pattern Takeaway

Lower Bound becomes
the foundation for:

- upper bound
- first occurrence
- insertion position
