# Binary Search Problem 11 — Find Peak Element

## Problem Statement

Given an integer array `nums`,

find a peak element.

Return its index.

A peak element is greater than its neighbors.

You may assume:

nums[-1] = -∞

nums[n] = -∞

# Problem Explanation

Example 1:

Input:

nums = [1,2,3,1]


Output:

2


Explanation:

nums[2]=3

3 is greater than neighbors.



Example 2:

Input:

nums = [1,2,1,3,5,6,4]


Output:

5


Explanation:

nums[5]=6

6 is a valid peak.

# Brute Force Approach

Traverse array.

Check whether current element
is greater than both neighbors.

```python
def find_peak_bruteforce(nums):
    n = len(nums)

    for i in range(n):
        left = float("-inf") if i == 0 else nums[i - 1]
        right = float("-inf") if i == n - 1 else nums[i + 1]

        if nums[i] > left and nums[i] > right:
            return i
```

# Complexity of Brute Force

Time Complexity:

O(n)

# Optimal Idea

Look at:

mid

If right neighbor is larger,

peak must exist
on the right.

Otherwise,

peak exists
on the left side
including mid.

# Dry Run

nums = [1,2,3,1]


mid = 1


nums[2] > nums[1]


Move right


Peak found.

```python
class Solution:
    def findPeakElement(self, nums):
        left = 0
        right = len(nums) - 1

        while left < right:
            mid = left + (right - left) // 2

            if nums[mid] < nums[mid + 1]:
                left = mid + 1

            else:
                right = mid

        return left
```

# Complexity Comparison

Brute Force:

Time → O(n)

Space → O(1)


Optimal:

Time → O(log n)

Space → O(1)

# Important Observation

You do NOT need
full sorting.

Use direction.

Move toward rise.
