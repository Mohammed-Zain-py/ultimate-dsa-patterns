# Binary Search Problem 12 — Sqrt(x)

## Problem Statement

Given a non-negative integer `x`,

return the integer square root.

The returned value should be:

⌊√x⌋

Do NOT use built-in exponent functions.

# Problem Explanation

Example 1:

Input:

x = 8


Output:

2


Explanation:

√8 = 2.828...

Floor value becomes 2.



Example 2:

Input:

x = 16


Output:

4


Explanation:

√16 = 4

# Brute Force Approach

Start from:

1

Keep checking:

number × number

until value exceeds x.

```python
def sqrt_bruteforce(x):
    answer = 0

    for num in range(x + 1):
        if num * num <= x:
            answer = num
        else:
            break

    return answer


print(sqrt_bruteforce(8))
```

# Complexity of Brute Force

Time Complexity:

O(√x)

# Optimal Idea

Search possible answers.

Range:

0 → x

If:

mid × mid ≤ x

store answer

move right.

# Dry Run

x = 8


left = 0

right = 8


mid = 4


16 > 8


Move left


mid = 2


Store answer


Final:

2

```python
class Solution:
    def mySqrt(self, x):
        left = 0
        right = x

        answer = 0

        while left <= right:
            mid = left + (right - left) // 2

            square = mid * mid

            if square <= x:
                answer = mid
                left = mid + 1

            else:
                right = mid - 1

        return answer
```

# Complexity Comparison

Brute Force:

Time → O(√x)

Space → O(1)


Optimal:

Time → O(log x)

Space → O(1)

# Important Observation

This is NOT searching
array elements.

This is:

Binary Search on Answers.

# Pattern Takeaway

Whenever answer exists
inside a range,

consider Binary Search.
