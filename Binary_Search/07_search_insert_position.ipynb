# Binary Search Problem 7 — Search Insert Position

## Problem Statement

Given a sorted array `nums`
and a target `target`,

return the index if found.

Otherwise,

return the index
where it should be inserted.

# Problem Explanation

Example 1:

Input:

nums = [1,3,5,6]

target = 5


Output:

2


Explanation:

Target already exists.



Example 2:

Input:

nums = [1,3,5,6]

target = 2


Output:

1


Explanation:

Insert 2 at index 1.



Example 3:

Input:

nums = [1,3,5,6]

target = 7


Output:

4


Explanation:

Insert at end.

# Brute Force Approach

Traverse array.

Return first position
where:

nums[i] ≥ target

```python
def search_insert_bruteforce(nums, target):
    for i in range(len(nums)):
        if nums[i] >= target:
            return i

    return len(nums)


print(search_insert_bruteforce([1, 3, 5, 6], 2))
```

# Complexity of Brute Force

Time Complexity:

O(n)

# Optimal Idea

This is exactly:

Lower Bound

Return first index
where:

nums[index] ≥ target

# Dry Run

nums = [1,3,5,6]

target = 2


mid = 1

nums[1]=3


Store answer

Move left


Final answer:

1

```python
class Solution:
    def searchInsert(self, nums, target):
        left = 0
        right = len(nums) - 1

        answer = len(nums)

        while left <= right:
            mid = left + (right - left) // 2

            if nums[mid] >= target:
                answer = mid
                right = mid - 1

            else:
                left = mid + 1

        return answer
```

# Complexity Comparison

Brute Force:

Time → O(n)

Space → O(1)


Optimal:

Time → O(log n)

Space → O(1)

# Important Observation

Search Insert Position = Lower Bound
