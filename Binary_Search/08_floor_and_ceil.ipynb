# Binary Search Problem 8 — Floor and Ceil

## Problem Statement

Given a sorted array `nums`
and a target `x`,

find:

Floor:
largest value ≤ x

Ceil:
smallest value ≥ x

Return:

[floor, ceil]

If not present,

return -1 for that value.

# Problem Explanation

Example 1:

Input:

nums = [1,2,4,6,10]

x = 5


Output:

[4,6]


Explanation:

Floor:

largest value ≤ 5 → 4

Ceil:

smallest value ≥ 5 → 6



Example 2:

Input:

nums = [1,2,4,6]

x = 0


Output:

[-1,1]


Explanation:

No floor exists.

Ceil becomes 1.

# Brute Force Approach

Traverse array.

Track:

largest valid floor

smallest valid ceil

```python
def floor_ceil_bruteforce(nums, x):
    floor = -1
    ceil = -1

    for num in nums:
        if num <= x:
            floor = num

        if num >= x and ceil == -1:
            ceil = num

    return [floor, ceil]


print(floor_ceil_bruteforce([1, 2, 4, 6, 10], 5))
```

# Complexity of Brute Force

Time Complexity:

O(n)

# Optimal Idea

Ceil:

Lower Bound


Floor:

Previous element
before Lower Bound

# Dry Run

nums = [1,2,4,6,10]

x = 5


Lower Bound:

index = 3


Ceil:

6


Floor:

index 2

value = 4

```python
class Solution:
    def getFloorAndCeil(self, nums, x):
        left = 0
        right = len(nums) - 1

        lower = len(nums)

        while left <= right:
            mid = left + (right - left) // 2

            if nums[mid] >= x:
                lower = mid
                right = mid - 1

            else:
                left = mid + 1

        ceil = nums[lower] if lower < len(nums) else -1

        if lower < len(nums) and nums[lower] == x:
            floor = nums[lower]

        elif lower > 0:
            floor = nums[lower - 1]

        else:
            floor = -1

        return [floor, ceil]
```

# Complexity Comparison

Brute Force:

Time → O(n)

Space → O(1)


Optimal:

Time → O(log n)

Space → O(1)

# Important Observation

Floor:

right side answer


Ceil:

left side answer

# Pattern Takeaway

Floor and Ceil

teach answer positioning

inside Binary Search.
