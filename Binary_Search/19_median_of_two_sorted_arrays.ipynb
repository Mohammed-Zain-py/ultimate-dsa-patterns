# Binary Search Problem 19 — Median of Two Sorted Arrays

## Problem Statement

Given two sorted arrays
`nums1`
and
`nums2`,

return the median
of both arrays.

Required complexity:

O(log(min(n,m)))

# Problem Explanation

Example 1:

Input:

nums1 = [1,3]

nums2 = [2]


Output:

2


Explanation:

Merged:

[1,2,3]

Median:

2



Example 2:

Input:

nums1 = [1,2]

nums2 = [3,4]


Output:

2.5


Explanation:

Merged:

[1,2,3,4]

Median:

(2+3)/2

# Brute Force Approach

Merge arrays.

Find median
using final array.

```python
def median_bruteforce(nums1, nums2):
    merged = sorted(nums1 + nums2)

    n = len(merged)

    if n % 2:
        return merged[n // 2]

    return (
        merged[n // 2 - 1]
        + merged[n // 2]
    ) / 2


print(
    median_bruteforce(
        [1, 2],
        [3, 4]
    )
)
```

# Complexity of Brute Force

Time Complexity:

O((n+m) log(n+m))

# Optimal Idea

Partition both arrays.

Left side:

contains half elements.

Median exists

between boundaries.

# Dry Run

nums1 = [1,2]

nums2 = [3,4]


Partition


Left:

[1,2]


Right:

[3,4]


Median:

2.5

```python
class Solution:
    def findMedianSortedArrays(self, nums1, nums2):
        if len(nums1) > len(nums2):
            nums1, nums2 = nums2, nums1

        n1 = len(nums1)
        n2 = len(nums2)

        left = 0
        right = n1

        while left <= right:
            cut1 = left + (right - left) // 2
            cut2 = (n1 + n2 + 1) // 2 - cut1

            l1 = float("-inf")
            l2 = float("-inf")

            r1 = float("inf")
            r2 = float("inf")

            if cut1 > 0:
                l1 = nums1[cut1 - 1]

            if cut2 > 0:
                l2 = nums2[cut2 - 1]

            if cut1 < n1:
                r1 = nums1[cut1]

            if cut2 < n2:
                r2 = nums2[cut2]

            if l1 <= r2 and l2 <= r1:
                if (n1 + n2) % 2:
                    return max(l1, l2)

                return (
                    max(l1, l2)
                    + min(r1, r2)
                ) / 2

            elif l1 > r2:
                right = cut1 - 1

            else:
                left = cut1 + 1


obj = Solution()

print(obj.findMedianSortedArrays([1, 2],[3, 4]))
```

# Complexity Comparison

Brute Force:

Time → O((n+m)log(n+m))

Space → O(n+m)


Optimal:

Time → O(log(min(n,m)))

Space → O(1)

# Important Observation

Do NOT merge.

Search

partition positions.

# Pattern Takeaway

Advanced Binary Search

can operate

on partitions.
