# Two Pointer Problem 29 — Longest Palindrome Length from Characters

## Problem Statement

Given a string `s`, return the length of the longest palindrome that can be built using its characters.

Example:

s = "abccccdd"

Output:
7

Explanation:
One possible palindrome is "dccaccd"

# Brute Force Approach

Count the frequency of each character.

Use all even counts completely.

For odd counts:
- use count - 1
- allow only one odd character in the center

```python
def longest_palindrome_length_bruteforce(s):
    freq = {}

    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1

    length = 0
    odd_found = False

    for count in freq.values():
        if count % 2 == 0:
            length += count
        else:
            length += count - 1
            odd_found = True

    if odd_found:
        length += 1

    return length


print(longest_palindrome_length_bruteforce("abccccdd"))
```

# Complexity of Brute Force

Time Complexity: O(n)

Space Complexity: O(n)

This is already efficient.

# Optimal Idea — Frequency + Center Logic

A palindrome can use:

- all pairs completely
- at most one unpaired character in center

# Key Observation

Every even count contributes fully.

Every odd count contributes:
count - 1

And one extra center character.

# Frequency Dry Run

s = "abccccdd"

| char | count | contribution |
|------|-------|--------------|
| a | 1 | 0 |
| b | 1 | 0 |
| c | 4 | 4 |
| d | 2 | 2 |

```python
class Solution:
    def longestPalindrome(self, s):
        freq = {}

        for ch in s:
            freq[ch] = freq.get(ch, 0) + 1

        length = 0
        odd_found = False

        for count in freq.values():
            length += (count // 2) * 2

            if count % 2 == 1:
                odd_found = True

        if odd_found:
            length += 1

        return length
```

# Complexity Comparison

Time → O(n)

Space → O(n)
