# Two Pointer Problem 12 — Compare Version Numbers

## Problem Statement

Given two version strings, compare them.

Return:
- `1` if version1 > version2
- `-1` if version1 < version2
- `0` if equal

Example:

version1 = "1.01"
version2 = "1.001"

Output: 0

# Brute Force Approach

Split both strings by `.`

Then compare each part after converting to integer.

```python
def compare_versions_bruteforce(version1, version2):
    v1 = version1.split(".")
    v2 = version2.split(".")

    n = max(len(v1), len(v2))

    for i in range(n):
        num1 = int(v1[i]) if i < len(v1) else 0
        num2 = int(v2[i]) if i < len(v2) else 0

        if num1 > num2:
            return 1
        elif num1 < num2:
            return -1

    return 0


print(compare_versions_bruteforce("1.01", "1.001"))
```

# Complexity of Brute Force

Time Complexity: O(n)

Space Complexity: O(n)

Because split creates extra lists.

# Optimal Idea — Two Pointers

Use two pointers to traverse both strings.

Extract each number between dots.

Compare chunk by chunk.

This avoids creating extra arrays.

# Pointer Movement Dry Run

version1 = "1.01"
version2 = "1.001"

| i | j | chunk1 | chunk2 | action |
|---|---|--------|--------|--------|
| 0 | 0 | 1 | 1 | continue |
| 2 | 2 | 1 | 1 | equal |

```python
class Solution:
    def compareVersion(self, version1, version2):
        i = 0
        j = 0

        while i < len(version1) or j < len(version2):
            num1 = 0
            num2 = 0

            while i < len(version1) and version1[i] != ".":
                num1 = num1 * 10 + int(version1[i])
                i += 1

            while j < len(version2) and version2[j] != ".":
                num2 = num2 * 10 + int(version2[j])
                j += 1

            if num1 > num2:
                return 1
            elif num1 < num2:
                return -1

            i += 1
            j += 1

        return 0
```

# Complexity Comparison

Brute Force:
- Time → O(n)
- Space → O(n)

Optimal:
- Time → O(n)
- Space → O(1)
