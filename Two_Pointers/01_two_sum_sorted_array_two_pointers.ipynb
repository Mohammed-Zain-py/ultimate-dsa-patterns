# Two Pointer Problem 1 — Two Sum II (Sorted Array)

## Problem Statement

Given a sorted array of integers and a target value, find two numbers such that they add up to the target.

Return their indexes.

Example:

nums = [2, 7, 11, 15]  
target = 9

Output: [0, 1]

# Brute Force Idea

Check every pair using nested loops.

This gives time complexity **O(n²)**.

But since the array is sorted, we can optimize using Two Pointers.

```python
arr = [2, 4, 7, 11, 15]
target = 15

for i in range(len(arr)):
    for j in range(i + 1, len(arr)):
        if arr[i] + arr[j] == target:
            print("Pair found:", arr[i], arr[j])
```

This works, but it is not optimal.

Since the array is already sorted, we can do better.

# Optimal Idea — Opposite Ends

Take one pointer from the left and one from the right.

- if sum is too small → move left
- if sum is too large → move right
- if sum matches → answer found

# Pointer Movement Dry Run

nums = [2, 7, 11, 15]  
target = 9

| left | right | nums[left] | nums[right] | sum | action |
|------|-------|------------|-------------|-----|--------|
| 0 | 3 | 2 | 15 | 17 | move right leftward |
| 0 | 2 | 2 | 11 | 13 | move right leftward |
| 0 | 1 | 2 | 7 | 9 | answer found |

```python
def two_sum_sorted(nums, target):
    left = 0
    right = len(nums) - 1

    while left < right:
        current_sum = nums[left] + nums[right]

        if current_sum == target:
            return [left, right]

        elif current_sum < target:
            left += 1

        else:
            right -= 1

    return [-1, -1]


nums = [2, 7, 11, 15]
target = 9

print(two_sum_sorted(nums, target))
```

# Why this works

Because the array is sorted:

- increasing left increases sum
- decreasing right decreases sum

This is the key observation.

# Complexity Analysis

Time Complexity: O(n)

Each pointer moves at most n times.

Space Complexity: O(1)

# Pattern Takeaway

This is the classic **Opposite Ends Two Pointer** pattern.

Whenever input is sorted and a pair is required, think of this first.
