# Two Pointer Problem 19 — Remove Duplicates from Sorted Array

## Problem Statement

Given a sorted array `nums`, remove duplicates in-place such that each unique element appears only once.

Return the number of unique elements.

Example:

nums = [1, 1, 2]

Output:
2

Modified array:
[1, 2, _]

# Brute Force Approach

Create a new array.

Add elements only if they are different from the last added element.

This works but uses extra space.

```python
def remove_duplicates_bruteforce(nums):
    if not nums:
        return 0, []

    result = [nums[0]]

    for i in range(1, len(nums)):
        if nums[i] != nums[i - 1]:
            result.append(nums[i])

    return len(result), result


print(remove_duplicates_bruteforce([1, 1, 2]))
```

# Complexity of Brute Force

Time Complexity: O(n)

Space Complexity: O(n)

But the problem requires in-place modification.

# Optimal Idea — Slow and Fast Pointers

Use:

- `fast` → scans all elements
- `slow` → points to next unique position

# Key Observation

Because the array is sorted,

duplicates will always be adjacent.

This makes comparison easy.

# Pointer Movement Dry Run

nums = [1, 1, 2]

| slow | fast | nums[slow-1] | nums[fast] | action |
|------|------|--------------|------------|--------|
| 1 | 1 | 1 | 1 | skip |
| 1 | 2 | 1 | 2 | place at slow |

```python
class Solution:
    def removeDuplicates(self, nums):
        if not nums:
            return 0

        slow = 1

        for fast in range(1, len(nums)):
            if nums[fast] != nums[fast - 1]:
                nums[slow] = nums[fast]
                slow += 1

        return slow
```

# Complexity Comparison

Brute Force:
- Time → O(n)
- Space → O(n)

Optimal:
- Time → O(n)
- Space → O(1)
