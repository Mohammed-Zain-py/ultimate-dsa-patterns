# Two Pointer Problem 22 — Merge Sorted Array

## Problem Statement

You are given two sorted arrays.

`nums1` has enough space at the end to hold all elements of `nums2`.

Merge `nums2` into `nums1` in sorted order.

Example:

nums1 = [1, 2, 3, 0, 0, 0]
m = 3

nums2 = [2, 5, 6]
n = 3

Output:
[1, 2, 2, 3, 5, 6]

# Brute Force Approach

Copy elements of `nums2` into `nums1`.

Then sort the entire array.

This works but is not the most optimal approach.

```python
def merge_sorted_array_bruteforce(nums1, m, nums2, n):
    nums1[m:] = nums2
    nums1.sort()
    return nums1


print(merge_sorted_array_bruteforce([1,2,3,0,0,0], 3, [2,5,6], 3))
```

# Complexity of Brute Force

Time Complexity: O((m+n) log(m+n))

Space Complexity: O(1)

Can we do better?

Yes — O(m+n)

# Optimal Idea — Two Pointers from the End

Since both arrays are already sorted,

start comparing from the end.

This avoids overwriting useful values in `nums1`.

# Key Observation

Largest remaining value must be at the end of either:

- nums1 valid part
- nums2

# Pointer Movement Dry Run

nums1 = [1,2,3,0,0,0]
nums2 = [2,5,6]

| i | j | k | chosen |
|---|---|---|--------|
| 2 | 2 | 5 | 6 |
| 2 | 1 | 4 | 5 |
| 2 | 0 | 3 | 3 |
| 1 | 0 | 2 | 2 |

```python
class Solution:
    def merge(self, nums1, m, nums2, n):
        i = m - 1
        j = n - 1
        k = m + n - 1

        while i >= 0 and j >= 0:
            if nums1[i] > nums2[j]:
                nums1[k] = nums1[i]
                i -= 1
            else:
                nums1[k] = nums2[j]
                j -= 1

            k -= 1

        while j >= 0:
            nums1[k] = nums2[j]
            j -= 1
            k -= 1

        return nums1
```

# Complexity Comparison

Brute Force:
- Time → O((m+n) log(m+n))

Optimal:
- Time → O(m+n)
- Space → O(1)
