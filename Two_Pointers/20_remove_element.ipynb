# Two Pointer Problem 20 — Remove Element

## Problem Statement

Given an array `nums` and a value `val`, remove all occurrences of `val` in-place.

Return the number of remaining elements.

Example:

nums = [3, 2, 2, 3]
val = 3

Output:
2

Modified array:
[2, 2, _, _]

# Brute Force Approach

Create a new array.

Copy only elements that are not equal to `val`.

This works but uses extra space.

```python
def remove_element_bruteforce(nums, val):
    result = []

    for num in nums:
        if num != val:
            result.append(num)

    return len(result), result


print(remove_element_bruteforce([3, 2, 2, 3], 3))
```

# Complexity of Brute Force

Time Complexity: O(n)

Space Complexity: O(n)

But the question asks for in-place modification.

# Optimal Idea — Slow and Fast Pointers

Use:

- `fast` → scans every element
- `slow` → points to next valid position

# Key Observation

Whenever `nums[fast]` is not equal to `val`,

copy it to `nums[slow]`

Then move `slow`.

# Pointer Movement Dry Run

nums = [3, 2, 2, 3]
val = 3

| slow | fast | nums[fast] | action |
|------|------|------------|--------|
| 0 | 0 | 3 | skip |
| 0 | 1 | 2 | copy to slow |
| 1 | 2 | 2 | copy to slow |
| 2 | 3 | 3 | skip |

```python
class Solution:
    def removeElement(self, nums, val):
        slow = 0

        for fast in range(len(nums)):
            if nums[fast] != val:
                nums[slow] = nums[fast]
                slow += 1

        return slow
```

# Complexity Comparison

Brute Force:
- Time → O(n)
- Space → O(n)

Optimal:
- Time → O(n)
- Space → O(1)
