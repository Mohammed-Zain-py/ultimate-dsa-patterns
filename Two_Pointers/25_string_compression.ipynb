# Two Pointer Problem 25 — String Compression

## Problem Statement

Given an array of characters `chars`, compress it in-place.

For repeated consecutive characters, store the character followed by its count.

Example:

chars = ["a","a","b","b","c","c","c"]

Output length:
6

Modified array:
["a","2","b","2","c","3"]

# Brute Force Approach

Create a new list.

Count consecutive characters.

Append character and count.

This works but uses extra space.

```python
def string_compression_bruteforce(chars):
    result = []
    i = 0

    while i < len(chars):
        current_char = chars[i]
        count = 0

        while i < len(chars) and chars[i] == current_char:
            count += 1
            i += 1

        result.append(current_char)

        if count > 1:
            for digit in str(count):
                result.append(digit)

    return len(result), result


print(string_compression_bruteforce(["a","a","b","b","c","c","c"]))
```

# Complexity of Brute Force

Time Complexity: O(n)

Space Complexity: O(n)

But the problem requires in-place compression.

# Optimal Idea — Slow and Fast Pointers

Use:

- `fast` → counts a group of same characters
- `slow` → writes compressed output

# Key Observation

For every group:

- write the character
- if count > 1, write its digits

# Pointer Movement Dry Run

chars = ["a","a","b","b","c","c","c"]

| slow | fast range | group | count | write |
|------|------------|-------|-------|-------|
| 0 | 0-1 | a | 2 | a2 |
| 2 | 2-3 | b | 2 | b2 |
| 4 | 4-6 | c | 3 | c3 |

```python
class Solution:
    def compress(self, chars):
        slow = 0
        fast = 0

        while fast < len(chars):
            current_char = chars[fast]
            count = 0

            while fast < len(chars) and chars[fast] == current_char:
                fast += 1
                count += 1

            chars[slow] = current_char
            slow += 1

            if count > 1:
                for digit in str(count):
                    chars[slow] = digit
                    slow += 1

        return slow
```

# Complexity Comparison

Brute Force:
- Time → O(n)
- Space → O(n)

Optimal:
- Time → O(n)
- Space → O(1)
