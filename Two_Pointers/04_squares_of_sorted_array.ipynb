# Two Pointer Problem 4 — Squares of a Sorted Array

## Problem Statement

Given a sorted array that may contain negative numbers, return a new array of the squares of each number in sorted order.

Example:

nums = [-4, -1, 0, 3, 10]

Output: [0, 1, 9, 16, 100]

# Important Observation

Even though the input array is sorted, squaring negative numbers can disturb the order.

Example:

[-4, -1, 0, 3, 10]

Squares become:

[16, 1, 0, 9, 100]

This is not sorted.

# Optimal Idea — Opposite Ends

The largest square will always come from either:

- left end
- right end

because the largest absolute value lies at the ends.

# Pointer Movement Dry Run

nums = [-4, -1, 0, 3, 10]

| left | right | abs(left) | abs(right) | chosen |
|------|-------|-----------|------------|--------|
| 0 | 4 | 4 | 10 | 100 |
| 0 | 3 | 4 | 3 | 16 |
| 1 | 3 | 1 | 3 | 9 |
| 1 | 2 | 1 | 0 | 1 |
| 2 | 2 | 0 | 0 | 0 |

Final = [0, 1, 9, 16, 100]

```python
def sorted_squares(nums):
    left = 0
    right = len(nums) - 1

    result = [0] * len(nums)
    index = len(nums) - 1

    while left <= right:
        left_square = nums[left] ** 2
        right_square = nums[right] ** 2

        if left_square > right_square:
            result[index] = left_square
            left += 1
        else:
            result[index] = right_square
            right -= 1

        index -= 1

    return result


nums = [-4, -1, 0, 3, 10]
print(sorted_squares(nums))
```

# Why this works

The array is sorted.

So the largest absolute value must lie at one of the two ends.

We compare both ends and fill the result array from the back.

# Complexity Analysis

Time Complexity: O(n)

Space Complexity: O(n)

# Pattern Takeaway

This is an excellent **Opposite Ends Two Pointer** problem for arrays.

Very frequently asked in interviews.
