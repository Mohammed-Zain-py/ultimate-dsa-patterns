# Two Pointer Problem 5 — Container With Most Water

## Problem Statement

You are given an array `height` where each element represents the height of a vertical line.

Find two lines that together with the x-axis form a container that holds the maximum amount of water.

Example:

height = [1,8,6,2,5,4,8,3,7]

Output: 49

# Brute Force Approach

Check every possible pair of lines.

For every pair `(i, j)`:

Area = min(height[i], height[j]) * (j - i)

Keep track of the maximum area.

```python
def max_area_bruteforce(height):
    max_water = 0

    for i in range(len(height)):
        for j in range(i + 1, len(height)):
            width = j - i
            current_height = min(height[i], height[j])
            area = width * current_height

            max_water = max(max_water, area)

    return max_water


height = [1,8,6,2,5,4,8,3,7]
print(max_area_bruteforce(height))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Space Complexity: O(1)

This becomes slow for large input sizes.

# Why optimize?

The brute force solution checks all pairs.

But can we avoid unnecessary checks?

Yes.

This is where the Two Pointer pattern becomes powerful.

# Optimal Idea — Opposite Ends

Start with the widest possible container.

Take:
- left pointer at start
- right pointer at end

After calculating area, move the pointer with the smaller height.

# Why move the smaller height?

Area depends on:

width × minimum height

Width always decreases when pointers move.

So the only hope of getting a larger area is by increasing the minimum height.

That is why we move the smaller height pointer.

# Pointer Movement Dry Run

height = [1,8,6,2,5,4,8,3,7]

| left | right | min height | width | area | action |
|------|-------|------------|-------|------|--------|
| 0 | 8 | 1 | 8 | 8 | move left |
| 1 | 8 | 7 | 7 | 49 | move right |
| 1 | 7 | 3 | 6 | 18 | move right |
| 1 | 6 | 8 | 5 | 40 | continue |

```python
# Just inside function 

def max_area_optimal(height):
    left = 0
    right = len(height) - 1
    max_water = 0

    while left < right:
        width = right - left
        current_height = min(height[left], height[right])
        area = width * current_height

        max_water = max(max_water, area)

        if height[left] < height[right]:
            left += 1
        else:
            right -= 1

    return max_water


height = [1,8,6,2,5,4,8,3,7]
print(max_area_optimal(height))
```

# Interview Platform Friendly Solution

```python
# This is inside a class
class Solution:
    def maxArea(self, height):
        left = 0
        right = len(height) - 1
        max_water = 0

        while left < right:
            width = right - left
            current_height = min(height[left], height[right])
            area = width * current_height

            max_water = max(max_water, area)

            if height[left] < height[right]:
                left += 1
            else:
                right -= 1

        return max_water
```

# Complexity Analysis

Brute Force: O(n²)

Optimal: O(n)

Huge optimization using Two Pointers.
