# Two Pointer Problem 26 — Remove Duplicates from Sorted Array II

## Problem Statement

Given a sorted array `nums`, remove duplicates in-place such that each element appears **at most twice**.

Return the new length.

Example:

nums = [1, 1, 1, 2, 2, 3]

Output:
5

Modified array:
[1, 1, 2, 2, 3, _]

# Brute Force Approach

Create a new array.

Track how many times each number has appeared.

Add only up to 2 times.

This works but uses extra space.

```python
def remove_duplicates_ii_bruteforce(nums):
    if len(nums) <= 2:
        return len(nums), nums

    result = []
    count = {}

    for num in nums:
        count[num] = count.get(num, 0) + 1

        if count[num] <= 2:
            result.append(num)

    return len(result), result


print(remove_duplicates_ii_bruteforce([1,1,1,2,2,3]))
```

# Complexity of Brute Force

Time Complexity: O(n)

Space Complexity: O(n)

But the problem requires in-place modification.

# Optimal Idea — Slow and Fast Pointers

Use:

- `fast` → scans the array
- `slow` → writes valid values

A value is valid if it is different from the value at `slow - 2`.

# Key Observation

Because the array is sorted,

if `nums[fast] != nums[slow - 2]`

then it is safe to keep it.

# Pointer Movement Dry Run

nums = [1, 1, 1, 2, 2, 3]

| slow | fast | nums[slow-2] | nums[fast] | action |
|------|------|--------------|------------|--------|
| 2 | 2 | 1 | 1 | skip |
| 2 | 3 | 1 | 2 | keep |
| 3 | 4 | 1 | 2 | keep |
| 4 | 5 | 2 | 3 | keep |

```python
class Solution:
    def removeDuplicates(self, nums):
        if len(nums) <= 2:
            return len(nums)

        slow = 2

        for fast in range(2, len(nums)):
            if nums[fast] != nums[slow - 2]:
                nums[slow] = nums[fast]
                slow += 1

        return slow
```

# Complexity Comparison

Brute Force:
- Time → O(n)
- Space → O(n)

Optimal:
- Time → O(n)
- Space → O(1)
