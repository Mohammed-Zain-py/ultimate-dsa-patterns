# Two Pointer Problem 3 — Reverse String

## Problem Statement

Given a string, reverse it using Two Pointers.

Example:

s = "python"

Output: "nohtyp"

# Brute Force Idea

Use slicing:

s[::-1]

This is easy but does not help build Two Pointer logic.

For DSA learning, we solve it using pointers.

# Optimal Idea — Opposite Ends

Take one pointer at the start and one at the end.

Swap both characters.

Move inward.

# Pointer Movement Dry Run

s = "python"

| left | right | before swap | after swap |
|------|-------|-------------|------------|
| 0 | 5 | p n | n p |
| 1 | 4 | y o | o y |
| 2 | 3 | t h | h t |

Final string = "nohtyp"

```python
def reverse_string(s):
    chars = list(s)

    left = 0
    right = len(chars) - 1

    while left < right:
        chars[left], chars[right] = chars[right], chars[left]

        left += 1
        right -= 1

    return "".join(chars)


print(reverse_string("python"))
print(reverse_string("madam"))
```

# Why this works

Each iteration swaps the outer characters.

After every swap, the unsolved part becomes smaller.

This continues until both pointers meet.

# Complexity Analysis

Time Complexity: O(n)

Space Complexity: O(n)

Extra space is used because strings are immutable in Python.

# Pattern Takeaway

This is a pure **Opposite Ends Two Pointer** problem.

This pattern is extremely important for:
- strings
- arrays
- palindrome logic
