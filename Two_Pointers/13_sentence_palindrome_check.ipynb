# Two Pointer Problem 13 — Valid Palindrome Sentence

## Problem Statement

Given a string, check whether it is a palindrome after removing all non-alphanumeric characters and ignoring case.

Example:

s = "A man, a plan, a canal: Panama"

Output: True

# Brute Force Approach

Create a cleaned string containing only letters and digits.

Convert to lowercase.

Then compare with its reverse.

```python
def sentence_palindrome_bruteforce(s):
    cleaned = ""

    for ch in s:
        if ch.isalnum():
            cleaned += ch.lower()

    return cleaned == cleaned[::-1]


print(sentence_palindrome_bruteforce("A man, a plan, a canal: Panama"))
```

# Complexity of Brute Force

Time Complexity: O(n)

Space Complexity: O(n)

Extra space is used for the cleaned string.

# Optimal Idea — Opposite Ends

Use two pointers from both ends.

Skip all non-alphanumeric characters.

Compare lowercase characters directly.

This avoids creating a new string.

# Pointer Movement Dry Run (partial)

s = "A man, a plan, a canal: Panama"

| left | right | char left | char right | action |
|------|-------|-----------|------------|--------|
| 0 | 29 | A | a | move both |
| 1 | 28 | space | m | skip left |
| 2 | 28 | m | m | move both |

```python
class Solution:
    def isPalindrome(self, s):
        left = 0
        right = len(s) - 1

        while left < right:
            while left < right and not s[left].isalnum():
                left += 1

            while left < right and not s[right].isalnum():
                right -= 1

            if s[left].lower() != s[right].lower():
                return False

            left += 1
            right -= 1

        return True
```

# Complexity Comparison

Brute Force:
- Time → O(n)
- Space → O(n)

Optimal:
- Time → O(n)
- Space → O(1)
