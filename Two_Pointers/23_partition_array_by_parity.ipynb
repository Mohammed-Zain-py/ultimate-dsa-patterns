# Two Pointer Problem 23 — Partition Array by Parity

## Problem Statement

Given an integer array `nums`, move all even numbers to the beginning and all odd numbers to the end.

The relative order does not matter.

Example:

nums = [3, 1, 2, 4]

Output:
[2, 4, 3, 1]

# Brute Force Approach

Create two separate lists:

- even numbers
- odd numbers

Then combine them.

This works but uses extra space.

```python
def partition_parity_bruteforce(nums):
    even = []
    odd = []

    for num in nums:
        if num % 2 == 0:
            even.append(num)
        else:
            odd.append(num)

    return even + odd


print(partition_parity_bruteforce([3, 1, 2, 4]))
```

# Complexity of Brute Force

Time Complexity: O(n)

Space Complexity: O(n)

Can we do it in-place?

Yes — using Two Pointers.

# Optimal Idea — Opposite Ends / Partition

Use two pointers:

- `left` → searches for odd number
- `right` → searches for even number

Swap them.

# Key Observation

Correct position:

- even → left side
- odd → right side

# Pointer Movement Dry Run

nums = [3, 1, 2, 4]

| left | right | nums[left] | nums[right] | action |
|------|-------|------------|-------------|--------|
| 0 | 3 | 3 | 4 | swap |
| 1 | 2 | 1 | 2 | swap |

```python
class Solution:
    def sortArrayByParity(self, nums):
        left = 0
        right = len(nums) - 1

        while left < right:
            if nums[left] % 2 > nums[right] % 2:
                nums[left], nums[right] = nums[right], nums[left]

            if nums[left] % 2 == 0:
                left += 1

            if nums[right] % 2 == 1:
                right -= 1

        return nums
```

# Complexity Comparison

Brute Force:
- Time → O(n)
- Space → O(n)

Optimal:
- Time → O(n)
- Space → O(1)
