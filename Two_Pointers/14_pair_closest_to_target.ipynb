# Two Pointer Problem 14 — Pair Closest to Target Sum

## Problem Statement

Given a sorted array and a target value, find the pair whose sum is closest to the target.

Example:

nums = [1, 3, 4, 7, 10]
target = 15

Output: (4, 10)

# Brute Force Approach

Check every possible pair.

For each pair, calculate:

absolute difference = |target - current_sum|

Track the minimum difference.

```python
def pair_closest_bruteforce(nums, target):
    best_pair = None
    min_diff = float("inf")

    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            current_sum = nums[i] + nums[j]
            diff = abs(target - current_sum)

            if diff < min_diff:
                min_diff = diff
                best_pair = (nums[i], nums[j])

    return best_pair


print(pair_closest_bruteforce([1, 3, 4, 7, 10], 15))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Space Complexity: O(1)

Too slow for large arrays.

# Optimal Idea — Opposite Ends

Because the array is sorted:

- if sum is too small → move left
- if sum is too large → move right

Track the closest pair while moving.

# Pointer Movement Dry Run

nums = [1, 3, 4, 7, 10]
target = 15

| left | right | sum | diff | action |
|------|-------|-----|------|--------|
| 0 | 4 | 11 | 4 | move left |
| 1 | 4 | 13 | 2 | move left |
| 2 | 4 | 14 | 1 | move left |
| 3 | 4 | 17 | 2 | move right |

```python
class Solution:
    def pairClosestToTarget(self, nums, target):
        left = 0
        right = len(nums) - 1

        min_diff = float("inf")
        best_pair = None

        while left < right:
            current_sum = nums[left] + nums[right]
            diff = abs(target - current_sum)

            if diff < min_diff:
                min_diff = diff
                best_pair = (nums[left], nums[right])

            if current_sum < target:
                left += 1
            else:
                right -= 1

        return best_pair
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)

Huge improvement using sorted property.

# Pattern Takeaway

Classic **sorted array opposite-end two-pointer problem**.

Very good for placements.
