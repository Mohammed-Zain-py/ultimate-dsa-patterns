# How Coding Platform Templates Work

This notebook is created to solve one of the most common beginner confusions in coding interviews.

A lot of students know the logic but still get confused on platforms like HackerRank and LeetCode because the code is written inside a class template.

This notebook explains everything in detail so that this confusion never happens again.

# The most common template

Most coding platforms provide code in this format:

```python
class Solution:
    def functionName(self, ...):
        ...

# The code

````python
class Solution:
    def addNumbers(self, a, b):
        return a + b
````

# Biggest confusion

Many students ask:

- Where is the object created?
- Why is self used?
- Should I use __init__?
- How are test cases passed?

These are all valid doubts.

# Important truth

You usually DO NOT create the object yourself.

The platform does it internally.

Example internal working:

```python
solution = Solution()
output = solution.addNumbers(2, 3)

# Why self is used

When a method is inside a class, Python automatically passes the object as the first argument.

That first argument is usually written as `self`.

This is why we write:

````python
class Solution:
    def multiply(self, a, b):
        return a * b
````

Here, `self` refers to the object created internally by the platform.

# When to use __init__

Use `__init__` only when the problem specifically requires object initialization.

Example:

````python
class Solution:
    def __init__(self):
        self.count = 0
````

For most DSA problems, `__init__` is NOT required.

You only need to implement the given function.
