# Two Pointer Problem 8 — Valid Palindrome II

## Problem Statement

Given a string `s`, return `True` if it can become a palindrome after deleting **at most one character**.

Example:

s = "abca"

Output: True

Explanation:
Remove 'c' → "aba"

# Brute Force Approach

Try removing every character one by one.

For each new string, check whether it is a palindrome.

This works but is slow.

```python
def is_palindrome(s):
    return s == s[::-1]


def valid_palindrome_bruteforce(s):
    if is_palindrome(s):
        return True

    for i in range(len(s)):
        new_string = s[:i] + s[i + 1:]

        if is_palindrome(new_string):
            return True

    return False


print(valid_palindrome_bruteforce("abca"))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Because for each deletion, we check the entire string again.

This is not efficient for large inputs.

# Optimal Idea — Opposite Ends with One Chance

Take two pointers from opposite ends.

If characters match → move inward.

If mismatch occurs → we get only one chance to delete either:

- left character
- right character

Then check if the remaining substring is palindrome.

# Pointer Movement Dry Run

s = "abca"

| left | right | s[left] | s[right] | action |
|------|-------|---------|----------|--------|
| 0 | 3 | a | a | move both |
| 1 | 2 | b | c | mismatch |
| - | - | - | - | try removing b OR c |

Check:
- "ca" → not palindrome
- "bc" → not palindrome
Actually valid substring check becomes:
- "ba"
- "ac"

For "abca", removing c gives "aba" → palindrome

```python
class Solution:
    def validPalindrome(self, s):
        def is_palindrome_range(left, right):
            while left < right:
                if s[left] != s[right]:
                    return False
                left += 1
                right -= 1
            return True

        left = 0
        right = len(s) - 1

        while left < right:
            if s[left] != s[right]:
                return (
                    is_palindrome_range(left + 1, right) or
                    is_palindrome_range(left, right - 1)
                )

            left += 1
            right -= 1

        return True
```

# Complexity Comparison

Brute Force:
- Time → O(n²)
- Space → O(n)

Optimal:
- Time → O(n)
- Space → O(1)
