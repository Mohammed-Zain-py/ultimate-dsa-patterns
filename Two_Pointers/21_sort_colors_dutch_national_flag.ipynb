# Two Pointer Problem 21 — Sort Colors (Dutch National Flag)

## Problem Statement

Given an array containing only `0`, `1`, and `2`, sort it in-place.

Example:

nums = [2, 0, 2, 1, 1, 0]

Output:
[0, 0, 1, 1, 2, 2]

# Brute Force Approach

Use sorting.

This is the most natural first thought.

But the goal is to solve it in one pass.

```python
def sort_colors_bruteforce(nums):
    nums.sort()
    return nums


print(sort_colors_bruteforce([2, 0, 2, 1, 1, 0]))
```

# Complexity of Brute Force

Time Complexity: O(n log n)

Space Complexity: O(1)

Can we do better?

Yes — O(n)

# Optimal Idea — Three Pointers

This is a special Two Pointer / partition-style problem.

We use:

- `low` → next position for 0
- `mid` → current scanning pointer
- `high` → next position for 2

# Key Observation

- 0 should go left
- 2 should go right
- 1 stays in middle

# Pointer Movement Dry Run

nums = [2, 0, 2, 1, 1, 0]

| low | mid | high | nums[mid] | action |
|-----|-----|------|-----------|--------|
| 0 | 0 | 5 | 2 | swap with high |
| 0 | 0 | 4 | 0 | swap with low |
| 1 | 1 | 4 | 0 | swap with low |
| 2 | 2 | 4 | 2 | swap with high |

```python
class Solution:
    def sortColors(self, nums):
        low = 0
        mid = 0
        high = len(nums) - 1

        while mid <= high:
            if nums[mid] == 0:
                nums[low], nums[mid] = nums[mid], nums[low]
                low += 1
                mid += 1

            elif nums[mid] == 1:
                mid += 1

            else:
                nums[mid], nums[high] = nums[high], nums[mid]
                high -= 1

        return nums
```

# Complexity Comparison

Brute Force:
- Time → O(n log n)

Optimal:
- Time → O(n)
- Space → O(1)

# Pattern Takeaway

This is one of the most important **partition-style Two Pointer problems**.

Very frequently asked.
