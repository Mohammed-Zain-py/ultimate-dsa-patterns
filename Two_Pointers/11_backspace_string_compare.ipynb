# Two Pointer Problem 11 — Backspace String Compare

## Problem Statement

Given two strings `s` and `t`, return `True` if they are equal when both are typed into empty text editors.

`#` means backspace.

Example:

s = "ab#c"
t = "ad#c"

Output: True

# Brute Force Approach

Simulate typing using a stack-like list.

For each character:
- if normal → push
- if # → pop if possible

Then compare final strings.

```python
def process_string(string):
    stack = []

    for ch in string:
        if ch == "#":
            if stack:
                stack.pop()
        else:
            stack.append(ch)

    return "".join(stack)


def backspace_compare_bruteforce(s, t):
    return process_string(s) == process_string(t)


print(backspace_compare_bruteforce("ab#c", "ad#c"))
```

# Complexity of Brute Force

Time Complexity: O(n)

Space Complexity: O(n)

Can we reduce the extra space?

Yes — using Two Pointers from the end.

# Optimal Idea — Opposite Ends from Back

Start both pointers from the end.

Skip characters based on backspace count.

This avoids building extra strings.

# Key Observation

When moving from the end:

- `#` means skip one character
- maintain a skip counter

# Pointer Movement Dry Run (partial)

s = "ab#c"
t = "ad#c"

| i | j | s[i] | t[j] | action |
|---|---|------|------|--------|
| 3 | 3 | c | c | compare |
| 2 | 2 | # | # | skip prev |
| 0 | 0 | a | a | compare |

```python
class Solution:
    def backspaceCompare(self, s, t):
        i = len(s) - 1
        j = len(t) - 1

        skip_s = 0
        skip_t = 0

        while i >= 0 or j >= 0:
            while i >= 0:
                if s[i] == "#":
                    skip_s += 1
                    i -= 1
                elif skip_s > 0:
                    skip_s -= 1
                    i -= 1
                else:
                    break

            while j >= 0:
                if t[j] == "#":
                    skip_t += 1
                    j -= 1
                elif skip_t > 0:
                    skip_t -= 1
                    j -= 1
                else:
                    break

            if i >= 0 and j >= 0 and s[i] != t[j]:
                return False

            if (i >= 0) != (j >= 0):
                return False

            i -= 1
            j -= 1

        return True
```

# Complexity Comparison

Brute Force:
- Time → O(n)
- Space → O(n)

Optimal:
- Time → O(n)
- Space → O(1)
