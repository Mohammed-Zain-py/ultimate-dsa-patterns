# Two Pointer Problem 18 — Move Zeroes

## Problem Statement

Given an integer array `nums`, move all `0`s to the end while maintaining the relative order of the non-zero elements.

Do this in-place.

Example:

nums = [0, 1, 0, 3, 12]

Output:
[1, 3, 12, 0, 0]

# Brute Force Approach

Create a new array.

First add all non-zero elements.

Then append zeroes at the end.

This works but uses extra space.

```python
def move_zeroes_bruteforce(nums):
    result = []

    for num in nums:
        if num != 0:
            result.append(num)

    zero_count = len(nums) - len(result)

    result.extend([0] * zero_count)

    return result


print(move_zeroes_bruteforce([0, 1, 0, 3, 12]))
```

# Complexity of Brute Force

Time Complexity: O(n)

Space Complexity: O(n)

But the problem asks for in-place modification.

# Optimal Idea — Slow and Fast Pointers

Use two pointers:

- `fast` → scans every element
- `slow` → points to the position where next non-zero should go

# Key Observation

Whenever `fast` finds a non-zero element,

swap it with `slow`

Then move `slow`.

# Pointer Movement Dry Run

nums = [0, 1, 0, 3, 12]

| slow | fast | nums[fast] | action |
|------|------|------------|--------|
| 0 | 0 | 0 | move fast |
| 0 | 1 | 1 | swap with slow |
| 1 | 2 | 0 | move fast |
| 1 | 3 | 3 | swap |
| 2 | 4 | 12 | swap |

```python
class Solution:
    def moveZeroes(self, nums):
        slow = 0

        for fast in range(len(nums)):
            if nums[fast] != 0:
                nums[slow], nums[fast] = nums[fast], nums[slow]
                slow += 1

        return nums
```

# Complexity Comparison

Brute Force:
- Time → O(n)
- Space → O(n)

Optimal:
- Time → O(n)
- Space → O(1)

# Pattern Takeaway

This is the first major **Slow–Fast Two Pointer** problem.

A must-know interview question.
