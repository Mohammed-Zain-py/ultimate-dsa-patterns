# Two Pointer Problem 24 — Max Consecutive Ones II

## Problem Statement

Given a binary array `nums`, return the maximum number of consecutive `1`s if you can flip at most one `0`.

Example:

nums = [1, 0, 1, 1, 0]

Output:
4

# Brute Force Approach

Try flipping every `0` one by one.

For each flip, count the maximum consecutive ones.

This works but is slow.

```python
def max_ones_bruteforce(nums):
    max_count = 0

    for i in range(len(nums)):
        temp = nums[:]

        if temp[i] == 0:
            temp[i] = 1

        count = 0
        best = 0

        for num in temp:
            if num == 1:
                count += 1
                best = max(best, count)
            else:
                count = 0

        max_count = max(max_count, best)

    return max_count


print(max_ones_bruteforce([1, 0, 1, 1, 0]))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Space Complexity: O(n)

Too slow for large arrays.

# Optimal Idea — Slow and Fast Pointers

Use a variable-size window.

Keep at most one zero inside the window.

- `fast` expands
- `slow` shrinks when zero count exceeds 1

# Pointer Movement Dry Run

nums = [1,0,1,1,0]

| slow | fast | zero_count | window_size |
|------|------|------------|-------------|
| 0 | 0 | 0 | 1 |
| 0 | 1 | 1 | 2 |
| 0 | 2 | 1 | 3 |
| 0 | 3 | 1 | 4 |
| 0 | 4 | 2 | shrink |

```python
class Solution:
    def findMaxConsecutiveOnes(self, nums):
        slow = 0
        zero_count = 0
        max_length = 0

        for fast in range(len(nums)):
            if nums[fast] == 0:
                zero_count += 1

            while zero_count > 1:
                if nums[slow] == 0:
                    zero_count -= 1
                slow += 1

            max_length = max(max_length, fast - slow + 1)

        return max_length
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(1)

# Pattern Takeaway

Excellent **slow–fast / window bridge** problem.

Very commonly asked.
