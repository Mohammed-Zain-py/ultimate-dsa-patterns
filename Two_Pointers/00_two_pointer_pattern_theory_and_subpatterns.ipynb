# Two Pointer Pattern — Theory and Sub-Patterns

Two Pointer is one of the most important DSA patterns for coding interviews.

This pattern is used whenever we need to process data using two positions (indexes) in an efficient way.

Instead of using nested loops, we use two indexes that move intelligently.

This often helps reduce time complexity from **O(n²) to O(n)**.

# Where is this pattern used?

Two Pointer is commonly used in:

- sorted arrays
- palindrome problems
- pair sum problems
- string processing
- array partitioning
- merging arrays
- slow-fast traversal

# The biggest idea

Instead of checking every pair using nested loops, we move two pointers based on conditions.

This reduces unnecessary work.

# Sub-Pattern 1: Opposite Ends

One pointer starts from the left.

Another starts from the right.

Both move toward each other.

[1, 2, 3, 4, 5, 6]
 ^             ^
 |             |
left         right



[1, 2, 3, 4, 5, 6]
    ^        ^
    |        |
   left     right

# Sub-Pattern 2: Slow and Fast

Both pointers move in the same direction.

But one moves slower than the other.

    [1, 2, 3, 4, 5, 6]
     ^   ^
     |   |
  left   right

      [1, 2, 3, 4, 5, 6]
          ^     ^
          |     |
        left    right

Common use cases:

- remove duplicates
- move zeroes
- in-place modifications
- linked list cycle detection (later)

# Sub-Pattern 3: Expand Around Center

This is mostly used in string problems.

Start with two pointers near the center and expand outward.

[1, 2, 3, 4, 5, 6]
       ^  ^
       |  |
    left  right


[1, 2, 3, 4, 5, 6]
    ^        ^
    |        |
  left      right

Used in:

- longest palindrome substring
- count palindrome substrings

# Sub-Pattern 4: Partition / Rearrangement

Used when elements need to be rearranged in-place.

Examples:

- sort colors
- segregate positives and negatives
- even odd arrangement

# Sub-Pattern 5: Merge Style

Used when working with two sorted sequences.

Examples:

- merge sorted arrays
- interval merge logic

# Most important interview tip

Whenever you see:

- sorted input
- pair problem
- palindrome
- opposite direction traversal

immediately think:

**Can Two Pointers be used here?**
