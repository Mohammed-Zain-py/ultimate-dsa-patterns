# Two Pointer Problem 7 — Reverse Vowels of a String

## Problem Statement

Given a string, reverse only the vowels in the string and return the new string.

Example:

s = "hello"

Output: "holle"

# Brute Force Approach

First, extract all vowels from the string.

Then reverse the vowel list.

Finally, place them back in the string.

This works but uses extra space.

```python
def reverse_vowels_bruteforce(s):
    vowels = "aeiouAEIOU"
    extracted = []

    for ch in s:
        if ch in vowels:
            extracted.append(ch)

    extracted.reverse()

    result = []
    index = 0

    for ch in s:
        if ch in vowels:
            result.append(extracted[index])
            index += 1
        else:
            result.append(ch)

    return "".join(result)


print(reverse_vowels_bruteforce("hello"))
```

# Complexity of Brute Force

Time Complexity: O(n)

Space Complexity: O(n)

Can we reduce the extra space?

Yes — using Two Pointers.

# Optimal Idea — Opposite Ends

Take one pointer from the left and one from the right.

Move them until both land on vowels.

Then swap.

Continue moving inward.

# Pointer Movement Dry Run

s = "hello"

| left | right | char left | char right | action |
|------|-------|-----------|------------|--------|
| 0 | 4 | h | o | move left |
| 1 | 4 | e | o | swap |
| 2 | 3 | l | l | move both |

```python
class Solution:
    def reverseVowels(self, s):
        vowels = "aeiouAEIOU"
        chars = list(s)

        left = 0
        right = len(chars) - 1

        while left < right:
            while left < right and chars[left] not in vowels:
                left += 1

            while left < right and chars[right] not in vowels:
                right -= 1

            chars[left], chars[right] = chars[right], chars[left]

            left += 1
            right -= 1

        return "".join(chars)
```

# Complexity Comparison

Brute Force:
- Time → O(n)
- Space → O(n)

Optimal:
- Time → O(n)
- Space → O(n)

Because strings are immutable in Python, list conversion is required.

# Pattern Takeaway

This is a classic **Opposite Ends Two Pointer string problem**.

Very useful for strengthening pointer movement logic.
