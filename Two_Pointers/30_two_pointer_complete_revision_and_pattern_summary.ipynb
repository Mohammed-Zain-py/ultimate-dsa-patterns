# What is Two Pointer?

Two Pointer is a problem-solving technique where we use two indexes to process data efficiently.

Instead of nested loops, we intelligently move pointers.

This often reduces:

O(n²) → O(n)

# Sub-Pattern 1 — Opposite Ends

Use when:

- input is sorted
- palindrome checking
- pair sum
- trapping water
- reverse string

Examples:
- Two Sum II
- Valid Palindrome
- Container With Most Water
- Trapping Rain Water
- Three Sum
- Four Sum

# Sub-Pattern 2 — Slow and Fast

Use when:

- in-place modification
- duplicate removal
- move elements
- compression
- partitioning

Examples:
- Move Zeroes
- Remove Duplicates
- Sort Colors
- String Compression

# Sub-Pattern 3 — Expand Around Center

Use when:

- palindrome substring
- count palindromes
- center expansion logic

Examples:
- Longest Palindromic Substring
- Count Palindromic Substrings

# Quick Recognition Checklist

Ask these:

1. Is the array sorted?
2. Do I need a pair / triplet?
3. Is palindrome involved?
4. Is in-place modification required?
5. Is substring center logic involved?

# Complexity Revision

Typical improvements:

- O(n²) → O(n)
- O(n³) → O(n²)
- O(n log n) → O(n)

This is why Two Pointer is extremely important.
