# Two Pointer Problem 16 — Three Sum Closest

## Problem Statement

Given an integer array `nums` and an integer `target`, find three integers such that the sum is closest to the target.

Return the sum of the three integers.

Example:

nums = [-1, 2, 1, -4]
target = 1

Output: 2

# Brute Force Approach

Check every possible triplet using 3 nested loops.

Track the triplet sum with the minimum absolute difference from the target.

```python
def three_sum_closest_bruteforce(nums, target):
    closest_sum = nums[0] + nums[1] + nums[2]
    n = len(nums)

    for i in range(n):
        for j in range(i + 1, n):
            for k in range(j + 1, n):
                current_sum = nums[i] + nums[j] + nums[k]

                if abs(target - current_sum) < abs(target - closest_sum):
                    closest_sum = current_sum

    return closest_sum


print(three_sum_closest_bruteforce([-1, 2, 1, -4], 1))
```

# Complexity of Brute Force

Time Complexity: O(n³)

Space Complexity: O(1)

Too slow for larger arrays.

# Optimal Idea — Sort + Two Pointers

Sort the array.

Fix one element.

Use two pointers for the remaining part.

This reduces one loop.

# Pointer Movement Dry Run

nums = [-4, -1, 1, 2]
target = 1

Fix i = 0 → -4

| left | right | sum | diff | action |
|------|-------|-----|------|--------|
| 1 | 3 | -3 | 4 | move left |
| 2 | 3 | -1 | 2 | closest so far |

```python
class Solution:
    def threeSumClosest(self, nums, target):
        nums.sort()
        closest_sum = nums[0] + nums[1] + nums[2]

        for i in range(len(nums) - 2):
            left = i + 1
            right = len(nums) - 1

            while left < right:
                current_sum = nums[i] + nums[left] + nums[right]

                if abs(target - current_sum) < abs(target - closest_sum):
                    closest_sum = current_sum

                if current_sum < target:
                    left += 1
                elif current_sum > target:
                    right -= 1
                else:
                    return current_sum

        return closest_sum
```

# Complexity Comparison

Brute Force:
- Time → O(n³)

Optimal:
- Time → O(n²)
