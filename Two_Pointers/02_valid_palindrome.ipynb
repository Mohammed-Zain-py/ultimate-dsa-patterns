# Two Pointer Problem 2 — Valid Palindrome

## Problem Statement

Given a string, check whether it reads the same from left to right and right to left.

Example:

s = "madam"

Output: True

# Brute Force Idea

Reverse the string and compare it with the original string.

This works, but it uses extra space.

A better way is to use Two Pointers.

# Optimal Idea — Opposite Ends

Take one pointer from the beginning and one from the end.

Compare characters.

- if same → move both pointers
- if different → not palindrome

# Pointer Movement Dry Run

s = "madam"

| left | right | s[left] | s[right] | action |
|------|-------|---------|----------|--------|
| 0 | 4 | m | m | move both |
| 1 | 3 | a | a | move both |
| 2 | 2 | d | d | stop |

Palindrome confirmed

```python
def is_palindrome(s):
    left = 0
    right = len(s) - 1

    while left < right:
        if s[left] != s[right]:
            return False

        left += 1
        right -= 1

    return True


print(is_palindrome("madam"))
print(is_palindrome("python"))
```

# Why this works

For a string to be palindrome:

first character = last character  
second character = second last character

and so on.

This is exactly what opposite-end pointers check efficiently.

# Complexity Analysis

Time Complexity: O(n)

Space Complexity: O(1)

# Pattern Takeaway

This is again the **Opposite Ends Two Pointer** pattern.

Important learning:

Same pattern can solve both:
- sorted arrays
- strings
