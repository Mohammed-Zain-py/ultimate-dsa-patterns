# Two Pointer Problem 15 — Three Sum

## Problem Statement

Given an integer array `nums`, return all unique triplets `[a, b, c]` such that:

a + b + c = 0

Example:

nums = [-1, 0, 1, 2, -1, -4]

Output:
[[-1, -1, 2], [-1, 0, 1]]

# Brute Force Approach

Check every possible triplet using 3 nested loops.

For each triplet, check whether the sum is zero.

Also avoid duplicates.

```python
def three_sum_bruteforce(nums):
    result = set()
    n = len(nums)

    for i in range(n):
        for j in range(i + 1, n):
            for k in range(j + 1, n):
                if nums[i] + nums[j] + nums[k] == 0:
                    triplet = tuple(sorted([nums[i], nums[j], nums[k]]))
                    result.add(triplet)

    return [list(triplet) for triplet in result]


print(three_sum_bruteforce([-1, 0, 1, 2, -1, -4]))
```

# Complexity of Brute Force

Time Complexity: O(n³)

Space Complexity: O(k)

This is too slow for large inputs.

# Optimal Idea — Sort + Two Pointers

Sort the array first.

Fix one element.

Then use two pointers on the remaining array.

This reduces one loop.

# Key Observation

For every fixed `nums[i]`:

We need two numbers whose sum is `-nums[i]`.

This becomes a Two Sum problem on the remaining sorted array.

# Pointer Movement Dry Run (partial)

Sorted nums = [-4, -1, -1, 0, 1, 2]

Fix i = 1 → nums[i] = -1

Target = 1

| left | right | sum | action |
|------|-------|-----|--------|
| 2 | 5 | 1 | triplet found |
| 3 | 4 | 1 | triplet found |

```python
class Solution:
    def threeSum(self, nums):
        nums.sort()
        result = []

        for i in range(len(nums)):
            if i > 0 and nums[i] == nums[i - 1]:
                continue

            left = i + 1
            right = len(nums) - 1

            while left < right:
                total = nums[i] + nums[left] + nums[right]

                if total == 0:
                    result.append([nums[i], nums[left], nums[right]])

                    left += 1
                    right -= 1

                    while left < right and nums[left] == nums[left - 1]:
                        left += 1

                    while left < right and nums[right] == nums[right + 1]:
                        right -= 1

                elif total < 0:
                    left += 1
                else:
                    right -= 1

        return result
```

# Complexity Comparison

Brute Force:
- Time → O(n³)

Optimal:
- Time → O(n²)

Very important optimization.
