# Two Pointer Problem 10 — Is Subsequence

## Problem Statement

Given two strings `s` and `t`, return `True` if `s` is a subsequence of `t`.

A subsequence means all characters of `s` appear in `t` in the same order.

Example:

s = "abc"
t = "ahbgdc"

Output: True

# Brute Force Idea

The most natural thought is to try matching every character of `s` by scanning through `t`.

Since order matters, we move step by step.

```python
def is_subsequence_bruteforce(s, t):
    i = 0

    for ch in t:
        if i < len(s) and s[i] == ch:
            i += 1

    return i == len(s)


print(is_subsequence_bruteforce("abc", "ahbgdc"))
```

# Complexity of Brute Force

Time Complexity: O(n)

Space Complexity: O(1)

This is already efficient.

The same idea is also the optimal approach.

# Optimal Idea — Two Pointers

Use one pointer for `s` and one pointer for `t`.

- if characters match → move both
- else → move only `t`

# Pointer Movement Dry Run

s = "abc"
t = "ahbgdc"

| i (s) | j (t) | s[i] | t[j] | action |
|-------|-------|------|------|--------|
| 0 | 0 | a | a | move both |
| 1 | 1 | b | h | move j |
| 1 | 2 | b | b | move both |
| 2 | 3 | c | g | move j |
| 2 | 4 | c | d | move j |
| 2 | 5 | c | c | move both |

```python
class Solution:
    def isSubsequence(self, s, t):
        i = 0
        j = 0

        while i < len(s) and j < len(t):
            if s[i] == t[j]:
                i += 1

            j += 1

        return i == len(s)
```

# Complexity

Time Complexity: O(n)

Space Complexity: O(1)
