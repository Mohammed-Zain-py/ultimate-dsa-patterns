# Two Pointer Problem 17 — Four Sum

## Problem Statement

Given an array `nums` and an integer `target`, return all unique quadruplets such that:

a + b + c + d = target

Example:

nums = [1, 0, -1, 0, -2, 2]
target = 0

Output:
[[-2, -1, 1, 2], [-2, 0, 0, 2], [-1, 0, 0, 1]]

# Brute Force Approach

Use 4 nested loops.

Check every possible quadruplet.

Store only unique results.

```python
def four_sum_bruteforce(nums, target):
    result = set()
    n = len(nums)

    for i in range(n):
        for j in range(i + 1, n):
            for k in range(j + 1, n):
                for l in range(k + 1, n):
                    total = nums[i] + nums[j] + nums[k] + nums[l]

                    if total == target:
                        quad = tuple(sorted([nums[i], nums[j], nums[k], nums[l]]))
                        result.add(quad)

    return [list(quad) for quad in result]


print(four_sum_bruteforce([1, 0, -1, 0, -2, 2], 0))
```

# Complexity of Brute Force

Time Complexity: O(n⁴)

Space Complexity: O(k)

Very slow for larger arrays.

# Optimal Idea — Sort + Nested Loop + Two Pointers

Sort the array.

Fix first two numbers.

Use two pointers for the remaining two numbers.

# Key Observation

After fixing two elements, the remaining problem becomes:

Find two numbers whose sum = remaining target

This is a Two Sum problem on a sorted array.

# Pointer Movement Dry Run (partial)

Sorted nums = [-2, -1, 0, 0, 1, 2]

Fix:
i = 0 → -2
j = 1 → -1

Remaining target = 3

| left | right | sum | action |
|------|-------|-----|--------|
| 2 | 5 | 2 | move left |
| 3 | 5 | 2 | move left |
| 4 | 5 | 3 | quad found |

```python
class Solution:
    def fourSum(self, nums, target):
        nums.sort()
        result = []
        n = len(nums)

        for i in range(n - 3):
            if i > 0 and nums[i] == nums[i - 1]:
                continue

            for j in range(i + 1, n - 2):
                if j > i + 1 and nums[j] == nums[j - 1]:
                    continue

                left = j + 1
                right = n - 1

                while left < right:
                    total = nums[i] + nums[j] + nums[left] + nums[right]

                    if total == target:
                        result.append([nums[i], nums[j], nums[left], nums[right]])

                        left += 1
                        right -= 1

                        while left < right and nums[left] == nums[left - 1]:
                            left += 1

                        while left < right and nums[right] == nums[right + 1]:
                            right -= 1

                    elif total < target:
                        left += 1
                    else:
                        right -= 1

        return result
```

# Complexity Comparison

Brute Force:
- Time → O(n⁴)

Optimal:
- Time → O(n³)

Huge improvement.
