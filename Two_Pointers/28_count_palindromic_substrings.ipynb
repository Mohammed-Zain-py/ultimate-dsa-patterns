# Two Pointer Problem 28 — Count Palindromic Substrings

## Problem Statement

Given a string `s`, return the total number of palindromic substrings.

Example:

s = "aaa"

Output:
6

Explanation:
"a", "a", "a", "aa", "aa", "aaa"

# Brute Force Approach

Generate all possible substrings.

Check each substring whether it is a palindrome.

Count all valid ones.

```python
def count_palindromes_bruteforce(s):
    count = 0

    for i in range(len(s)):
        for j in range(i, len(s)):
            substring = s[i:j + 1]

            if substring == substring[::-1]:
                count += 1

    return count


print(count_palindromes_bruteforce("aaa"))
```

# Complexity of Brute Force

Time Complexity: O(n³)

Too slow for long strings.

# Optimal Idea — Expand Around Center

Every palindrome has a center.

For every index:

- expand as odd-length palindrome
- expand as even-length palindrome

Count all expansions.

# Center Expansion Dry Run

s = "aaa"

Centers:
- index 0
- index 1
- index 2
- gaps between them

# Center Expansion Dry Run

s = "aaa"

Centers:
- index 0
- index 1
- index 2
- gaps between them

```python
class Solution:
    def countSubstrings(self, s):
        count = 0

        def expand(left, right):
            nonlocal count

            while left >= 0 and right < len(s) and s[left] == s[right]:
                count += 1
                left -= 1
                right += 1

        for i in range(len(s)):
            expand(i, i)       # odd length
            expand(i, i + 1)   # even length

        return count
```

# Complexity Comparison

Brute Force:
- Time → O(n³)

Optimal:
- Time → O(n²)
- Space → O(1)
