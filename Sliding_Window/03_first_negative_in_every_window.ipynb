# Sliding Window Problem 3 — First Negative Number in Every Window

## Problem Statement

Given an array and a window size `k`, find the first negative number in every window of size `k`.

If a window does not contain a negative number, return 0.

Example:

nums = [12, -1, -7, 8, -15, 30, 16, 28]
k = 3

Output:
[-1, -1, -7, -15, -15, 0]

# Brute Force Approach

For every window of size `k`, scan all elements.

Find the first negative number.

If none found → return 0.

```python
def first_negative_bruteforce(nums, k):
    result = []

    for i in range(len(nums) - k + 1):
        found = False

        for j in range(i, i + k):
            if nums[j] < 0:
                result.append(nums[j])
                found = True
                break

        if not found:
            result.append(0)

    return result


print(first_negative_bruteforce([12, -1, -7, 8, -15, 30, 16, 28], 3))
```

# Complexity of Brute Force

Time Complexity: O(n * k)

Repeated scanning for every window.

# Optimal Idea — Sliding Window + Queue

Maintain a list (queue) of negative numbers inside the current window.

- add when entering window
- remove when leaving window

# Key Observation

The first negative number in the window is always at the front of the queue.

# Window Movement Dry Run

nums = [12, -1, -7, 8, -15, 30, 16, 28]

Window size = 3

| window | negatives | output |
|--------|----------|--------|
| [12,-1,-7] | [-1,-7] | -1 |
| [-1,-7,8] | [-1,-7] | -1 |
| [-7,8,-15] | [-7,-15] | -7 |
| [8,-15,30] | [-15] | -15 |
| [-15,30,16] | [-15] | -15 |
| [30,16,28] | [] | 0 |

```python
from collections import deque

class Solution:
    def firstNegative(self, nums, k):
        result = []
        queue = deque()

        for i in range(len(nums)):
            if nums[i] < 0:
                queue.append(i)

            if i >= k - 1:
                if queue:
                    result.append(nums[queue[0]])
                else:
                    result.append(0)

                if queue and queue[0] == i - k + 1:
                    queue.popleft()

        return result
```

# Complexity Comparison

Brute Force:
- Time → O(n * k)

Optimal:
- Time → O(n)
- Space → O(k)

# Pattern Takeaway

This is a classic **fixed-size sliding window + queue problem**.

Very commonly asked.
