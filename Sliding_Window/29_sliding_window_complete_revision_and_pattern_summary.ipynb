# What is Sliding Window?

Sliding Window is a technique used for
continuous ranges:

- subarrays
- substrings
- fixed ranges
- variable ranges

It often optimizes:

O(n²) → O(n)

# Subtype 1 — Fixed Size Window

Use when the size is constant.

Common signals:
- size k
- every window of size k
- max / min / average of size k

# Subtype 2 — Variable Size Window

Use when the size changes based on condition.

Examples:
- longest substring
- minimum window
- at most k distinct

# Subtype 3 — Counting Windows

Very important for placements.

Common trick:

exactly(k) = at_most(k) - at_most(k-1)

# Subtype 4 — Frequency Map Windows

Used for:
- anagrams
- character replacement
- distinct counts
- minimum window substring

# Recognition Checklist

Ask these:

1. continuous range?
2. longest / shortest?
3. exactly k?
4. at most k?
5. size fixed?
