# Sliding Window Problem 15 — Fruit Into Baskets

## Problem Statement

You are given an array `fruits`
where each value represents a fruit type.

You have 2 baskets,
and each basket can contain only one type of fruit.

Find the length of the longest contiguous subarray
containing at most 2 distinct fruit types.

Example:

fruits = [1, 2, 1]

Output:
3

# Brute Force Approach

Check every possible subarray.

For each subarray,
count distinct fruit types.

If distinct count <= 2,
update maximum length.

```python
def fruit_baskets_bruteforce(fruits):
    max_length = 0

    for i in range(len(fruits)):
        distinct = set()

        for j in range(i, len(fruits)):
            distinct.add(fruits[j])

            if len(distinct) <= 2:
                max_length = max(max_length, j - i + 1)
            else:
                break

    return max_length


print(fruit_baskets_bruteforce([1, 2, 1]))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Too slow for larger inputs.

# Optimal Idea — Variable Size Sliding Window + Frequency Map

Maintain a window with at most 2 distinct fruit types.

- expand right
- if distinct > 2 → shrink left

# Window Dry Run

fruits = [1, 2, 1]

| left | right | window | distinct |
|------|-------|--------|---------|
| 0 | 0 | [1] | 1 |
| 0 | 1 | [1,2] | 2 |
| 0 | 2 | [1,2,1] | 2 |

```python
class Solution:
    def totalFruit(self, fruits):
        freq = defaultdict(int)
        left = 0
        max_length = 0

        for right in range(len(fruits)):
            freq[fruits[right]] += 1

            while len(freq) > 2:
                freq[fruits[left]] -= 1

                if freq[fruits[left]] == 0:
                    del freq[fruits[left]]

                left += 1

            max_length = max(max_length, right - left + 1)

        return max_length
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(2)

# Pattern Takeaway

This is one of the most important **at most K distinct sliding window** problems.

Very frequently asked.
