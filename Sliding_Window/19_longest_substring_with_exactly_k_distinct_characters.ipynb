# Sliding Window Problem 19 — Longest Substring with Exactly K Distinct Characters

## Problem Statement

Given a string `s` and an integer `k`,
return the length of the longest substring
that contains exactly `k` distinct characters.

Example:

s = "aabacbebebe"
k = 3

Output:
7

Explanation:
"cbebebe" is the longest valid substring

# Brute Force Approach

Check every possible substring.

Count distinct characters.

If distinct count == k,
update maximum length.

```python
def longest_exactly_k_bruteforce(s, k):
    max_length = 0

    for i in range(len(s)):
        distinct = set()

        for j in range(i, len(s)):
            distinct.add(s[j])

            if len(distinct) == k:
                max_length = max(max_length, j - i + 1)

            elif len(distinct) > k:
                break

    return max_length


print(longest_exactly_k_bruteforce("aabacbebebe", 3))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Too slow for long strings.

# Optimal Idea — Variable Size Sliding Window + Frequency Map

Maintain a window with at most `k` distinct characters.

Whenever distinct count == k,
update maximum length.

# Window Dry Run

s = "aabacbebebe"
k = 3

| left | right | window | distinct |
|------|-------|--------|---------|
| 0 | 0 | a | 1 |
| 0 | 1 | aa | 1 |
| 0 | 2 | aab | 2 |
| 0 | 3 | aaba | 2 |
| 0 | 4 | aabac | 3 |
| 0 | 5 | aabacb | 3 |

```python
from collections import defaultdict

class Solution:
    def longestKSubstr(self, s, k):
        freq = defaultdict(int)
        left = 0
        max_length = -1

        for right in range(len(s)):
            freq[s[right]] += 1

            while len(freq) > k:
                freq[s[left]] -= 1

                if freq[s[left]] == 0:
                    del freq[s[left]]

                left += 1

            if len(freq) == k:
                max_length = max(max_length, right - left + 1)

        return max_length
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(k)

# Pattern Takeaway

This is the classic **exactly K distinct sliding window** problem.

Very frequently asked.
