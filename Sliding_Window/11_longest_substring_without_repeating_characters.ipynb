# Sliding Window Problem 11 — Longest Substring Without Repeating Characters

## Problem Statement

Given a string `s`,
find the length of the longest substring without repeating characters.

Example:

s = "abcabcbb"

Output:
3

Explanation:
"abc" is the longest substring

# Brute Force Approach

Check every possible substring.

For each substring,
verify whether all characters are unique.

Track the maximum length.

```python
def longest_unique_bruteforce(s):
    max_length = 0

    for i in range(len(s)):
        seen = set()

        for j in range(i, len(s)):
            if s[j] in seen:
                break

            seen.add(s[j])
            max_length = max(max_length, j - i + 1)

    return max_length


print(longest_unique_bruteforce("abcabcbb"))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Too slow for long strings.

# Optimal Idea — Variable Size Sliding Window + Set

Maintain a window with unique characters.

- expand right
- if duplicate found → shrink from left
- keep window unique

# Window Dry Run

s = "abcabcbb"

| left | right | window | max |
|------|-------|--------|-----|
| 0 | 0 | a | 1 |
| 0 | 1 | ab | 2 |
| 0 | 2 | abc | 3 |
| 0 | 3 | abca | duplicate |
| 1 | 3 | bca | 3 |

```python
class Solution:
    def lengthOfLongestSubstring(self, s):
        window = set()
        left = 0
        max_length = 0

        for right in range(len(s)):
            while s[right] in window:
                window.remove(s[left])
                left += 1

            window.add(s[right])
            max_length = max(max_length, right - left + 1)

        return max_length
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(min(n, charset))
