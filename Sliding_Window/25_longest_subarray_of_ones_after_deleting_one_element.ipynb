# Sliding Window Problem 25 — Longest Subarray of 1's After Deleting One Element

## Problem Statement

Given a binary array `nums`,
return the length of the longest subarray containing only `1`s
after deleting exactly one element.

Example:

nums = [1, 1, 0, 1]

Output:
3

# Brute Force Approach

Delete each element one by one.

For every deletion,
find the longest continuous subarray of 1's.

Track the maximum.

```python
def longest_ones_delete_bruteforce(nums):
    max_length = 0

    for i in range(len(nums)):
        temp = nums[:i] + nums[i + 1:]

        current = 0
        best = 0

        for num in temp:
            if num == 1:
                current += 1
                best = max(best, current)
            else:
                current = 0

        max_length = max(max_length, best)

    return max_length


print(longest_ones_delete_bruteforce([1, 1, 0, 1]))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Too slow for large arrays.

# Optimal Idea — Sliding Window with At Most One Zero

This is an **at most 1 zero** problem.

Maintain a window containing at most one `0`.

The final answer is:

window size - 1

# Why Minus 1?

Because the problem says:

delete exactly one element

So even if the window is all 1's,
one element must still be removed.

```python
class Solution:
    def longestSubarray(self, nums):
        left = 0
        zero_count = 0
        max_length = 0

        for right in range(len(nums)):
            if nums[right] == 0:
                zero_count += 1

            while zero_count > 1:
                if nums[left] == 0:
                    zero_count -= 1
                left += 1

            max_length = max(max_length, right - left)

        return max_length
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(1)

# Pattern Takeaway

Excellent **at most one violation sliding window** problem.

Very frequently asked.
