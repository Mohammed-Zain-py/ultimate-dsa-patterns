# Sliding Window Problem 5 — Maximum Average Subarray I

## Problem Statement

Given an integer array `nums` and an integer `k`,
find the contiguous subarray of size `k`
that has the maximum average value.

Return the maximum average.

Example:

nums = [1, 12, -5, -6, 50, 3]
k = 4

Output:
12.75

# Brute Force Approach

Check every subarray of size `k`.

For each window:
- calculate sum
- divide by `k`
- track maximum average

```python
def max_average_bruteforce(nums, k):
    max_avg = float("-inf")

    for i in range(len(nums) - k + 1):
        current_sum = 0

        for j in range(i, i + k):
            current_sum += nums[j]

        avg = current_sum / k
        max_avg = max(max_avg, avg)

    return max_avg


print(max_average_bruteforce([1, 12, -5, -6, 50, 3], 4))
```

# Complexity of Brute Force

Time Complexity: O(n * k)

Repeated summation makes it slow.

# Optimal Idea — Fixed Size Sliding Window

Instead of recalculating sum for every window,

reuse previous sum.

window_sum = old_sum - outgoing + incoming

# Window Dry Run

nums = [1, 12, -5, -6, 50, 3]
k = 4

| window | sum | average |
|--------|-----|---------|
| [1,12,-5,-6] | 2 | 0.5 |
| [12,-5,-6,50] | 51 | 12.75 |
| [-5,-6,50,3] | 42 | 10.5 |

```python

class Solution:
    def findMaxAverage(self, nums, k):
        window_sum = sum(nums[:k])
        max_sum = window_sum

        for right in range(k, len(nums)):
            window_sum += nums[right]
            window_sum -= nums[right - k]

            max_sum = max(max_sum, window_sum)

        return max_sum / k
```

# Complexity Comparison

Brute Force:
- Time → O(n * k)

Optimal:
- Time → O(n)
- Space → O(1)

# Pattern Takeaway

Classic **fixed-size sliding window sum / average problem**.

Very commonly asked.
