# Sliding Window Problem 28 — Grumpy Bookstore Owner

## Problem Statement

The owner of a bookstore is grumpy for some minutes.

If the owner is grumpy, customers in that minute are unsatisfied.

The owner can use a secret technique for exactly `minutes`
to keep themselves not grumpy.

Return the maximum number of satisfied customers.

Example:

customers = [1,0,1,2,1,1,7,5]
grumpy    = [0,1,0,1,0,1,0,1]
minutes = 3

Output:
16

# Brute Force Approach

Try applying the secret technique
for every possible window of size `minutes`.

Calculate total satisfied customers each time.

```python
def max_satisfied_bruteforce(customers, grumpy, minutes):
    max_satisfied = 0
    n = len(customers)

    for start in range(n - minutes + 1):
        current = 0

        for i in range(n):
            if grumpy[i] == 0 or (start <= i < start + minutes):
                current += customers[i]

        max_satisfied = max(max_satisfied, current)

    return max_satisfied


print(max_satisfied_bruteforce(
    [1,0,1,2,1,1,7,5],
    [0,1,0,1,0,1,0,1],
    3
))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Too slow for large inputs.

# Optimal Idea — Fixed Size Sliding Window

Split the answer into 2 parts:

1. always satisfied customers
2. extra customers saved by the technique window

# Key Formula

answer = base satisfied + max extra window

```python
class Solution:
    def maxSatisfied(self, customers, grumpy, minutes):
        base = 0

        for i in range(len(customers)):
            if grumpy[i] == 0:
                base += customers[i]

        extra = 0

        for i in range(minutes):
            if grumpy[i] == 1:
                extra += customers[i]

        max_extra = extra

        for right in range(minutes, len(customers)):
            if grumpy[right] == 1:
                extra += customers[right]

            if grumpy[right - minutes] == 1:
                extra -= customers[right - minutes]

            max_extra = max(max_extra, extra)

        return base + max_extra
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(1)
