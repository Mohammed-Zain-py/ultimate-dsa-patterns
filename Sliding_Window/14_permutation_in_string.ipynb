# Sliding Window Problem 14 — Permutation in String

## Problem Statement

Given two strings `s1` and `s2`,
return `True` if `s2` contains any permutation of `s1`.

Example:

s1 = "ab"
s2 = "eidbaooo"

Output:
True

Explanation:
"ba" is a permutation of "ab"

# Brute Force Approach

Generate every substring of size `len(s1)` from `s2`.

Sort and compare with sorted `s1`.

If any match is found, return `True`.

```python
def check_permutation_bruteforce(s1, s2):
    k = len(s1)

    for i in range(len(s2) - k + 1):
        window = s2[i:i + k]

        if sorted(window) == sorted(s1):
            return True

    return False


print(check_permutation_bruteforce("ab", "eidbaooo"))
```

# Complexity of Brute Force

Time Complexity: O(n * k log k)

Sorting every window is expensive.

# Optimal Idea — Sliding Window + Frequency Map

Use a fixed-size window of length `len(s1)`.

Compare frequency counts instead of sorting.

# Window Dry Run

s1 = "ab"
s2 = "eidbaooo"

Window size = 2

| window | permutation? |
|--------|-------------|
| ei | no |
| id | no |
| db | no |
| ba | yes |

```python
class Solution:
    def checkInclusion(self, s1, s2):
        k = len(s1)

        if k > len(s2):
            return False

        target = Counter(s1)
        window = Counter()

        for i in range(len(s2)):
            window[s2[i]] += 1

            if i >= k:
                left_char = s2[i - k]
                window[left_char] -= 1

                if window[left_char] == 0:
                    del window[left_char]

            if i >= k - 1:
                if window == target:
                    return True

        return False
```

# Complexity Comparison

Brute Force:
- Time → O(n * k log k)

Optimal:
- Time → O(n)
- Space → O(26)
