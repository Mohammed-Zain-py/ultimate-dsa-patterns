# Sliding Window Problem 2 — Maximum Sum Subarray of Size K

## Problem Statement

Given an array `nums` and an integer `k`, find the maximum sum of any contiguous subarray of size `k`.

Example:

nums = [2, 1, 5, 1, 3, 2]
k = 3

Output:
9

Explanation:
Subarray [5, 1, 3] gives the maximum sum = 9

# Brute Force Approach

Check every possible subarray of size `k`.

For each window, calculate the sum separately.

This works but repeats a lot of work.

```python
def max_sum_subarray_bruteforce(nums, k):
    max_sum = float("-inf")

    for i in range(len(nums) - k + 1):
        current_sum = 0

        for j in range(i, i + k):
            current_sum += nums[j]

        max_sum = max(max_sum, current_sum)

    return max_sum


print(max_sum_subarray_bruteforce([2, 1, 5, 1, 3, 2], 3))
```

# Complexity of Brute Force

Time Complexity: O(n * k)

Because for every starting index, we sum `k` elements again.

# Optimal Idea — Fixed Size Sliding Window

Instead of recalculating the entire sum,

reuse the previous window sum.

When window slides:

new_sum = old_sum - outgoing + incoming

# Window Movement Dry Run

nums = [2, 1, 5, 1, 3, 2]
k = 3

| window | sum |
|--------|-----|
| [2,1,5] | 8 |
| [1,5,1] | 7 |
| [5,1,3] | 9 |
| [1,3,2] | 6 |

```python

class Solution:
    def maxSumSubarray(self, nums, k):
        window_sum = sum(nums[:k])
        max_sum = window_sum

        for right in range(k, len(nums)):
            window_sum += nums[right]
            window_sum -= nums[right - k]

            max_sum = max(max_sum, window_sum)

        return max_sum
```

# Complexity Comparison

Brute Force:
- Time → O(n * k)

Optimal:
- Time → O(n)
- Space → O(1)
