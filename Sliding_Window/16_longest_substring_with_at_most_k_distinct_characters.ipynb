# Sliding Window Problem 16 — Longest Substring with At Most K Distinct Characters

## Problem Statement

Given a string `s` and an integer `k`,
return the length of the longest substring
that contains at most `k` distinct characters.

Example:

s = "eceba"
k = 2

Output:
3

Explanation:
"ece" is the longest valid substring

# Brute Force Approach

Check every possible substring.

For each substring,
count distinct characters.

If distinct count <= k,
update maximum length.

```python
def longest_k_distinct_bruteforce(s, k):
    max_length = 0

    for i in range(len(s)):
        distinct = set()

        for j in range(i, len(s)):
            distinct.add(s[j])

            if len(distinct) <= k:
                max_length = max(max_length, j - i + 1)
            else:
                break

    return max_length


print(longest_k_distinct_bruteforce("eceba", 2))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Too slow for long strings.

# Optimal Idea — Variable Size Sliding Window + Frequency Map

Maintain a window with at most `k` distinct characters.

- expand right
- if distinct > k → shrink left
- track maximum valid length

# Window Dry Run

s = "eceba"
k = 2

| left | right | window | distinct |
|------|-------|--------|---------|
| 0 | 0 | e | 1 |
| 0 | 1 | ec | 2 |
| 0 | 2 | ece | 2 |
| 0 | 3 | eceb | 3 |
| 1 | 3 | ceb | 3 |
| 2 | 3 | eb | 2 |

```python
class Solution:
    def lengthOfLongestSubstringKDistinct(self, s, k):
        freq = defaultdict(int)
        left = 0
        max_length = 0

        for right in range(len(s)):
            freq[s[right]] += 1

            while len(freq) > k:
                freq[s[left]] -= 1

                if freq[s[left]] == 0:
                    del freq[s[left]]

                left += 1

            max_length = max(max_length, right - left + 1)

        return max_length
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(k)

# Pattern Takeaway
