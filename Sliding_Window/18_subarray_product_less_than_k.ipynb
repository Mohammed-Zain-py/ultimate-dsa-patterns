# Sliding Window Problem 18 — Subarray Product Less Than K

## Problem Statement

Given an array of positive integers `nums`
and an integer `k`,
return the number of contiguous subarrays
where the product of all elements is strictly less than `k`.

Example:

nums = [10, 5, 2, 6]
k = 100

Output:
8

# Brute Force Approach

Check every possible subarray.

For each subarray,
calculate the product.

If product < k,
increase the count.

```python
def count_subarrays_bruteforce(nums, k):
    count = 0

    for i in range(len(nums)):
        product = 1

        for j in range(i, len(nums)):
            product *= nums[j]

            if product < k:
                count += 1
            else:
                break

    return count


print(count_subarrays_bruteforce([10, 5, 2, 6], 100))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Too slow for larger arrays.

# Optimal Idea — Variable Size Sliding Window

Maintain a window where product < k.

- expand right
- if product >= k → shrink left
- add valid subarrays ending at right

# Key Observation

For every valid window:

number of valid subarrays ending at `right`

= window size

# Window Dry Run

nums = [10,5,2,6]
k = 100

| left | right | product | valid subarrays added |
|------|-------|---------|----------------------|
| 0 | 0 | 10 | 1 |
| 0 | 1 | 50 | 2 |
| 0 | 2 | 100 | shrink |
| 1 | 2 | 10 | 2 |
| 1 | 3 | 60 | 3 |

```python
class Solution:
    def numSubarrayProductLessThanK(self, nums, k):
        if k <= 1:
            return 0

        left = 0
        product = 1
        count = 0

        for right in range(len(nums)):
            product *= nums[right]

            while product >= k:
                product //= nums[left]
                left += 1

            count += right - left + 1

        return count
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(1)

# Pattern Takeaway

This is one of the most important **counting + variable-size sliding window** problems.

Very frequently asked.
