# Sliding Window Problem 7 — Contains Duplicate II

## Problem Statement

Given an integer array `nums` and an integer `k`,
return `True` if there are two distinct indices `i` and `j`
such that:

- nums[i] == nums[j]
- |i - j| <= k

Example:

nums = [1, 2, 3, 1]
k = 3

Output:
True

# Brute Force Approach

For every element,
check the next `k` elements.

If a duplicate is found, return `True`.

```python
def contains_duplicate_bruteforce(nums, k):
    for i in range(len(nums)):
        for j in range(i + 1, min(i + k + 1, len(nums))):
            if nums[i] == nums[j]:
                return True

    return False


print(contains_duplicate_bruteforce([1, 2, 3, 1], 3))
```

# Complexity of Brute Force

Time Complexity: O(n * k)

This becomes slow for large `k`.

# Optimal Idea — Fixed Size Sliding Window + Set

Maintain a window of size at most `k`.

Use a set to store current window elements.

If incoming element already exists → duplicate found.

# Window Dry Run

nums = [1, 2, 3, 1]
k = 3

| window set | incoming | result |
|-----------|----------|--------|
| {1} | 2 | continue |
| {1,2} | 3 | continue |
| {1,2,3} | 1 | duplicate |

```python
class Solution:
    def containsNearbyDuplicate(self, nums, k):
        window = set()

        for i in range(len(nums)):
            if nums[i] in window:
                return True

            window.add(nums[i])

            if len(window) > k:
                window.remove(nums[i - k])

        return False
```

# Complexity Comparison

Brute Force:
- Time → O(n * k)

Optimal:
- Time → O(n)
- Space → O(k)

# Pattern Takeaway

Excellent **fixed-size sliding window + set** problem.
