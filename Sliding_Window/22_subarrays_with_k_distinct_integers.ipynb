# Sliding Window Problem 22 — Subarrays with K Distinct Integers

## Problem Statement

Given an integer array `nums` and an integer `k`,
return the number of good subarrays.

A good subarray contains exactly `k` distinct integers.

Example:

nums = [1, 2, 1, 2, 3]
k = 2

Output:
7

# Brute Force Approach

Check every possible subarray.

Count distinct integers.

If distinct count == k,
increase the answer.

```python
def count_exactly_k_bruteforce(nums, k):
    count = 0

    for i in range(len(nums)):
        distinct = set()

        for j in range(i, len(nums)):
            distinct.add(nums[j])

            if len(distinct) == k:
                count += 1
            elif len(distinct) > k:
                break

    return count


print(count_exactly_k_bruteforce([1, 2, 1, 2, 3], 2))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Too slow for large arrays.

# Optimal Idea — Exactly K = At Most K - At Most (K - 1)

This is the most important trick.

exactly(k) = at_most(k) - at_most(k-1)

# Why This Works

Subarrays with exactly `k` distinct elements

= all subarrays with at most `k`
minus
all subarrays with at most `k-1`

```python
from collections import defaultdict

class Solution:
    def subarraysWithKDistinct(self, nums, k):
        def at_most(k):
            freq = defaultdict(int)
            left = 0
            count = 0

            for right in range(len(nums)):
                freq[nums[right]] += 1

                while len(freq) > k:
                    freq[nums[left]] -= 1

                    if freq[nums[left]] == 0:
                        del freq[nums[left]]

                    left += 1

                count += right - left + 1

            return count

        return at_most(k) - at_most(k - 1)
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(k)

# Pattern Takeaway

This is one of the most important **medium-hard exact K sliding window problems**.
