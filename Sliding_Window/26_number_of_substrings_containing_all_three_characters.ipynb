# Sliding Window Problem 26 — Number of Substrings Containing All Three Characters

## Problem Statement

Given a string `s` consisting only of characters `a`, `b`, and `c`,
return the number of substrings containing at least one occurrence of all three characters.

Example:

s = "abcabc"

Output:
10

# Brute Force Approach

Check every possible substring.

For each substring,
verify whether it contains `a`, `b`, and `c`.

If yes, increase count.

```python
def count_substrings_bruteforce(s):
    count = 0

    for i in range(len(s)):
        chars = set()

        for j in range(i, len(s)):
            chars.add(s[j])

            if len(chars) == 3:
                count += 1

    return count


print(count_substrings_bruteforce("abcabc"))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Too slow for long strings.

# Optimal Idea — Variable Size Sliding Window

Maintain a window that contains all three characters.

Whenever valid:
all substrings ending at `right`
and starting from `left` onward are valid.

# Key Observation

If window `[left ... right]` is valid,

then all substrings:

[left ... right]
[left+1 ... right]
[left+2 ... right]

are also valid.

Count added = len(s) - right

```python
from collections import defaultdict

class Solution:
    def numberOfSubstrings(self, s):
        freq = defaultdict(int)
        left = 0
        count = 0
        n = len(s)

        for right in range(n):
            freq[s[right]] += 1

            while len(freq) == 3:
                count += n - right

                freq[s[left]] -= 1

                if freq[s[left]] == 0:
                    del freq[s[left]]

                left += 1

        return count
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(3)
