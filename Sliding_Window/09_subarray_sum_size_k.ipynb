# Sliding Window Problem 9 — Sum of Every Subarray of Size K

## Problem Statement

Given an array `nums` and an integer `k`,
return a list containing the sum of every contiguous subarray of size `k`.

Example:

nums = [1, 2, 3, 4, 5]
k = 3

Output:
[6, 9, 12]

# Brute Force Approach

For every starting index,
sum the next `k` elements.

Store the sum in the result list.

```python
def subarray_sum_bruteforce(nums, k):
    result = []

    for i in range(len(nums) - k + 1):
        current_sum = 0

        for j in range(i, i + k):
            current_sum += nums[j]

        result.append(current_sum)

    return result


print(subarray_sum_bruteforce([1, 2, 3, 4, 5], 3))
```

# Complexity of Brute Force

Time Complexity: O(n * k)

Repeated summation causes extra work.

# Optimal Idea — Fixed Size Sliding Window

Reuse the previous window sum.

When sliding:
- remove outgoing element
- add incoming element

# Window Dry Run

nums = [1, 2, 3, 4, 5]
k = 3

| window | sum |
|--------|-----|
| [1,2,3] | 6 |
| [2,3,4] | 9 |
| [3,4,5] | 12 |

```python
class Solution:
    def subarraySums(self, nums, k):
        result = []

        window_sum = sum(nums[:k])
        result.append(window_sum)

        for right in range(k, len(nums)):
            window_sum += nums[right]
            window_sum -= nums[right - k]

            result.append(window_sum)

        return result
```

# Complexity Comparison

Brute Force:
- Time → O(n * k)

Optimal:
- Time → O(n)
- Space → O(n)
