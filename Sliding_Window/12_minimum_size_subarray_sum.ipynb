# Sliding Window Problem 12 — Minimum Size Subarray Sum

## Problem Statement

Given an array of positive integers `nums`
and an integer `target`,
return the minimum length of a contiguous subarray
whose sum is greater than or equal to `target`.

If no such subarray exists, return 0.

Example:

nums = [2, 3, 1, 2, 4, 3]
target = 7

Output:
2

Explanation:
[4, 3] is the minimum valid subarray

# Brute Force Approach

Check every possible subarray.

For each subarray,
calculate the sum.

If sum >= target,
update minimum length.

```python
def min_subarray_bruteforce(nums, target):
    min_length = float("inf")

    for i in range(len(nums)):
        current_sum = 0

        for j in range(i, len(nums)):
            current_sum += nums[j]

            if current_sum >= target:
                min_length = min(min_length, j - i + 1)
                break

    return 0 if min_length == float("inf") else min_length


print(min_subarray_bruteforce([2, 3, 1, 2, 4, 3], 7))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Too slow for larger arrays.

# Optimal Idea — Variable Size Sliding Window

- expand right until sum >= target
- shrink from left while still valid
- track minimum window length

# Window Dry Run

nums = [2,3,1,2,4,3]
target = 7

| left | right | sum | length |
|------|-------|-----|--------|
| 0 | 0 | 2 | - |
| 0 | 1 | 5 | - |
| 0 | 2 | 6 | - |
| 0 | 3 | 8 | 4 |
| 1 | 3 | 6 | shrink stop |
| 1 | 4 | 10 | 4 |
| 2 | 4 | 7 | 3 |
| 3 | 4 | 6 | stop |
| 3 | 5 | 9 | 3 |
| 4 | 5 | 7 | 2 |

```python
class Solution:
    def minSubArrayLen(self, target, nums):
        left = 0
        current_sum = 0
        min_length = float("inf")

        for right in range(len(nums)):
            current_sum += nums[right]

            while current_sum >= target:
                min_length = min(min_length, right - left + 1)
                current_sum -= nums[left]
                left += 1

        return 0 if min_length == float("inf") else min_length
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(1)

# Pattern Takeaway

This is one of the most important **variable-size sliding window minimum problems**.

Very frequently asked.
