# Sliding Window Problem 20 — Minimum Window Substring

## Problem Statement

Given two strings `s` and `t`,
return the minimum window substring of `s`
such that every character in `t`
(including duplicates) is included.

If no such window exists, return an empty string.

Example:

s = "ADOBECODEBANC"
t = "ABC"

Output:
"BANC"

# Brute Force Approach

Check every possible substring.

For each substring,
verify whether all characters of `t` are present.

Track the minimum valid length.

```python
from collections import Counter

def min_window_bruteforce(s, t):
    target = Counter(t)
    min_window = ""

    for i in range(len(s)):
        freq = Counter()

        for j in range(i, len(s)):
            freq[s[j]] += 1

            valid = True
            for ch in target:
                if freq[ch] < target[ch]:
                    valid = False
                    break

            if valid:
                window = s[i:j + 1]

                if min_window == "" or len(window) < len(min_window):
                    min_window = window
                break

    return min_window


print(min_window_bruteforce("ADOBECODEBANC", "ABC"))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Very slow for long strings.

# Optimal Idea — Variable Size Sliding Window + Frequency Map

- expand right until all required characters are included
- then shrink from left to find the minimum valid window

# Key Observation

Use a `need` counter.

When all characters are matched:

matched == required

# Window Dry Run

s = "ADOBECODEBANC"
t = "ABC"

Valid windows found:
- ADOBEC
- CODEBA
- BANC

Minimum = BANC

```python
from collections import Counter, defaultdict

class Solution:
    def minWindow(self, s, t):
        need = Counter(t)
        window = defaultdict(int)

        required = len(need)
        formed = 0

        left = 0
        min_len = float("inf")
        start = 0

        for right in range(len(s)):
            char = s[right]
            window[char] += 1

            if char in need and window[char] == need[char]:
                formed += 1

            while formed == required:
                if right - left + 1 < min_len:
                    min_len = right - left + 1
                    start = left

                left_char = s[left]
                window[left_char] -= 1

                if left_char in need and window[left_char] < need[left_char]:
                    formed -= 1

                left += 1

        return "" if min_len == float("inf") else s[start:start + min_len]
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(52)
