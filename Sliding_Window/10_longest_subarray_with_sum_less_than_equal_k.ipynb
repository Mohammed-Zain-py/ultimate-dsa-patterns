# Sliding Window Problem 10 — Longest Subarray with Sum Less Than or Equal to K

## Problem Statement

Given an array of positive integers `nums` and an integer `k`,
find the length of the longest contiguous subarray
whose sum is less than or equal to `k`.

Example:

nums = [1, 2, 1, 0, 1, 1, 0]
k = 4

Output:
5

# Brute Force Approach

Check every possible subarray.

For each subarray, calculate the sum.

If sum <= k, update maximum length.

```python
def longest_subarray_bruteforce(nums, k):
    max_length = 0

    for i in range(len(nums)):
        current_sum = 0

        for j in range(i, len(nums)):
            current_sum += nums[j]

            if current_sum <= k:
                max_length = max(max_length, j - i + 1)

    return max_length


print(longest_subarray_bruteforce([1, 2, 1, 0, 1, 1, 0], 4))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Too slow for large arrays.

# Optimal Idea — Variable Size Sliding Window

Now the window size changes.

- expand right
- if sum > k, shrink from left
- track maximum valid length

# Window Dry Run

nums = [1,2,1,0,1,1,0]
k = 4

| left | right | window sum | length |
|------|-------|-----------|--------|
| 0 | 0 | 1 | 1 |
| 0 | 1 | 3 | 2 |
| 0 | 2 | 4 | 3 |
| 0 | 3 | 4 | 4 |
| 0 | 4 | 5 | shrink |
| 1 | 4 | 4 | 4 |
| 1 | 6 | 4 | 6 |

```python
class Solution:
    def longestSubarray(self, nums, k):
        left = 0
        current_sum = 0
        max_length = 0

        for right in range(len(nums)):
            current_sum += nums[right]

            while current_sum > k:
                current_sum -= nums[left]
                left += 1

            max_length = max(max_length, right - left + 1)

        return max_length
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(1)

# Pattern Takeaway

This is the perfect **bridge from fixed-size to variable-size sliding window**.

Very important conceptually.
