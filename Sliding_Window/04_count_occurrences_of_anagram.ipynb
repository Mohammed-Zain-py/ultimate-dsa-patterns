# Sliding Window Problem 4 — Count Occurrences of Anagram

## Problem Statement

Given a string `text` and a string `pattern`, count how many substrings of `text`
are anagrams of `pattern`.

Example:

text = "forxxorfxdofr"
pattern = "for"

Output:
3

# Brute Force Approach

Generate every substring of size equal to the pattern length.

Sort both strings and compare.

If equal → anagram found.

```python
def count_anagrams_bruteforce(text, pattern):
    count = 0
    k = len(pattern)

    for i in range(len(text) - k + 1):
        window = text[i:i + k]

        if sorted(window) == sorted(pattern):
            count += 1

    return count


print(count_anagrams_bruteforce("forxxorfxdofr", "for"))
```

# Complexity of Brute Force

Time Complexity: O(n * k log k)

Sorting every window is expensive.

# Optimal Idea — Fixed Window + Frequency Map

Use a frequency dictionary for the pattern.

Maintain a sliding window of size `k`.

Compare frequency maps.

# Key Observation

Two strings are anagrams if their frequency counts are equal.

# Window Dry Run

text = "forxxorfxdofr"
pattern = "for"

Window size = 3

| window | anagram? |
|--------|----------|
| for | yes |
| orx | no |
| rxx | no |
| xxo | no |
| xor | no |
| orf | yes |
| rfx | no |
| xdo | no |
| dof | no |
| ofr | yes |

```python
from collections import Counter

class Solution:
    def countAnagrams(self, text, pattern):
        k = len(pattern)
        pattern_count = Counter(pattern)
        window_count = Counter()

        count = 0

        for i in range(len(text)):
            window_count[text[i]] += 1

            if i >= k:
                left_char = text[i - k]
                window_count[left_char] -= 1

                if window_count[left_char] == 0:
                    del window_count[left_char]

            if i >= k - 1:
                if window_count == pattern_count:
                    count += 1

        return count
```

# Complexity Comparison

Brute Force:
- Time → O(n * k log k)

Optimal:
- Time → O(n * 26) ≈ O(n)
- Space → O(26)

# Pattern Takeaway

Classic **fixed-size sliding window + hashmap** problem.
