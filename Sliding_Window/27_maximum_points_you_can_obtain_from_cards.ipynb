# Sliding Window Problem 27 — Maximum Points You Can Obtain from Cards

## Problem Statement

You are given an integer array `cardPoints`
and an integer `k`.

You can take exactly `k` cards
from the beginning or the end.

Return the maximum score.

Example:

cardPoints = [1,2,3,4,5,6,1]
k = 3

Output:
12

# Brute Force Approach

Try every possible split:

- take `i` cards from left
- take `k-i` cards from right

Find the maximum total.

```python
def max_score_bruteforce(cardPoints, k):
    max_score = 0
    n = len(cardPoints)

    for i in range(k + 1):
        left_sum = sum(cardPoints[:i])
        right_sum = sum(cardPoints[n - (k - i):])

        max_score = max(max_score, left_sum + right_sum)

    return max_score


print(max_score_bruteforce([1,2,3,4,5,6,1], 3))
```

# Complexity of Brute Force

Time Complexity: O(k²)

Repeated summation increases cost.

# Optimal Idea — Convert to Minimum Subarray Problem

Instead of taking `k` cards,

leave `n-k` cards in the middle.

So:

maximum score = total sum - minimum window sum of size (n-k)

# Why This Works

Taking cards from ends

is equivalent to

removing one continuous middle subarray.

```python
class Solution:
    def maxScore(self, cardPoints, k):
        n = len(cardPoints)

        if k == n:
            return sum(cardPoints)

        window_size = n - k
        total_sum = sum(cardPoints)

        window_sum = sum(cardPoints[:window_size])
        min_sum = window_sum

        for right in range(window_size, n):
            window_sum += cardPoints[right]
            window_sum -= cardPoints[right - window_size]

            min_sum = min(min_sum, window_sum)

        return total_sum - min_sum
```

# Complexity Comparison

Brute Force:
- Time → O(k²)

Optimal:
- Time → O(n)
- Space → O(1)

# Pattern Takeaway

Excellent **transformation-based sliding window** problem.

Very frequently asked.
