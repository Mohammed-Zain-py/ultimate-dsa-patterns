# Sliding Window Problem 6 — Maximum Number of Vowels in a Substring of Given Length

## Problem Statement

Given a string `s` and an integer `k`,
return the maximum number of vowels in any substring of length `k`.

Example:

s = "abciiidef"
k = 3

Output:
3

# Brute Force Approach

Check every substring of size `k`.

For each window, count vowels manually.

Track the maximum count.

```python
def max_vowels_bruteforce(s, k):
    vowels = "aeiou"
    max_count = 0

    for i in range(len(s) - k + 1):
        count = 0

        for j in range(i, i + k):
            if s[j] in vowels:
                count += 1

        max_count = max(max_count, count)

    return max_count


print(max_vowels_bruteforce("abciiidef", 3))
```

# Complexity of Brute Force

Time Complexity: O(n * k)

Repeated counting makes it slower.

# Optimal Idea — Fixed Size Sliding Window

Keep track of vowel count in the current window.

When window slides:
- subtract outgoing character if vowel
- add incoming character if vowel

# Window Dry Run

s = "abciiidef"
k = 3

| window | vowel count |
|--------|------------|
| abc | 1 |
| bci | 1 |
| cii | 2 |
| iii | 3 |
| iid | 2 |
| ide | 2 |
| def | 1 |

```python
class Solution:
    def maxVowels(self, s, k):
        vowels = set("aeiou")
        count = 0

        for i in range(k):
            if s[i] in vowels:
                count += 1

        max_count = count

        for right in range(k, len(s)):
            if s[right] in vowels:
                count += 1

            if s[right - k] in vowels:
                count -= 1

            max_count = max(max_count, count)

        return max_count
```

# Complexity Comparison

Brute Force:
- Time → O(n * k)

Optimal:
- Time → O(n)
- Space → O(1)

# Pattern Takeaway

Excellent **fixed-size string sliding window** problem.
