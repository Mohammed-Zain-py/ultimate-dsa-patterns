# How to Identify Sliding Window Problems

This notebook is one of the most important notebooks in the Sliding Window module.

The biggest goal is to recognize:

**When should we think Sliding Window?**

# The biggest signal

Whenever the problem talks about:

- subarray
- substring
- continuous segment
- window of size k
- longest / shortest continuous range

immediately think:

**Can Sliding Window be used?**

# Recognition Checklist

Ask these questions:

1. Is the range continuous?
2. Is the size fixed?
3. Does the size change based on condition?
4. Is longest / minimum asked?
5. Is frequency involved?

# Fixed Size Signals

Examples of phrases:

- size k
- every window of size k
- average of k elements
- sum of k consecutive numbers

These usually indicate:

**Fixed Size Sliding Window**

# Variable Size Signals

Examples:

- longest substring
- smallest subarray
- maximum length with condition
- at most k distinct
- exactly k

# Frequency Map Signals

Examples:

- anagram
- repeating characters
- distinct characters
- character count condition

# Quick Comparison

If it asks for:
- pair → Two Pointer
- continuous range → Sliding Window

# Common Interview Keywords

Watch for these words:

- longest
- shortest
- maximum
- minimum
- substring
- subarray
- continuous

# Final Takeaway

Sliding Window is all about avoiding repeated recalculation of overlapping ranges.

This is what gives O(n) optimization.
