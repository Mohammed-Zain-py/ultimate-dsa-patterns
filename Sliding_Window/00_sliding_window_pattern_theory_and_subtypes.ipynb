# Sliding Window Pattern — Theory and Subtypes

Sliding Window is one of the most important DSA patterns for coding interviews.

It is mainly used for:

- subarrays
- substrings
- continuous segments
- fixed size ranges
- variable size ranges

This pattern helps reduce repeated work and often optimizes:

O(n²) → O(n)

# The Core Idea

Instead of checking every possible subarray or substring,

we maintain a **window**

and slide it intelligently.

A window is simply a continuous range:

[left ... right]

# Why is it called Sliding Window?

Because the range keeps moving.

Example:

[1, 2, 3] → [2, 3, 4] → [3, 4, 5]

The window slides by moving left and right pointers.

# Subtype 1 — Fixed Size Window

The window size remains constant.

Example:

Find maximum sum subarray of size k

For k = 3

[1,2,3] → [2,3,4] → [3,4,5]

Common problems:

- max sum of size k
- average of size k
- first negative in every window

# Subtype 2 — Variable Size Window

The window size changes based on conditions.

Example:

Longest substring without repeating characters

Common problems:

- longest substring
- shortest subarray
- minimum window

# Subtype 3 — Frequency Map Window

Used when character frequency matters.

Usually combined with dictionary / hashmap.

Examples:

- anagram search
- minimum window substring

# Difference from Two Pointers

Two Pointer:
- pair / opposite ends
- in-place modification

Sliding Window:
- continuous segment
- range based optimization
