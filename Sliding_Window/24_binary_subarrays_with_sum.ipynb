# Sliding Window Problem 24 — Binary Subarrays With Sum

## Problem Statement

Given a binary array `nums` and an integer `goal`,
return the number of non-empty subarrays with sum exactly equal to `goal`.

Example:

nums = [1, 0, 1, 0, 1]
goal = 2

Output:
4

# Brute Force Approach

Check every possible subarray.

Calculate sum.

If sum == goal,
increase answer.

```python
def binary_sum_bruteforce(nums, goal):
    count = 0

    for i in range(len(nums)):
        current_sum = 0

        for j in range(i, len(nums)):
            current_sum += nums[j]

            if current_sum == goal:
                count += 1
            elif current_sum > goal:
                break

    return count


print(binary_sum_bruteforce([1, 0, 1, 0, 1], 2))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Too slow for large arrays.

# Optimal Idea — Exactly Goal Trick

Again use:

exact(goal) = at_most(goal) - at_most(goal - 1)

# Why This Works

All subarrays with exact sum = goal

= all subarrays with sum <= goal
minus
all subarrays with sum <= goal-1

```python
class Solution:
    def numSubarraysWithSum(self, nums, goal):
        def at_most(goal):
            if goal < 0:
                return 0

            left = 0
            current_sum = 0
            count = 0

            for right in range(len(nums)):
                current_sum += nums[right]

                while current_sum > goal:
                    current_sum -= nums[left]
                    left += 1

                count += right - left + 1

            return count

        return at_most(goal) - at_most(goal - 1)
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(1)

# Pattern Takeaway

This is one of the most important **exact sum counting sliding window** problems.

Very frequently asked.
