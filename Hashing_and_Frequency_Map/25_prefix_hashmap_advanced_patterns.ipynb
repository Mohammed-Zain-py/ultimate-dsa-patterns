# Hashing Problem 25 — Prefix Hashmap Advanced Patterns

This notebook summarizes the most important
advanced hashmap interview patterns.

Especially:

- prefix sum
- prefix remainder
- prefix xor

# Pattern 1 — Prefix Sum

Used for:

- subarray sum equals k
- longest zero sum subarray
- cumulative tracking

# Core Formula

If:

prefix_j - prefix_i = k

Then:

prefix_i = prefix_j - k

# Pattern 2 — Prefix Remainder

Used for:

- divisible by k
- repeated remainder problems

# Core Formula

If:

(prefix_j - prefix_i) % k == 0

Then:

prefix_j % k == prefix_i % k

# Pattern 3 — Prefix XOR

Used for:

- xor equals k
- xor matching problems

# Core Formula

If:

prefix_xor ^ previous_xor = k

Then:

previous_xor = prefix_xor ^ k

# Most Important Interview Skill

Do NOT memorize problems.

Memorize:

- pattern recognition
- transformations
- repeated prefix logic

# Common Signals

Watch for:

- subarray count
- divisible by k
- xor equal to k
- longest subarray
- repeated prefix

# Final Takeaway

Prefix hashmap problems are among
the highest-value DSA interview topics.
