# Hashing Problem 21 — Count Subarrays With Equal Number of 0's and 1's

## Problem Statement

Given a binary array `nums`,
count the number of subarrays
having equal numbers of 0's and 1's.

Example:

nums = [0,1,0]

Output:
2

# Brute Force Approach

Generate every possible subarray.

Count:
- number of 0's
- number of 1's

If counts become equal,
increase answer.

```python
def count_equal_zero_one_bruteforce(nums):
    count = 0

    for i in range(len(nums)):
        zeros = 0
        ones = 0

        for j in range(i, len(nums)):
            if nums[j] == 0:
                zeros += 1
            else:
                ones += 1

            if zeros == ones:
                count += 1

    return count


print(count_equal_zero_one_bruteforce([0,1,0]))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Too slow for large arrays.

# Optimal Idea — Convert 0 into -1

Replace:

0 → -1

Now the problem becomes:

find subarrays with sum = 0

# Why This Works

Suppose:

0 → -1
1 → +1

Then:

equal 0's and 1's

means total sum becomes 0.

# Dry Run

nums = [0,1,0]

Converted:

[-1,1,-1]

Prefix sums:

-1
0
-1

Repeated prefix sum found.

Valid subarrays exist.

```python
class Solution:
    def countSubarrWithEqualZeroAndOne(self, nums):
        prefix_count = {
            0: 1
        }

        current_sum = 0
        count = 0

        for num in nums:
            if num == 0:
                current_sum += -1
            else:
                current_sum += 1

            if current_sum in prefix_count:
                count += prefix_count[current_sum]

            prefix_count[current_sum] = (
                prefix_count.get(current_sum, 0) + 1
            )

        return count
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(n)
