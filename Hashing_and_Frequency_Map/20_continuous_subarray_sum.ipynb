# Hashing Problem 20 — Continuous Subarray Sum

## Problem Statement

Given an integer array `nums`
and an integer `k`,

return True if the array has a continuous subarray
of size at least 2
whose sum is a multiple of `k`.

Example:

nums = [23,2,4,6,7]
k = 6

Output:
True

# Brute Force Approach

Generate every possible subarray
of size at least 2.

Calculate the sum.

If sum % k == 0,
return True.

```python
def check_subarray_sum_bruteforce(nums, k):
    n = len(nums)

    for i in range(n):
        current_sum = 0

        for j in range(i, n):
            current_sum += nums[j]

            if j - i + 1 >= 2:
                if current_sum % k == 0:
                    return True

    return False


print(check_subarray_sum_bruteforce([23,2,4,6,7], 6))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Too slow for large inputs.

# Optimal Idea — Prefix Sum Remainder Hashmap

If:

(prefix_sum_j - prefix_sum_i) % k == 0

Then:

prefix_sum_j % k == prefix_sum_i % k

# Key Observation

If the same remainder appears again,

the subarray between them
has sum divisible by k.

# Dry Run

nums = [23,2,4,6,7]
k = 6

Prefix sums:

23 → remainder 5
25 → remainder 1
29 → remainder 5

Remainder 5 appeared before.

Subarray between them:
[2,4]

Sum = 6

```python
class Solution:
    def checkSubarraySum(self, nums, k):
        remainder_index = {
            0: -1
        }

        current_sum = 0

        for i in range(len(nums)):
            current_sum += nums[i]

            remainder = current_sum % k

            if remainder in remainder_index:
                if i - remainder_index[remainder] >= 2:
                    return True
            else:
                remainder_index[remainder] = i

        return False
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(k)

# Pattern Takeaway

This is one of the most important
prefix remainder hashmap problems.

Very commonly asked in interviews.
