# Hashing Problem 18 — Find the Duplicate Number

## Problem Statement

Given an array `nums`
containing `n + 1` integers
where each integer is between `1` and `n` inclusive,

return the duplicate number.

Example:

nums = [1,3,4,2,2]

Output:
2

# Brute Force Approach

Compare every pair.

If two elements are equal,
return that number.

```python
def find_duplicate_bruteforce(nums):
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] == nums[j]:
                return nums[i]


print(find_duplicate_bruteforce([1,3,4,2,2]))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Because every pair is checked.

# Better Idea — Hash Set

Store visited numbers in a set.

If a number already exists,
it is the duplicate.

# Why This Works

Set lookup is O(1).

So duplicate detection becomes very fast.

# Dry Run

nums = [1,3,4,2,2]

Visited:

1
3
4
2

2 already exists → duplicate found

```python
class Solution:
    def findDuplicate(self, nums):
        seen = set()

        for num in nums:
            if num in seen:
                return num

            seen.add(num)
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Hash Set:
- Time → O(n)
- Space → O(n)
