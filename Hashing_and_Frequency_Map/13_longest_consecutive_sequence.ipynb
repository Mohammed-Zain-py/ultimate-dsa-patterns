# Hashing Problem 13 — Longest Consecutive Sequence

## Problem Statement

Given an unsorted array `nums`,
return the length of the longest consecutive sequence.

Example:

nums = [100,4,200,1,3,2]

Output:
4

Explanation:
The sequence [1,2,3,4] has length 4.

# Brute Force Approach

For every number,
keep checking whether the next consecutive number exists.

Track the longest sequence.

```python
def longest_consecutive_bruteforce(nums):
    longest = 0

    for num in nums:
        current = num
        length = 1

        while current + 1 in nums:
            current += 1
            length += 1

        longest = max(longest, length)

    return longest


print(longest_consecutive_bruteforce([100,4,200,1,3,2]))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Because membership checking in a list is O(n).

# Optimal Idea — Hash Set

Store all numbers in a set.

Only start counting when:

number - 1 does NOT exist.

This identifies the beginning of a sequence.

# Key Observation

If `num - 1` exists,

then current number is NOT the start of a sequence.

So skip it.

# Dry Run

nums = [100,4,200,1,3,2]

Set:

{1,2,3,4,100,200}

Sequence starts:
1 → length 4
100 → length 1
200 → length 1

Longest = 4

```python
class Solution:
    def longestConsecutive(self, nums):
        num_set = set(nums)
        longest = 0

        for num in num_set:
            if num - 1 not in num_set:
                current = num
                length = 1

                while current + 1 in num_set:
                    current += 1
                    length += 1

                longest = max(longest, length)

        return longest
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(n)
