# Hashing Problem 16 — Isomorphic Strings

## Problem Statement

Given two strings `s` and `t`,
determine whether they are isomorphic.

Two strings are isomorphic if:

- characters from `s` can be replaced to get `t`
- each character maps to exactly one character
- no two characters map to the same character

Example:

s = "egg"
t = "add"

Output:
True

# Brute Force Thought Process

Try building character mappings manually.

For every character:
check whether mapping remains consistent.

```python
def isomorphic_bruteforce(s, t):
    map_st = {}

    for i in range(len(s)):
        ch1 = s[i]
        ch2 = t[i]

        if ch1 in map_st:
            if map_st[ch1] != ch2:
                return False
        else:
            map_st[ch1] = ch2

    values = []

    for key in map_st:
        if map_st[key] in values:
            return False

        values.append(map_st[key])

    return True


print(isomorphic_bruteforce("egg", "add"))
```

# Problem in Brute Force Logic

Checking duplicate mapped values
using a list is inefficient.

We need faster lookup.

# Optimal Idea — Two Hashmaps

Use:

s → t mapping
t → s mapping

Both directions must remain consistent.

# Why Two Hashmaps?

Example:

a → x
b → x

This is invalid.

Two hashmaps help detect such conflicts instantly.

# Dry Run

s = "egg"
t = "add"

Mappings:

e → a
g → d

Reverse:

a → e
d → g

All mappings remain consistent.

```python
class Solution:
    def isIsomorphic(self, s, t):
        map_st = {}
        map_ts = {}

        for i in range(len(s)):
            ch1 = s[i]
            ch2 = t[i]

            if ch1 in map_st:
                if map_st[ch1] != ch2:
                    return False
            else:
                map_st[ch1] = ch2

            if ch2 in map_ts:
                if map_ts[ch2] != ch1:
                    return False
            else:
                map_ts[ch2] = ch1

        return True
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(n)
