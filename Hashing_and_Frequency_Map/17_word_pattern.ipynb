# Hashing Problem 17 — Word Pattern

## Problem Statement

Given a pattern and a string `s`,
determine whether `s` follows the same pattern.

Example:

pattern = "abba"
s = "dog cat cat dog"

Output:
True

# Brute Force Thought Process

Split the string into words.

Try building mappings manually.

Check whether mappings remain consistent.

```python
def word_pattern_bruteforce(pattern, s):
    words = s.split()

    if len(pattern) != len(words):
        return False

    mapping = {}

    for i in range(len(pattern)):
        ch = pattern[i]
        word = words[i]

        if ch in mapping:
            if mapping[ch] != word:
                return False
        else:
            mapping[ch] = word

    used = []

    for key in mapping:
        if mapping[key] in used:
            return False

        used.append(mapping[key])

    return True


print(word_pattern_bruteforce(
    "abba",
    "dog cat cat dog"
))
```

# Problem in Brute Force Logic

Checking duplicate mapped words
using a list is inefficient.

We need faster reverse checking.

# Optimal Idea — Two Hashmaps

Use:

pattern → word
word → pattern

Both mappings must remain valid.

# Why Two Hashmaps?

Example:

a → dog
b → dog

Invalid.

Two hashmaps detect this instantly.

# Dry Run

pattern = "abba"

Words:
["dog","cat","cat","dog"]

Mappings:

a → dog
b → cat

Reverse mappings:

dog → a
cat → b

```python
class Solution:
    def wordPattern(self, pattern, s):
        words = s.split()

        if len(pattern) != len(words):
            return False

        map_pw = {}
        map_wp = {}

        for i in range(len(pattern)):
            ch = pattern[i]
            word = words[i]

            if ch in map_pw:
                if map_pw[ch] != word:
                    return False
            else:
                map_pw[ch] = word

            if word in map_wp:
                if map_wp[word] != ch:
                    return False
            else:
                map_wp[word] = ch

        return True
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(n)
