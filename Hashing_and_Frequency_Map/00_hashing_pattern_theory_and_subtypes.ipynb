# Hashing & Frequency Map — Theory and Subtypes

Hashing is one of the most powerful techniques in Data Structures.

It allows us to store and retrieve data in constant time.

This pattern is heavily used in:

- counting problems
- duplicate detection
- frequency tracking
- prefix sum optimizations

# What is Hashing?

Hashing means storing data using a key → value mapping.

In Python, this is done using:

- dictionary (dict)
- set

```python
# basic examples

freq = {}
freq["a"] = 1

seen = set()
seen.add(10)

print(freq, seen)
```

# Why Hashing?

Without hashing:

search → O(n)

With hashing:

search → O(1)

This is a huge optimization.

# Common Use Cases

Hashing is used for:

- frequency counting
- checking duplicates
- storing visited elements
- fast lookup
- grouping data

# Subtype 1 — Frequency Map

Track how many times each element appears.

Example:
"aabcc"

a → 2  
b → 1  
c → 2

```python
from collections import Counter

s = "aabcc"
print(Counter(s))
```

# Subtype 2 — Set Based Problems

Used when we only care about existence.

Example:
check duplicates

```python
nums = [1, 2, 3, 1]

seen = set()

for num in nums:
    if num in seen:
        print("duplicate found")
    seen.add(num)
```

# Subtype 3 — Prefix + Hashmap

Used for advanced problems like:

- subarray sum = k
- longest subarray with condition
- cumulative logic

This will be very important later.

# Complexity

Time:
O(1) average for insert, search

Space:
O(n)
