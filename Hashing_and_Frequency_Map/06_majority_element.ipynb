# Hashing Problem 6 — Majority Element

## Problem Statement

Given an array `nums`,
return the element that appears more than `n // 2` times.

You may assume that the majority element always exists.

Example:

nums = [3, 2, 3]

Output:
3

# Brute Force Approach

For every element,
count how many times it appears.

If frequency > n // 2,
return that element.

```python
def majority_element_bruteforce(nums):
    n = len(nums)

    for i in range(n):
        count = 0

        for j in range(n):
            if nums[i] == nums[j]:
                count += 1

        if count > n // 2:
            return nums[i]


print(majority_element_bruteforce([3, 2, 3]))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Because every element is counted repeatedly.

# Optimal Idea — Frequency Map

Store frequencies using a hashmap.

As soon as any frequency becomes greater than n // 2,
return that element.

# Why This Works

Hashmap updates and lookups are O(1).

So counting becomes very efficient.

# Dry Run

nums = [3, 2, 3]

Frequency map updates:

3 → 1
2 → 1
3 → 2

n // 2 = 1

3 appears more than 1 time.

```python
class Solution:
    def majorityElement(self, nums):
        freq = {}
        limit = len(nums) // 2

        for num in nums:
            freq[num] = freq.get(num, 0) + 1

            if freq[num] > limit:
                return num
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(n)
