# Hashing Problem 22 — Longest Subarray With Sum Zero

## Problem Statement

Given an array `nums`,
find the length of the longest subarray
with sum equal to 0.

Example:

nums = [15,-2,2,-8,1,7,10,23]

Output:
5

# Brute Force Approach

Generate every possible subarray.

Calculate its sum.

If sum becomes 0,
update maximum length.

```python
def longest_zero_sum_bruteforce(nums):
    longest = 0

    for i in range(len(nums)):
        current_sum = 0

        for j in range(i, len(nums)):
            current_sum += nums[j]

            if current_sum == 0:
                longest = max(longest, j - i + 1)

    return longest


print(longest_zero_sum_bruteforce(
    [15,-2,2,-8,1,7,10,23]
))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Too slow for large arrays.

# Optimal Idea — Prefix Sum Hashmap

If the same prefix sum appears again,

the subarray between them
has sum equal to 0.

# Why This Works

Suppose:

prefix_sum at index i = 10
#
prefix_sum at index j = 10

Then:

sum between i+1 and j = 0

# Most Important Trick

Store ONLY the first occurrence
of a prefix sum.

Why?

Because earlier index gives longer subarray.

# Dry Run

nums = [15,-2,2,-8,1,7,10,23]

Prefix sums:

15
13
15
7
8
15

Prefix sum 15 repeated.

Longest zero-sum subarray found.

```python
class Solution:
    def maxLen(self, nums):
        prefix_index = {}

        current_sum = 0
        longest = 0

        for i in range(len(nums)):
            current_sum += nums[i]

            if current_sum == 0:
                longest = i + 1

            if current_sum in prefix_index:
                length = i - prefix_index[current_sum]
                longest = max(longest, length)
            else:
                prefix_index[current_sum] = i

        return longest
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(n)
