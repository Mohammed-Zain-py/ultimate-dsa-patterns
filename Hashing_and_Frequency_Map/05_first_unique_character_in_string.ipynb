# Hashing Problem 5 — First Unique Character in a String

## Problem Statement

Given a string `s`,
return the index of the first non-repeating character.

If it does not exist, return -1.

Example:

s = "leetcode"

Output:
0

# Brute Force Approach

For every character,
count how many times it appears in the string.

The first character with frequency 1
is the answer.

```python
def first_unique_bruteforce(s):
    for i in range(len(s)):
        count = 0

        for j in range(len(s)):
            if s[i] == s[j]:
                count += 1

        if count == 1:
            return i

    return -1


print(first_unique_bruteforce("leetcode"))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Because for every character,
we scan the string again.

# Optimal Idea — Frequency Map

First:
store frequencies of all characters.

Then:
scan again and find the first character
with frequency 1.

# Why This Works

Hashmap lookup is O(1).

So frequency checking becomes extremely fast.

# Dry Run

s = "leetcode"

Frequency map:

l → 1
e → 3
t → 1
c → 1
o → 1
d → 1

First frequency 1 character:
l

Index = 0

```python
class Solution:
    def firstUniqChar(self, s):
        freq = {}

        for ch in s:
            freq[ch] = freq.get(ch, 0) + 1

        for i in range(len(s)):
            if freq[s[i]] == 1:
                return i

        return -1
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(n)

# Pattern Takeaway

Classic frequency-map interview problem.
