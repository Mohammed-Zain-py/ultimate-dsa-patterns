# Hashing Problem 11 — Count Pairs With Given Sum

## Problem Statement

Given an array `nums`
and an integer `target`,
count the number of pairs
whose sum equals `target`.

Example:

nums = [1, 5, 7, -1, 5]
target = 6

Output:
3

# Brute Force Approach

Check every possible pair.

If pair sum equals target,
increase count.

```python
def count_pairs_bruteforce(nums, target):
    count = 0

    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] + nums[j] == target:
                count += 1

    return count


print(count_pairs_bruteforce([1, 5, 7, -1, 5], 6))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Because every pair is checked.

# Optimal Idea — Frequency Lookup

For every number:

required = target - current number

Check how many times required
has already appeared.

# Why This Works

Instead of searching the array repeatedly,

we store frequencies of previously seen numbers.

# Dry Run

nums = [1,5,7,-1,5]
target = 6

| current | required | pairs added |
|---------|----------|-------------|
| 1 | 5 | 0 |
| 5 | 1 | 1 |
| 7 | -1 | 0 |
| -1 | 7 | 1 |
| 5 | 1 | 1 |

Total = 3

```python
class Solution:
    def countPairs(self, nums, target):
        freq = {}
        count = 0

        for num in nums:
            required = target - num

            if required in freq:
                count += freq[required]

            freq[num] = freq.get(num, 0) + 1

        return count
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(n)
