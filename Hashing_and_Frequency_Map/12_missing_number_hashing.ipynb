# Hashing Problem 12 — Missing Number

## Problem Statement

Given an array `nums`
containing `n` distinct numbers
in the range `[0, n]`,

return the missing number.

Example:

nums = [3, 0, 1]

Output:
2

# Brute Force Approach

For every number from 0 to n,
search whether it exists in the array.

If not found,
that is the missing number.

```python
def missing_number_bruteforce(nums):
    n = len(nums)

    for number in range(n + 1):
        found = False

        for num in nums:
            if num == number:
                found = True
                break

        if not found:
            return number


print(missing_number_bruteforce([3, 0, 1]))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Because every number is searched repeatedly.

# Optimal Idea — Hash Set Lookup

Store all numbers in a set.

Then check which number from 0 to n
is missing.

# Why This Works

Set lookup is O(1).

So existence checking becomes very fast.

# Dry Run

nums = [3,0,1]

Set:

{0,1,3}

Check:

0 → exists
1 → exists
2 → missing

```python
class Solution:
    def missingNumber(self, nums):
        seen = set(nums)
        n = len(nums)

        for number in range(n + 1):
            if number not in seen:
                return number
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(n)
