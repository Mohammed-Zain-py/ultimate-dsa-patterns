# Hashing Problem 19 — First Missing Positive

## Problem Statement

Given an unsorted integer array `nums`,
return the smallest missing positive integer.

Example:

nums = [3,4,-1,1]

Output:
2

# Brute Force Approach

Start checking from 1 upward.

For every number,
search whether it exists in the array.

The first missing number is the answer.

```python
def first_missing_positive_bruteforce(nums):
    current = 1

    while True:
        found = False

        for num in nums:
            if num == current:
                found = True
                break

        if not found:
            return current

        current += 1


print(first_missing_positive_bruteforce([3,4,-1,1]))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Because every number is searched repeatedly.

# Better Idea — Hash Set

Store all positive numbers in a set.

Then check from 1 upward
to find the first missing value.

# Why This Works

Set lookup is O(1).

So existence checking becomes fast.

# Dry Run

nums = [3,4,-1,1]

Positive set:

{1,3,4}

Check:

1 → exists
2 → missing

```python
class Solution:
    def firstMissingPositive(self, nums):
        positive = set()

        for num in nums:
            if num > 0:
                positive.add(num)

        current = 1

        while True:
            if current not in positive:
                return current

            current += 1
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Hash Set:
- Time → O(n)
- Space → O(n)

# Important Note

There is a harder optimal approach:

Cyclic Sort / Index Placement

with:
- Time → O(n)
- Space → O(1)
