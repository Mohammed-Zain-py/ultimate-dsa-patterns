# Hashing Problem 9 — Frequency Counter Basics

This notebook focuses on mastering
manual frequency map logic.

This is one of the most important skills
for hashmap problems.

# Why Frequency Counting Matters

Frequency maps are used in:

- anagrams
- duplicates
- majority element
- substring problems
- counting problems
- sliding window

# Basic Frequency Counting Logic

Core formula:

freq[element] = freq.get(element, 0) + 1

```python
nums = [1, 2, 2, 3, 1, 2]

freq = {}

for num in nums:
    freq[num] = freq.get(num, 0) + 1

print(freq)
```

# Character Frequency Counting

The same logic works for strings.

```python
s = "banana"

freq = {}

for ch in s:
    freq[ch] = freq.get(ch, 0) + 1

print(freq)
```

# Frequency Decrement Logic

Very important for:

- sliding window
- removing elements
- dynamic frequency tracking

```python
freq = {
    "a": 3,
    "b": 2
}

freq["a"] -= 1

if freq["a"] == 0:
    del freq["a"]

print(freq)
```

# Common Beginner Mistakes

1. forgetting initialization
2. decrementing below zero
3. forgetting to delete zero frequencies
4. mixing string and integer keys
