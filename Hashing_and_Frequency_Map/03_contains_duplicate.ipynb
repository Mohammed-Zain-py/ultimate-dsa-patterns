# Hashing Problem 3 — Contains Duplicate

## Problem Statement

Given an integer array `nums`,
return `True` if any value appears at least twice.

Return `False` if every element is distinct.

Example:

nums = [1, 2, 3, 1]

Output:
True

# Brute Force Approach

Check every pair.

If any pair contains equal elements,
return True.

```python
def contains_duplicate_bruteforce(nums):
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] == nums[j]:
                return True

    return False


print(contains_duplicate_bruteforce([1, 2, 3, 1]))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Because every pair is checked.

# Optimal Idea — Use a Set

A set stores only unique elements.

If an element already exists in the set,
duplicate found.

# Why This Works

Set lookup is O(1).

So we can detect duplicates instantly.

# Dry Run

nums = [1, 2, 3, 1]

| current | set | duplicate? |
|---------|-----|------------|
| 1 | {} | no |
| 2 | {1} | no |
| 3 | {1,2} | no |
| 1 | {1,2,3} | yes |

```python
class Solution:
    def containsDuplicate(self, nums):
        seen = set()

        for num in nums:
            if num in seen:
                return True

            seen.add(num)

        return False
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(n)

# Pattern Takeaway

This is the most basic and important set/hashmap lookup problem.

Very frequently asked.
