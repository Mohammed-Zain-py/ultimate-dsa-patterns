# Hashing Problem 2 — Two Sum

## Problem Statement

Given an array `nums` and an integer `target`,
return the indices of two numbers such that they add up to `target`.

Example:

nums = [2, 7, 11, 15]
target = 9

Output:
[0, 1]

# Brute Force Approach

Check every possible pair.

If pair sum equals target,
return indices.

```python
def two_sum_bruteforce(nums, target):
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] + nums[j] == target:
                return [i, j]


print(two_sum_bruteforce([2, 7, 11, 15], 9))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Because we check every pair.

# Optimal Idea — Hashmap Lookup

For every number:

required = target - current_number

Check whether required already exists in hashmap.

# Why This Works

Instead of searching the entire array again,

we store previously seen numbers in a hashmap.

Lookup becomes O(1).

# Dry Run

nums = [2, 7, 11, 15]
target = 9

| current | required | hashmap | found? |
|---------|----------|----------|--------|
| 2 | 7 | {} | no |
| 7 | 2 | {2:0} | yes |

```python
class Solution:
    def twoSum(self, nums, target):
        lookup = {}

        for i, num in enumerate(nums):
            required = target - num

            if required in lookup:
                return [lookup[required], i]

            lookup[num] = i
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(n)
