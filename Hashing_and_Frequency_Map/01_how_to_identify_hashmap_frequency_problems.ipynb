# How to Identify Hashing & Frequency Map Problems

This is one of the most important notebooks in this module.

Goal:
Understand **when to use hashing instantly**.

# The Biggest Signal

Whenever the problem involves:

- counting
- duplicates
- frequency
- lookup
- pairs

Think:

**Can I use a hashmap?**

# Recognition Checklist

Ask these questions:

1. Do I need frequency count?
2. Do I need fast lookup?
3. Am I checking duplicates?
4. Am I grouping elements?
5. Am I storing previously seen values?

# Common Keywords

Watch for:

- count
- frequency
- duplicates
- anagram
- pairs
- sum
- distinct

# Pattern 1 — Frequency Counting

Used when:

- count occurrences
- compare strings
- anagrams

```python
from collections import Counter

s = "banana"
print(Counter(s))
```

# Pattern 2 — Set for Lookup

Used when:

- checking duplicates
- finding existence

```python
nums = [1, 2, 3, 1]

seen = set()

for num in nums:
    if num in seen:
        print("duplicate found")
    seen.add(num)
```

# Pattern 3 — Hashmap for Pair Problems

Used when:

- two sum
- complement lookup

```python
nums = [2, 7, 11, 15]
target = 9

lookup = {}

for i, num in enumerate(nums):
    if target - num in lookup:
        print("pair found")
    lookup[num] = i
```

# Pattern 4 — Prefix + Hashmap

Used for:

- subarray sum
- cumulative conditions
- advanced counting

# Final Takeaway

If a problem involves:

- counting
- matching
- tracking previous data

Hashing is almost always the answer.
