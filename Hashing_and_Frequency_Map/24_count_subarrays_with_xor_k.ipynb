# Hashing Problem 24 — Count Subarrays With XOR Equal To K

## Problem Statement

Given an array `nums`
and an integer `k`,

return the number of subarrays
whose XOR equals `k`.

Example:

nums = [4,2,2,6,4]
k = 6

Output:
4

# Brute Force Approach

Generate every possible subarray.

Calculate XOR.

If XOR equals `k`,
increase count.

```python
def count_xor_bruteforce(nums, k):
    count = 0

    for i in range(len(nums)):
        current_xor = 0

        for j in range(i, len(nums)):
            current_xor ^= nums[j]

            if current_xor == k:
                count += 1

    return count


print(count_xor_bruteforce([4,2,2,6,4], 6))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Too slow for large arrays.

# Optimal Idea — Prefix XOR Hashmap

Suppose:

prefix_xor ^ previous_prefix_xor = k

Then:

previous_prefix_xor = prefix_xor ^ k

# Why This Works

XOR property:

A ^ B = C

Then:

A ^ C = B

# Dry Run

nums = [4,2,2,6,4]
k = 6

Prefix XOR values:

4
6
4
2
6

Required lookup:

prefix_xor ^ k

```python
class Solution:
    def subarraysXor(self, nums, k):
        xor_count = {
            0: 1
        }

        current_xor = 0
        count = 0

        for num in nums:
            current_xor ^= num

            required = current_xor ^ k

            if required in xor_count:
                count += xor_count[required]

            xor_count[current_xor] = (
                xor_count.get(current_xor, 0) + 1
            )

        return count
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(n)
