# Hashing Problem 23 — Subarray Sum Divisible By K

## Problem Statement

Given an integer array `nums`
and an integer `k`,

return the number of subarrays
whose sum is divisible by `k`.

Example:

nums = [4,5,0,-2,-3,1]
k = 5

Output:
7

# Brute Force Approach

Generate every possible subarray.

Calculate its sum.

If sum % k == 0,
increase count.

```python
def subarrays_divisible_bruteforce(nums, k):
    count = 0

    for i in range(len(nums)):
        current_sum = 0

        for j in range(i, len(nums)):
            current_sum += nums[j]

            if current_sum % k == 0:
                count += 1

    return count


print(subarrays_divisible_bruteforce(
    [4,5,0,-2,-3,1],
    5
))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Too slow for large arrays.

# Optimal Idea — Prefix Remainder Hashmap

If:

(prefix_j - prefix_i) % k == 0

Then:

prefix_j % k == prefix_i % k

# Why This Works

Same remainders mean:

their difference is divisible by k.

# Dry Run

nums = [4,5,0,-2,-3,1]

Prefix remainders:

4
4
4
2
4
0

Remainder 4 repeats multiple times.

Multiple valid subarrays exist.

```python
class Solution:
    def subarraysDivByK(self, nums, k):
        remainder_count = {
            0: 1
        }

        current_sum = 0
        count = 0

        for num in nums:
            current_sum += num

            remainder = current_sum % k

            if remainder in remainder_count:
                count += remainder_count[remainder]

            remainder_count[remainder] = (
                remainder_count.get(remainder, 0) + 1
            )

        return count
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n)
- Space → O(k)
