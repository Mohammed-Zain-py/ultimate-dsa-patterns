# Hashing Problem 7 — Intersection of Two Arrays

## Problem Statement

Given two integer arrays `nums1` and `nums2`,
return their intersection.

Each element in the result must be unique.

Example:

nums1 = [1, 2, 2, 1]
nums2 = [2, 2]

Output:
[2]

# Brute Force Approach

Compare every element of nums1
with every element of nums2.

Store common unique elements.

```python
def intersection_bruteforce(nums1, nums2):
    result = []

    for i in range(len(nums1)):
        for j in range(len(nums2)):
            if nums1[i] == nums2[j] and nums1[i] not in result:
                result.append(nums1[i])

    return result


print(intersection_bruteforce([1, 2, 2, 1], [2, 2]))
```

# Complexity of Brute Force

Time Complexity: O(n * m)

Because every pair is checked.

# Optimal Idea — Use Set Lookup

Store elements of one array in a set.

Then check whether elements of the second array exist in that set.

# Why This Works

Set lookup is O(1).

So searching becomes very fast.

# Dry Run

nums1 = [1, 2, 2, 1]
nums2 = [2, 2]

Set from nums1:

{1, 2}

Check nums2:

2 exists → add to answer

```python
class Solution:
    def intersection(self, nums1, nums2):
        seen = set(nums1)
        result = set()

        for num in nums2:
            if num in seen:
                result.add(num)

        return list(result)
```

# Complexity Comparison

Brute Force:
- Time → O(n * m)

Optimal:
- Time → O(n + m)
- Space → O(n)

# Pattern Takeaway

Classic set lookup interview problem.

Very important for hashmap fundamentals.
