# Hashing Problem 10 — Sort Characters By Frequency

## Problem Statement

Given a string `s`,
sort it in decreasing order
based on character frequency.

Example:

s = "tree"

Output:
"eert"

# Brute Force Approach

Count frequencies manually.

Repeatedly find the maximum frequency character,
append it,
then remove it.

```python
def frequency_sort_bruteforce(s):
    freq = {}

    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1

    result = ""

    while freq:
        max_char = ""
        max_freq = 0

        for ch in freq:
            if freq[ch] > max_freq:
                max_freq = freq[ch]
                max_char = ch

        result += max_char * max_freq

        del freq[max_char]

    return result


print(frequency_sort_bruteforce("tree"))
```

# Complexity of Brute Force

Time Complexity: O(n²)

Because we repeatedly search for maximum frequency.

# Better Idea — Frequency + Sorting

Store frequencies first.

Then sort characters based on frequency.

# Key Observation

We do NOT sort the string itself.

We sort based on:

frequency values.

# Dry Run

s = "tree"

Frequency map:

t → 1
r → 1
e → 2

Sorted order:

e → 2
t → 1
r → 1

```python
class Solution:
    def frequencySort(self, s):
        freq = {}

        for ch in s:
            freq[ch] = freq.get(ch, 0) + 1

        chars = list(freq.items())

        chars.sort(key=lambda x: x[1], reverse=True)

        result = ""

        for ch, count in chars:
            result += ch * count

        return result
```

# Complexity Comparison

Brute Force:
- Time → O(n²)

Optimal:
- Time → O(n log n)
- Space → O(n)
