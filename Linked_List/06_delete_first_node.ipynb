# Delete First Node

## Problem Statement


Delete first node.


Input


```text
head = [10,20,30]
```


Output


```text
20 → 30
```


Explanation


```text
Move head forward.

Old head disappears.
```

## Problem Explanation


Original


```text
head
 ↓

10 → 20 → 30
```


Delete


```text
10
```


Update


```text
head = head.next
```


Result


```text
head
 ↓

20 → 30
```


Observation


No traversal required.

## Brute Force


Convert to list.

Remove first.

Rebuild.


Simple.

Not actual Linked List.

````python
values = [10, 20, 30]

values.pop(0)

print(values)
````

## Complexity


```text
Time  → O(N)

Space → O(N)
```

## Optimal Approach


Move head.


```text
head = head.next
```


Done.


No pointer reconnect.

## Dry Run


Initial


```text
head
 ↓

10 → 20 → 30
```


Move


```text
head
 ↓

20 → 30
```


Old node removed.

````python
class Solution:

    def deleteFirst(self, head):

        if not head:
            return None

        return head.next
````

## Complexity


```text
Time  → O(1)

Space → O(1)
```

## Comparison


| Delete | Traversal |
|---|---|
| First | No |
| End | Yes |

## Takeaway


```text
Delete First

↓

Move Head

↓

Done
```


Deleting first node is easier than inserting in middle.
