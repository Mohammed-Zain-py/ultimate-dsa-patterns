# Remove Duplicates from Sorted Linked List

## Problem Statement


Remove duplicate nodes.


Keep one occurrence.


Input


```text
1 → 1 → 2 → 3 → 3
```


Output


```text
1 → 2 → 3
```


Explanation


```text
List is sorted.

Duplicates appear together.
```

## Problem Explanation


Original


```text
1 → 1 → 2 → 3 → 3
```


Remove


```text
extra repeated nodes
```


Final


```text
1 → 2 → 3
```


Observation


Sorted order helps.


Equal values stay adjacent.


So:


```text
current

vs

current.next
```

## Brute Force


Store values.


Remove duplicates.


Build again.


Extra work.

````python
values = [1, 1, 2, 3, 3]

result = []

for value in values:

    if value not in result:

        result.append(value)

print(result)
````

## Complexity


```text
Time  → O(N)

Space → O(N)
```

## Optimal Approach


Traverse.


Compare:


```text
current.data

==

current.next.data
```


Duplicate


```text
Skip next
```


Else


```text
Move
```

## Dry Run


Initial


```text
1 → 1 → 2 → 3 → 3
↑
```


Skip


```text
1 ─────→ 2
```


Move


```text
1 → 2 → 3 → 3
```


Skip


```text
3 ─────→ NULL
```


Final


```text
1 → 2 → 3
```

````python
class Solution:

    def deleteDuplicates(self, head):

        current = head

        while current and current.next:

            if current.data == current.next.data:

                current.next = current.next.next

            else:

                current = current.next

        return head
````

## Complexity


```text
Time  → O(N)

Space → O(1)
```
