# Middle of Linked List

## Problem Statement


Find middle node.


If two middle nodes exist:

Return second middle.


Input


```text
head = [1,2,3,4,5]
```


Output


```text
3
```


Input


```text
head = [1,2,3,4,5,6]
```


Output


```text
4
```


Explanation


```text
Move pointers
at different speeds
```

## Optimal Approach


Use two pointers.


```text
slow += 1

fast += 2
```


Stop:


```text
fast == NULL
```


Return:


```text
slow
```

````python
class Solution:

    def middleNode(self, head):

        slow = head

        fast = head

        while fast and fast.next:

            slow = slow.next

            fast = fast.next.next

        return slow
````

## Complexity


```text
Time  → O(N)

Space → O(1)
```
