# Merge Two Sorted Lists

## Problem Statement


Merge two sorted Linked Lists.


Input


```text
list1 = 1 → 3 → 5

list2 = 2 → 4 → 6
```


Output


```text
1 → 2 → 3 → 4 → 5 → 6
```


Explanation


```text
Compare nodes.

Attach smaller.

Move forward.
```

## Optimal Approach


Create dummy.


Compare heads.


Attach smaller.


Move attached pointer.

````python
class Solution:

    def mergeTwoLists(self, list1, list2):

        dummy = Node(0)

        tail = dummy

        while list1 and list2:

            if list1.data <= list2.data:

                tail.next = list1

                list1 = list1.next

            else:

                tail.next = list2

                list2 = list2.next

            tail = tail.next

        tail.next = list1 or list2

        return dummy.next
````

## Complexity


```text

Space → O(1)
```
