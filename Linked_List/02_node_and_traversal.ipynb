# Node and Traversal

## Problem Statement


Build a Linked List.

Start from head.

Visit every node.

Stop correctly.


Input


```text
head = [10,20,30]
```


Output


```text
10
20
30
```


Explanation


```text
Visit each node once
using next reference
```

## Problem Explanation


Node Structure


```text
┌─────────────┐
│ data | next │
└─────────────┘
```


Example


```text
head
 ↓

[10|•] → [20|•] → [30|NULL]
```


Traversal means:


```text
Current Node

↓

Print

↓

Move to next

↓

Repeat
```

## Brute Force


Store values in Python list.


Traverse normally.


Used only to compare.

````python
values = [10, 20, 30]

for value in values:
    print(value)
````

## Complexity


```text
Time  → O(N)

Space → O(N)
```

## Optimal Approach


Create nodes.

Connect nodes.

Move until NULL.


Flow


```text
head
 ↓

current

↓

current.next

↓

NULL
```

## Dry Run


Initial


```text
head
 ↓
10 → 20 → 30
```


Iteration 1


```text
current = 10
print(10)
```


Iteration 2


```text
current = 20
print(20)
```


Iteration 3


```text
current = 30
print(30)
```


Iteration 4


```text
current = NULL

STOP
```

````python
class Node:

    def __init__(self, data):
        self.data = data
        self.next = None


class Solution:

    def traverse(self, head):

        current = head

        while current:

            print(current.data)

            current = current.next
````

## Complexity


```text
Time  → O(N)

Space → O(1)
```

## Comparison


| Method | Extra Space |
|---|---|
| Python List | O(N) |
| Traversal | O(1) |

## Takeaway


```text
head
 ↓
current
 ↓
next
 ↓
NULL
```


Traversal is the base operation.

Almost every Linked List interview question starts here.
