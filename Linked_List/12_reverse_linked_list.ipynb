# Reverse Linked List

## Problem Statement


Reverse the entire Linked List.


Input


```text
1 → 2 → 3 → 4 → 5
```


Output


```text
5 → 4 → 3 → 2 → 1
```


Explanation


```text
Reverse connections.

Do NOT reverse values.
```

## Problem Explanation


Original


```text
1 → 2 → 3 → 4 → 5
```


Wrong


```text
5 → 4 → 3 → 2 → 1

(values swapped)
```


Correct


```text
1 ← 2 ← 3 ← 4 ← 5
```


We change:


```text
next direction
```


Need 3 pointers:


```text
prev

current

next_node
```

## Brute Force


Store values.

Reverse.

Rebuild.


Works.

Not interview expectation.

````python
values = [1, 2, 3, 4, 5]

values.reverse()

print(values)
````

## Complexity


```text
Time  → O(N)

Space → O(N)
```

## Optimal Approach


Maintain:


```text
prev

current

next
```


Pointer Flip


```text
current.next

↓

prev
```


Move all pointers.

## Dry Run


Initial


```text
NULL ← 1 → 2 → 3 → 4 → 5
       ↑
    current
```


Step 1


```text
NULL ← 1   2 → 3 → 4 → 5
       ↑
      prev
```


Step 2


```text
NULL ← 1 ← 2 → 3 → 4 → 5
           ↑
         prev
```


Final


```text
NULL ← 1 ← 2 ← 3 ← 4 ← 5
```


Return


```text
5
```

````python
class Solution:

    def reverseList(self, head):

        prev = None

        current = head

        while current:

            next_node = current.next

            current.next = prev

            prev = current

            current = next_node

        return prev
````

## Complexity


```text
Time  → O(N)

Space → O(1)
```
