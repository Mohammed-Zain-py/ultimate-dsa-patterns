## Dry Run


Detect


```text
1 → 2 → 3 → 4 → 5
        ↑       ↓
        └───────┘
```


Meet


```text
slow == fast
```


Reset


```text
one → head
```


Move


```text
step

↓

step
```


Return


```text
3
```

````python
class Solution:

    def detectCycle(self, head):

        slow = head

        fast = head

        while fast and fast.next:

            slow = slow.next

            fast = fast.next.next

            if slow == fast:

                break

        else:

            return None

        start = head

        while start != slow:

            start = start.next

            slow = slow.next

        return start
````
