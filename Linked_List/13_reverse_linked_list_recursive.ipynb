# Reverse Linked List (Recursive)

## Problem Statement


Reverse Linked List.


But use recursion.


Input


```text
1 → 2 → 3 → 4 → 5
```


Output


```text
5 → 4 → 3 → 2 → 1
```


Explanation


```text
Go till end.

Reverse while returning.
```

## Problem Explanation


Recursive thinking:


```text
Reverse smaller problem

↓

Attach current later
```


Visual


```text
1 → 2 → 3 → 4 → 5

Call

↓

2 → 3 → 4 → 5

Call

↓

3 → 4 → 5
```


Then unwind


```text
5 ← 4 ← 3 ← 2 ← 1
```


Observation


Work happens while returning.

## Brute Force


Use stack.


Pop.


Reconnect.


Extra memory.

````python
values = [1, 2, 3, 4, 5]

stack = []

for value in values:
    stack.append(value)

while stack:
    print(stack.pop())
````

## Complexity


```text
Time  → O(N)

Space → O(N)
```

## Optimal Approach


Base Case


```text
head == NULL

OR

head.next == NULL
```


Recursive Step


```text
Reverse Remaining

↓

Attach Current
```

## Dry Run


Calls


```text
1

↓

2

↓

3

↓

4

↓

5
```


Return


```text
5

↓

5 ← 4

↓

5 ← 4 ← 3

↓

5 ← 4 ← 3 ← 2

↓

5 ← 4 ← 3 ← 2 ← 1
```

````python
class Solution:

    def reverseList(self, head):

        if not head or not head.next:
            return head

        new_head = self.reverseList(head.next)

        head.next.next = head

        head.next = None

        return new_head
````

## Complexity


```text
Time  → O(N)

Space → O(N)
```

## Comparison


| Method | Space |
|---|---|
| Iterative | O(1) |
| Recursive | O(N) |

## Takeaway


```text
Go Deep

↓

Reach End

```
