# Delete Last Node

## Problem Statement


Delete the last node.


Input


```text
head = [10,20,30]
```


Output


```text
10 → 20
```


Explanation


```text
Move to second last node.

Break connection.
```

## Problem Explanation


Original


```text
head
 ↓

10 → 20 → 30
```


Delete


```text
30
```


Result


```text
head
 ↓

10 → 20 → NULL
```


Important


Do NOT move to last.

Move to second last.


Reason


```text
Need access to next pointer
```

## Brute Force


Convert.

Remove last.

Rebuild.


Easy.

Not actual Linked List.

````python
values = [10, 20, 30]

values.pop()

print(values)
````

## Complexity


```text
Time  → O(N)

Space → O(N)
```

## Optimal Approach


Traverse until:


```text
current.next.next == NULL
```


Disconnect


```text
current.next = NULL
```

## Dry Run


Initial


```text
head
 ↓

10 → 20 → 30
```


Move


```text
head
 ↓

10 → 20
     ↑
```


Disconnect


```text
10 → 20 → NULL
```

````python
class Solution:

    def deleteLast(self, head):

        if not head:
            return None

        if not head.next:
            return None

        current = head

        while current.next.next:

            current = current.next

        current.next = None

        return head
````

## Complexity


```text
Time  → O(N)

Space → O(1)
```

## Comparison


| Delete | Traversal |
|---|---|
| First | No |
| Last | Yes |

## Takeaway


```text
Delete First

↓

Move Head


Delete Last

↓

Find Previous
```


Deleting from end requires thinking one step earlier.
