# Linked List — Theory and Subtypes

## What is a Linked List?

Linked List is a linear data structure.

Elements are connected using references.

Elements are NOT stored continuously in memory.


Array

```text
[10][20][30][40]
```


Linked List

```text
[10|•] → [20|•] → [30|•] → [40|NULL]
```


Each block is called:

• Node


Each Node stores:

• Data  
• Reference to next node

## Why Do We Need Linked List?

Imagine inserting at beginning.


Array

```text
[10][20][30][40]

Insert 5

↓

[5][10][20][30][40]
```

Every element shifts.


Time Complexity

```text
O(N)
```


Now Linked List


```text
5 → 10 → 20 → 30 → 40
```

Only connection changes.


Time Complexity

```text
O(1)
```


Core Idea:

Trade fast access for flexible insertion.

## Array vs Linked List

| Feature | Array | Linked List |
|---|---|---|
| Memory | Continuous | Dynamic |
| Access | O(1) | O(N) |
| Insert Front | O(N) | O(1) |
| Delete Front | O(N) | O(1) |


Quick Rule

```text
Frequent Insert/Delete
↓

Think Linked List
```

## Anatomy of a Node

Visual


```text
┌─────────────┐
│ data | next │
└─────────────┘
```


Example


```text
┌─────────┐
│  15 | • │
└─────────┘
```


Meaning

```text
data = 15

next = address of next node
```

## Traversal

Linked Lists cannot jump.


This exists in array

```python
arr[3]
```


This does NOT exist in Linked List.


Movement


```text
head
 ↓
10 → 20 → 30
 ↓
next
```


Continue until


```text
NULL
```

## Type 1 — Singly Linked List


```text
10 → 20 → 30 → NULL
```


Movement

```text
Forward Only
```


Interview Frequency

★★★★★

## Type 2 — Doubly Linked List


```text
NULL ← 10 ⇄ 20 ⇄ 30 → NULL
```


Movement

```text
Forward + Backward
```


Extra Memory

```text
prev pointer
```


Interview Frequency

★★★★☆

## Type 3 — Circular Linked List


```text
      ┌──────────────┐
      ↓              │
10 → 20 → 30 → 40 ───┘
```


Notice:

Last node does NOT point to NULL.


Instead:


```text
tail.next = head
```
