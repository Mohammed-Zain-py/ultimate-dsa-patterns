# Insert Node at End

## Problem Statement


Insert a new node at the end.


Input


```text
head = [10,20,30]

value = 40
```


Output


```text
10 → 20 → 30 → 40
```


Explanation


```text
Traverse until last node.

Attach new node.
```

## Problem Explanation


Current List


```text
head
 ↓

10 → 20 → 30 → NULL
```


Create


```text
new = 40
```


Move


```text
10 → 20 → 30
          ↑
```


Connect


```text
10 → 20 → 30 → 40
```


Observation


Beginning insertion updates head.

End insertion requires traversal.

## Brute Force


Convert to list.

Append.

Rebuild.


Not actual Linked List.

````python
values = [10, 20, 30]

values.append(40)

print(values)
````

## Complexity


```text
Time  → O(N)

Space → O(N)
```

## Optimal Approach


Move to tail.


Attach node.


```text
current.next = new
```


Final


```text
head
 ↓

10 → 20 → 30 → 40
```

## Dry Run


Initial


```text
10 → 20 → 30
↑
```


Move


```text
10 → 20 → 30
     ↑
```


Move


```text
10 → 20 → 30
          ↑
```


Attach


```text
10 → 20 → 30 → 40
```

````python
class Node:

    def __init__(self, data):
        self.data = data
        self.next = None


class Solution:

    def insertAtEnd(self, head, value):

        new = Node(value)

        if not head:
            return new

        current = head

        while current.next:
            current = current.next

        current.next = new

        return head
````

## Complexity


```text
Time  → O(N)

Space → O(1)
```

## Comparison


| Operation | Beginning | End |
|---|---|---|
| Traversal | No | Yes |
| Complexity | O(1) | O(N) |

## Takeaway


```text
Insert Front

↓

Change Head


Insert End

↓

Find Tail
```


Small difference.
