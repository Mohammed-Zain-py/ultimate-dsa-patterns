# Find Length of Linked List

## Problem Statement


Find total number of nodes.


Input


```text
head = [10,20,30,40]
```


Output


```text
4
```


Explanation


```text
Visit every node.

Count.
```

## Problem Explanation


Example


```text
head
 ↓

10 → 20 → 30 → 40
```


Count


```text
1 → 2 → 3 → 4
```


Stop


```text
NULL
```


Observation


Unlike arrays:

```text
No length property
```


Must traverse.

## Brute Force


Convert.

Call len().


Simple.

Not actual Linked List.

````python
values = [10, 20, 30, 40]

print(len(values))
````

## Complexity


```text
Time  → O(N)

Space → O(N)
```

## Optimal Approach


Start:


```text
count = 0
```


For every node:


```text
count += 1
```


Move:


```text
current = current.next
```


Stop at NULL.

## Dry Run


Initial


```text
10 → 20 → 30 → 40
↑
count = 0
```


Move


```text
10 ✓ → count = 1
20 ✓ → count = 2
30 ✓ → count = 3
40 ✓ → count = 4
```


Return


```text
4
```

````python
class Solution:

    def length(self, head):

        count = 0

        current = head

        while current:

            count += 1

            current = current.next

        return count
````

## Complexity


```text
Time  → O(N)

Space → O(1)
```

## Comparison


| Structure | Length |
|---|---|
| Array | O(1) |
| Linked List | O(N) |

## Takeaway


```text
Traverse

↓

Count

↓

Return
```
