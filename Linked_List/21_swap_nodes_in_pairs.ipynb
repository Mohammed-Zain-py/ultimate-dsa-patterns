# Swap Nodes in Pairs

## Problem Statement


Swap every two adjacent nodes.


Do NOT swap values.


Input


```text
1 → 2 → 3 → 4
```


Output


```text
2 → 1 → 4 → 3
```


Explanation


```text
Change connections.

Keep node values unchanged.
```

## Problem Explanation


Original


```text
1 → 2 → 3 → 4
```


Wrong


```text
2 → 1 → 4 → 3

(values swapped)
```


Correct


```text
2 ← 1   4 ← 3
```


One Pair


```text
first

second
```


Goal


```text
second → first
```

## Brute Force


Store values.


Swap values.


Rebuild.


Works.

Question violated.

````python
values = [1, 2, 3, 4]

for i in range(0, len(values) - 1, 2):

    values[i], values[i + 1] = values[i + 1], values[i]

print(values)
````

## Complexity


```text
Time  → O(N)

Space → O(N)
```

## Optimal Approach


Use dummy.


For each pair:


```text
prev

↓

first

↓

second
```


Reconnect


```text
prev → second

first → next

second → first
```

## Dry Run


Initial


```text
D → 1 → 2 → 3 → 4
```


Swap


```text
D → 2 → 1 → 3 → 4
```


Swap Again


```text
D → 2 → 1 → 4 → 3
```


Return


```text
2 → 1 → 4 → 3
```

````python
class Node:

    def __init__(self, data=0, next=None):

        self.data = data

        self.next = next


class Solution:

    def swapPairs(self, head):

        dummy = Node(0, head)

        prev = dummy

        while prev.next and prev.next.next:

            first = prev.next

            second = first.next

            first.next = second.next

            second.next = first

            prev.next = second

            prev = first

        return dummy.next
````

## Complexity


```text
Time  → O(N)

Space → O(1)
```
