# Intersection of Two Linked Lists

## Problem Statement


Find the node where two Linked Lists intersect.


Input


```text
A

1 → 2 ↘
        8 → 9
3 → 4 ↗

B
```


Output


```text
8
```


Explanation


```text
Intersection means

same node

not same value
```

## Problem Explanation


Wrong Thinking


```text
8 == 8
```


Correct Thinking


```text
address(A)

==

address(B)
```


Example


```text
A

1 → 2 ↘
        8 → 9

3 → 4 ↗

B
```


Observation


After intersection:

remaining path is identical.

## Brute Force


For every node in A


Traverse B


Compare references.


Nested traversal.

````python
A = [1, 2]

B = [3, 4]

print("Compare every node")
````

## Complexity


```text
Time  → O(N × M)

Space → O(1)
```

## Optimal Approach


Pointer A


```text
End

↓

Go to B
```


Pointer B


```text
End

↓

Go to A
```


Rule


```text
Equal distance

↓

Meet
```

## Dry Run


Start


```text
A → A → A
↑

B → B
↑
```


Switch


```text
A → B

B → A
```


Eventually


```text
Meet

↓

Intersection
```

````python
class Solution:

    def getIntersectionNode(self, headA, headB):

        a = headA

        b = headB

        while a != b:

            a = a.next if a else headB

            b = b.next if b else headA

        return a
````

## Complexity


```text
Time  → O(N+M)

Space → O(1)
```
