# Search in Linked List

## Problem Statement


Search whether a value exists.


Input


```text
head = [10,20,30,40]

target = 30
```


Output


```text
True
```


Explanation


```text
Visit every node.

Stop if found.
```

## Problem Explanation


Array


```text
Can Jump
```


Linked List


```text
Must Traverse
```


Example


```text
head
 ↓

10 → 20 → 30 → 40
```


Search


```text
target = 30
```


Traversal


```text
10

↓

20

↓

30

✓ Found
```


Observation


No direct access.

## Brute Force


Convert.

Search.

Return.


Simple.

Not actual Linked List.

````python
values = [10, 20, 30, 40]

print(30 in values)
````

## Complexity


```text
Time  → O(N)

Space → O(N)
```

## Optimal Approach


Start from head.


Compare.


Move.


Stop when:

```text
current == None
```

## Dry Run


Search


```text
target = 30
```


Step 1


```text
10 ✗
```


Step 2


```text
20 ✗
```


Step 3


```text
30 ✓
```


Return


```text
True
```

````python
class Solution:

    def search(self, head, target):

        current = head

        while current:

            if current.data == target:
                return True

            current = current.next

        return False
````

## Complexity


```text
Time  → O(N)

Space → O(1)
```

## Comparison


| Structure | Search |
|---|---|
| Array | O(1) |
| Linked List | O(N) |

## Takeaway


```text
Search

↓

Traverse

↓

Compare

↓

Stop
```


Linked Lists trade search speed for flexible updates.
