# How to Identify Linked List Problems

## Why Learn Identification?

Most interview failures happen before coding.

People know algorithms.

But they fail to recognize the pattern.


Goal:

```text
Problem
   ↓
Detect Pattern
   ↓
Choose Approach
   ↓
Code
```

## Signal 1 — You See "head"

Input like:


```text
head = [1,2,3,4]
```


or


```text
Node* head
```


Usually means:

```text
Linked List
```


Interview Examples:

• Reverse List  
• Delete Node  
• Detect Cycle

## Signal 2 — No Indexing

Question says:


```text
Move through nodes
Traverse
Current
Previous
Next
```


BUT never says:


```python
arr[i]
```


Think:


```text
Pointer Movement
```

## Signal 3 — Connection Changes

Watch for words:


• Reverse  
• Reorder  
• Remove  
• Merge  
• Insert  
• Rotate  


These usually mean:


```text
Update pointers
```

## Signal 4 — Fast Slow Pointer

If question asks:


```text
Middle

Cycle

Nth from End

Palindrome
```


Think:


```text
slow → 1 step

fast → 2 steps
```

## Signal 5 — Dummy Node

Question contains:


```text
Insert

Delete

Merge
```


Common trick:


```text
dummy → head
```

## Pattern Mapping


```text
Reverse
↓
Pointer Flip


Middle
↓
Fast Slow


Merge
↓
Multiple Pointers


Cycle
↓
Floyd
```

## Real Interview Examples


Question:

```text
Remove 5th node from end
```

Pattern:

```text
Fast Slow
```


Question:

```text
Merge two sorted lists
```

Pattern:

```text
Dummy Node
```


Question:

```text
Reverse K Group
```

Pattern:

```text
Pointer Manipulation
```

## Common Wrong Thinking


Wrong

```text
Looks sequential

→ Use array
```


Correct

```text
Node movement

→ Use Linked List
```

## Takeaway


```text
head
next
reverse
merge
cycle

↓

Linked List
```


Recognition is half the solution.
