# Insert Node at Position

## Problem Statement


Insert a node at given position.


Input


```text
head = [10,20,30]

position = 2

value = 15
```


Output


```text
10 → 15 → 20 → 30
```


Explanation


```text
Move to previous node.

Reconnect pointers.
```

## Problem Explanation


Original


```text
head
 ↓

10 → 20 → 30
```


Insert


```text
15
```


Result


```text
10 → 15 → 20 → 30
```


Two links change.


```text
new.next = current.next

current.next = new
```


Important


Wrong order breaks list.

## Brute Force


Convert to list.

Insert.

Rebuild.


Easy.

Not actual Linked List.

````python
values = [10, 20, 30]

values.insert(1, 15)

print(values)
````

## Complexity


```text
Time  → O(N)

Space → O(N)
```

## Optimal Approach


Move until:


```text
position - 1
```


Reconnect carefully.


```text
new.next = current.next

current.next = new
```

## Dry Run


Initial


```text
10 → 20 → 30
↑
```


Move


```text
10 → 20 → 30
     ↑
```


Attach


```text
10 → 15 → 20 → 30
```


Stop

````python
class Node:

    def __init__(self, data):
        self.data = data
        self.next = None


class Solution:

    def insertAtPosition(self, head, position, value):

        if position == 1:

            new = Node(value)

            new.next = head

            return new

        current = head

        for _ in range(position - 2):

            current = current.next

        new = Node(value)

        new.next = current.next

        current.next = new

        return head
````

## Complexity


```text
Time  → O(N)

Space → O(1)
```

## Comparison

|---|---|
| Beginning | 1 |
| End | 1 |
| Position | 2 |
