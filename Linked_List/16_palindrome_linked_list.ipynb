# Palindrome Linked List

## Problem Statement


Check whether Linked List reads same from both ends.


Input


```text
1 → 2 → 2 → 1
```


Output


```text
True
```


Input


```text
1 → 2 → 3
```


Output


```text
False
```


Explanation


```text
Compare first half
with reversed second half
```

## Problem Explanation


Palindrome means:


```text
Forward

=

Backward
```


Example


```text
1 → 2 → 2 → 1
```


Observation


Linked Lists cannot move backward.


So:


```text
Find Middle

↓

Reverse Right Half

↓

Compare
```


Reuse previous patterns.

## Brute Force

Reverse.

## Optimal Approach


Phase 1


```text
Find Middle
```


Phase 2


```text
Reverse Right Half
```


Phase 3


```text
Compare
```


Stop on mismatch.

## Dry Run


Initial


```text
1 → 2 → 2 → 1
```


Middle


```text
1 → 2 | 2 → 1
```


Reverse Right


```text
1 → 2 ← 2 ← 1
```


Compare


```text
1 = 1 ✓

2 = 2 ✓
```


Return


```text
True
```

````python
class Solution:

    def palindrome(self, head):

        slow = fast = head

        while fast and fast.next:

            slow = slow.next

            fast = fast.next.next

        prev = None

        while slow:

            nxt = slow.next

            slow.next = prev

            prev = slow

            slow = nxt

        left = head

        right = prev

        while right:

            if left.data != right.data:
                return False

            left = left.next

            right = right.next

        return True
````

## Complexity


```text
Time  → O(N)

Space → O(1)
```
