# Delete Node at Position

## Problem Statement


Delete node at given position.


Input


```text
head = [10,20,30,40]

position = 3
```


Output


```text
10 → 20 → 40
```


Explanation


```text
Move to previous node.

Skip target node.
```

## Problem Explanation


Original


```text
head
 ↓

10 → 20 → 30 → 40
```


Delete


```text
30
```


Reconnect


```text

```


Final


```text
head
 ↓

```


Actual pointer update


```text
current.next = current.next.next
```


Important


Delete previous link.

Not current node.

## Complexity


```text
Time  → O(N)

Space → O(N)
```

## Optimal Approach


Move to:


```text
position - 1
```


Reconnect


```text
current.next

↓

current.next.next
```

## Dry Run


Initial


```text
10 → 20 → 30 → 40
↑
```


Move


```text
10 → 20 → 30 → 40
     ↑
```


Skip


```text
20 ─────→ 40
```


Result


```text
10 → 20 → 40
```

````python
class Solution:

    def deleteAtPosition(self, head, position):

        if not head:
            return None

        if position == 1:
            return head.next

        current = head

        for _ in range(position - 2):

            current = current.next

        current.next = current.next.next

        return head
````

## Complexity


```text
Time  → O(N)

Space → O(1)
```
