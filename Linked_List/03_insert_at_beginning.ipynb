# Insert Node at Beginning

## Problem Statement


Insert a new node at the beginning.


Input


```text
head = [10,20,30]

value = 5
```


Output


```text
5 → 10 → 20 → 30
```


Explanation


```text
New node becomes head.

Old head becomes next.
```

## Problem Explanation


Original


```text
head
 ↓

10 → 20 → 30
```


Create


```text
new = 5
```


Connect


```text
5 → 10 → 20 → 30
```


Update


```text
head = new
```


Observation


Insertion at beginning is the easiest
Linked List insertion.

## Brute Force


Store values in list.

Insert at front.

Not real Linked List.

````python
values = [10, 20, 30]

values.insert(0, 5)

print(values)
````

## Complexity


```text
Time  → O(N)

Space → O(N)
```

## Optimal Approach


Only update one pointer.


```text
new.next = head

head = new
```


Final


```text
head
 ↓

5 → 10 → 20 → 30
```

## Dry Run


Initial


```text
head
 ↓

10 → 20 → 30
```


Step 1


```text
new = 5
```


Step 2


```text
5 → 10 → 20 → 30
```


Step 3


```text
head = 5
```

````python
class Node:

    def __init__(self, data):
        self.data = data
        self.next = None


class Solution:

    def insertAtBeginning(self, head, value):

        new = Node(value)

        new.next = head

        return new
````

## Complexity


```text
Time  → O(1)

Space → O(1)
```

## Comparison


| Operation | Array | Linked List |
|---|---|---|
| Insert Front | O(N) | O(1) |
