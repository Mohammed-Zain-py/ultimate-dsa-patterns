# Python Prerequisites for DSA

Before starting Data Structures and Algorithms, it is very important to be comfortable with Python fundamentals.

DSA is not about memorizing code.

It is about solving problems efficiently.

For that, we must first be strong with Python basics like:

- Lists
- Strings
- Dictionaries
- Sets
- Loops
- Functions
- Indexing and Slicing

This notebook is a quick yet detailed revision of all Python concepts that are heavily used in DSA.

# 1. Lists

Lists are the most used data type in DSA.

In Python, arrays are mostly handled using lists.

We use them for:
- storing elements
- traversing data
- sorting
- applying two pointers
- sliding window
- binary search

```python
arr = [10, 20, 30, 40, 50]

print("Original List:", arr)
print("First Element:", arr[0])
print("Last Element:", arr[-1])
print("Length:", len(arr))
```

### Indexing

Indexing means accessing elements using their position.

Python uses **0-based indexing**.

This is extremely important in DSA because most problems are index-based.

```python
print(arr[1])
print(arr[3])
```

### Slicing

Slicing is used to extract a portion of the list.

Syntax:

`list[start:end]`

The end index is excluded.

```python
print(arr[1:4])
print(arr[:3])
print(arr[2:])
```

### Common List Operations

```python
nums = [1, 2, 3]

nums.append(4)
print("After append:", nums)

nums.insert(1, 10)
print("After insert:", nums)

nums.pop()
print("After pop:", nums)

nums.remove(10)
print("After remove:", nums)
```

These operations are frequently used in problems like:

- stack simulation
- queue simulation
- dynamic array manipulation
- insert / delete problems

# 2. Strings

Strings are heavily used in interview questions.

Examples:
- palindrome
- substring
- sliding window
- frequency count
- compression

```python
s = "python"

print("First character:", s[0])
print("Last character:", s[-1])
print("Substring:", s[1:4])
```

Strings are **immutable**.

This means we cannot directly change a character.

```python
word = "hello"

new_word = word[:1] + "a" + word[2:]

print(new_word)
```

# 3. Dictionaries

Dictionaries are one of the MOST IMPORTANT structures in DSA.

They are used for:

- hashing
- frequency counting
- constant time lookup
- maps

```python
freq = {}

arr = [1, 2, 1, 3, 2, 1]

for num in arr:
    freq[num] = freq.get(num, 0) + 1

print(freq)
```

This concept is extremely important for:

- Two Sum
- frequency counter
- hashmap pattern
- anagram problems

# 4. Sets

Sets are used when we want:

- unique elements
- fast lookup
- duplicate removal

```python
nums = [1, 2, 2, 3, 3, 4]

unique_nums = set(nums)

print(unique_nums)
```

# 5. Loops

Loops are the heart of problem solving.

Without strong loops, DSA becomes difficult.

```python
for i in range(5):
    print(i)
```

```python
i = 0

while i < 5:
    print(i)
    i += 1
```

Nested loops are important for brute force solutions.

```python
for i in range(3):
    for j in range(3):
        print(i, j)
```

# 6. Functions

Functions help in code reusability and cleaner logic.

```python
def find_max(arr):
    return max(arr)

print(find_max([10, 20, 5, 40]))
```
