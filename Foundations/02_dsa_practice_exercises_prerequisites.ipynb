# Practice Exercises Before Starting DSA

Before we begin learning Data Structures and Algorithms, it is important to make sure our Python basics are strong.

In DSA, most mistakes do not happen because of logic.

They happen because of weak basics like:

- incorrect loop conditions
- wrong indexing
- slicing mistakes
- edge case handling

This notebook contains practice exercises that every beginner should be comfortable with before moving ahead.

# Exercise 1: Sum of All Elements in a List

Problem:
Given a list of numbers, find the sum of all elements.

Example:
[10, 20, 30, 40]

Output:
100

```python
arr = [10, 20, 30, 40]

total = 0

for num in arr:
    total += num

print("Sum of all elements:", total)
```

This exercise helps in understanding:

- list traversal
- loop iteration
- accumulator pattern

This pattern is used in many DSA problems.

# Exercise 2: Find Maximum Element

Problem:
Find the largest element in the list.

```python
arr = [5, 12, 7, 25, 9]

maximum = arr[0]

for num in arr:
    if num > maximum:
        maximum = num

print("Maximum element:", maximum)
```

This is one of the most important basic logic-building exercises.

Many DSA optimization problems start with this thinking.

# Exercise 3: Reverse a String

Problem:
Reverse the given string.

```python
# Note this is the pythonic version we will see the actual logic further
s = "python"

reversed_string = s[::-1]

print("Reversed string:", reversed_string)
```

String reversal is extremely important for:

- palindrome problems
- two pointer pattern
- string manipulation

# Exercise 4: Count Vowels in a String

```python
s = "data structures"

count = 0

for ch in s:
    if ch.lower() in "aeiou":
        count += 1

print("Number of vowels:", count)
```

This improves confidence with:

- string traversal
- conditional logic
- membership checks

# Exercise 5: Access Elements Using Indexing

Problem:
Print all elements with their indexes.

```python
arr = [10, 20, 30, 40]

for i in range(len(arr)):
    print("Index:", i, "Value:", arr[i])
```

This is one of the MOST IMPORTANT skills for DSA.

Almost every problem depends on correct indexing.

# Exercise 6: Nested Loop Practice

Problem:
Print all possible pairs.

```python
arr = [1, 2, 3]

for i in range(len(arr)):
    for j in range(len(arr)):
        print(arr[i], arr[j])
```

Nested loops are very important because many brute force DSA solutions start from this approach.

Later we optimize them using better patterns.
