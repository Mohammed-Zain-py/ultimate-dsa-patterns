# Problem 1: Count Number of Digits

Given a number, count how many digits it contains.

Example:
12345 → 5

```python
num = 12345
count = 0

while num > 0:
    num //= 10
    count += 1

print("Number of digits:", count)
```

This problem helps in understanding integer division and loop-based number processing.

# Problem 2: Reverse a Number

Example:
1234 → 4321

```python
num = 1234
reverse_num = 0

while num > 0:
    digit = num % 10
    reverse_num = reverse_num * 10 + digit
    num //= 10

print("Reversed number:", reverse_num)
```

This logic is extremely important because it is later used in palindrome number problems.

# Problem 3: Check Palindrome Number

A number is palindrome if it reads the same from both sides.

Example:
121 → palindrome

```python
num = 121
original = num
reverse_num = 0

while num > 0:
    digit = num % 10
    reverse_num = reverse_num * 10 + digit
    num //= 10

if original == reverse_num:
    print("Palindrome number")
else:
    print("Not a palindrome")
```

This problem is frequently asked in coding rounds.

# Problem 4: Check Prime Number

A prime number is divisible only by 1 and itself.

```python
num = 17
is_prime = True

for i in range(2, num):
    if num % i == 0:
        is_prime = False
        break

if is_prime:
    print("Prime number")
else:
    print("Not a prime number")
```

This introduces divisor checking logic and early break optimization.

# Problem 5: Fibonacci Series

Print first 10 Fibonacci numbers.

```python
a, b = 0, 1

for _ in range(10):
    print(a, end=" ")
    a, b = b, a + b
```

This is one of the most commonly asked beginner logic questions.

# Problem 6: Factorial

Example:
5! = 120

```python
num = 5
fact = 1

for i in range(1, num + 1):
    fact *= i

print("Factorial:", fact)
```

Factorial is an important foundation for recursion later.
