# Basic Recursion for Problem Solving

Recursion is one of the most important concepts in DSA.

A function is said to be recursive when it calls itself.

Recursion helps us solve problems by breaking them into smaller subproblems.

Many advanced DSA topics like:

- backtracking
- trees
- divide and conquer
- dynamic programming

depend heavily on recursion.

# Important Terms

Before solving problems, let us understand two important terms.

**Base Case**  
The condition where recursion stops.

**Recursive Case**  
The part where the function calls itself.

# Example 1: Print Numbers from 1 to n

```python
def print_numbers(n):
    if n == 0:
        return

    print_numbers(n - 1)
    print(n)

print_numbers(5)
```

Here,

- `n == 0` is the base case
- `print_numbers(n - 1)` is the recursive call

This teaches how the recursive stack builds and unwinds.

# Example 2: Factorial Using Recursion

Factorial of n = n × (n-1) × (n-2) ...

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1

    return n * factorial(n - 1)

print("Factorial:", factorial(5))
```

This is one of the most common recursion interview questions.

# Example 3: Fibonacci Using Recursion

```python
def fibonacci(n):
    if n == 0:
        return 0
    if n == 1:
        return 1

    return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(6))
```

This shows how recursion can branch into multiple calls.

Later this will help in understanding recursion trees.

# Example 4: Sum of First n Natural Numbers

```python
def sum_n(n):
    if n == 0:
        return 0

    return n + sum_n(n - 1)

print(sum_n(5))
```

# How recursion works internally

Each recursive call is stored in memory.

This memory structure is called the **call stack**.

Every function waits for the smaller problem to finish first.

# Visual intuition of factorial recursion

f(n)=n⋅f(n−1)

This formula is the exact recursive relation for factorial.

Understanding recursive relations will help a lot in future DSA topics.

# Important Note

Recursion is powerful, but every recursive solution uses extra stack space.

That is why space complexity becomes important.
